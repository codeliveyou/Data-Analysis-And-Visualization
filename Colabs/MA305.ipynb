```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the workout data from the CSV file
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/superstore_final_dataset.csv", encoding='ISO-8859-1')

# Convert 'Order_Date' column to datetime format
data['Order_Date'] = pd.to_datetime(data['Order_Date'])

# Top Selling Products
top_products = data.groupby('Product_Name').Sales.sum().nlargest(10)
product_colors = ['g' if value > top_products.median() else 'r' for value in top_products]

plt.figure(figsize=(12, 7))
top_products.sort_values().plot(kind='barh', color=product_colors)

plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')

plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
