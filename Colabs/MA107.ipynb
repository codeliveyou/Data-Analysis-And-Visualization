```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load csv into a dataframe
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv')

# Filter dataframe by salary_in_usd to be exceeding $100,000
filtered_df = df[df['salary_in_usd'] > 100000]

# Gain a count of each job of this filtered dataframe
job_title_count = filtered_df['job_title'].value_counts()

# Set a threshold for what can be displayed on pie chart
threshold = .005

# Gain a sum of all the jobs which total less than 0.5% of the data set
smaller_jobs = job_title_count[job_title_count / job_title_count.sum() < threshold]
other_category = smaller_jobs.sum()

# Create an 'Other' category in the job_title_count data frame
job_title_count = job_title_count[~job_title_count.index.isin(smaller_jobs.index)]
job_title_count['Other'] = other_category

# Configure and create pie chart
plt.figure(figsize=(12,8))
plt.pie(job_title_count, labels = job_title_count.index, autopct='%1.1f%%', textprops={'fontsize': 8})

# Create title for plot
plt.title('Percentage of Occupations Exceeding $100,000 Salary')

# Display plot
plt.axis('equal')
plt.show()
```
