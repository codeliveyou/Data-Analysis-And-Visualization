```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)

# parsing dates
df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['month'] = df['date'].dt.month

# Create a histogram of the 'month' column in the DataFrame
counts, bins = np.histogram(df['month'], bins=range(1, 14))

# Plot the histogram
fig, ax = plt.subplots()
ax.bar(bins[:-1], counts, width=0.8)

# Set the title and axis labels
ax.set_title('Number of matches played per month')
ax.set_xlabel('Number of months')
ax.set_ylabel('Matches played')

# Set the x-tick labels to the month names
ax.set_xticks(range(1, 13))
ax.set_xticklabels(range(1, 13))

plt.show()
```
