```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/dataset_olympics.csv')

# Top 10 most frequent participants based on total participations in different events
top_participants_events = df['Name'].value_counts().nlargest(10)

# Top 10 most frequent participants based on total participations in Olympics
top_participants_olympics = df.groupby('Name')['Year'].nunique().nlargest(10)

# Plotting the bar plots
plt.figure(figsize=(12, 6))

# Bar plot for top participants in different events
plt.subplot(1, 2, 1)
top_participants_events.plot(kind='bar', color='blue')
plt.title('Top 10 Participants by Total Participations (Events)')
plt.xlabel('Participant')
plt.ylabel('Total Participations')

# Bar plot for top participants in Olympics
plt.subplot(1, 2, 2)
top_participants_olympics.plot(kind='bar', color='green')
plt.title('Top Participants by Total Olympics Participations')
plt.xlabel('Participant')
plt.ylabel('Total Olympics Participations')

plt.tight_layout()
plt.show()
```
