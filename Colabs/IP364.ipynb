```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral6
from bokeh.transform import factor_cmap
from bokeh.layouts import column, row
import warnings
warnings.filterwarnings('ignore')

# Read the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/amazon_customer_behavior_survey.csv")

output_notebook()

# 1. Bar chart for Purchase Frequency by age
# Group the data by age and Purchase Frequency
grouped_data = df.groupby(['age', 'Purchase_Frequency']).size().unstack(fill_value=0)

# Find the top purchase frequencies for each age group
top_frequencies_age = grouped_data.sum().nlargest(3).index.tolist()

# Prepare the data
ages = df['age'].unique().astype(str)
purchase_frequencies = df['Purchase_Frequency'].unique()
source_age = ColumnDataSource(data={'age': ages})

for freq in purchase_frequencies:
    source_age.add(grouped_data[freq].tolist(), name=freq)

# Create a bar chart
hover_age = HoverTool(tooltips=[("Age", "@age"), ("Few times a Month: ", f"@{{{purchase_frequencies[0]}}}"), ("Once a Month: ", f"@{{{purchase_frequencies[1]}}}"), ("Multiple times a week: ", f"@{{{purchase_frequencies[3]}}}")])
p_age = figure(x_range=ages, title="Purchase Frequency by Age", tools=[hover_age], height=400, width=400)
p_age.vbar(x='age', top=purchase_frequencies[0], width=0.2, source=source_age, legend_label=purchase_frequencies[0], color=Spectral6[1])
p_age.vbar(x='age', top=purchase_frequencies[1], width=0.2, source=source_age, legend_label=purchase_frequencies[1], color=Spectral6[3])
p_age.vbar(x='age', top=purchase_frequencies[3], width=0.2, source=source_age, legend_label=purchase_frequencies[3], color=Spectral6[5])

# 2. Bar chart for Purchase Frequency by Gender
# Group the data by Gender and Purchase Frequency
grouped_gender_data = df.groupby(['Gender', 'Purchase_Frequency']).size().unstack(fill_value=0)

# Prepare the data
genders = df['Gender'].unique()
source_gender = ColumnDataSource(data={'genders': genders})

for freq in purchase_frequencies:
    source_gender.add(grouped_gender_data[freq].tolist(), name=freq)

# Create a bar chart
hover_gender = HoverTool(tooltips=[("Gender", "@genders"), ("Few times a Month: ", f"@{{{purchase_frequencies[0]}}}"), ("Once a Month: ", f"@{{{purchase_frequencies[1]}}}"), ("Multiple times a week: ", f"@{{{purchase_frequencies[3]}}}")])
p_gender = figure(x_range=genders, title="Purchase Frequency by Gender", tools=[hover_gender], height=400, width=400)
p_gender.vbar(x='genders', top=purchase_frequencies[0], width=0.2, source=source_gender, legend_label=purchase_frequencies[0], color=Spectral6[1])
p_gender.vbar(x='genders', top=purchase_frequencies[1], width=0.2, source=source_gender, legend_label=purchase_frequencies[1], color=Spectral6[3])
p_gender.vbar(x='genders', top=purchase_frequencies[3], width=0.2, source=source_gender, legend_label=purchase_frequencies[3], color=Spectral6[5])

p_age.xaxis.major_label_orientation = "vertical"
# Display the plots in a layout
show(column(row(p_age, p_gender)))
```
