```python
# Primary data structure libraries
import pandas as pd
import numpy as np

# Primary plotting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv')

# Count the number of crimes for each category
category_counts = df['Category'].value_counts()

# Sort categories by frequency
category_counts = category_counts.sort_values()

# Create a gradient color list (one color per category)
colors = plt.cm.viridis(np.linspace(0, 1, len(category_counts.index)))

# Create a bar plot
plt.figure(figsize=(8,10))
bars = plt.barh(category_counts.index, category_counts.values, color=colors)

# Add gridlines
plt.grid(axis='x')

# Add labels and title
plt.xlabel('Number of crimes')
plt.ylabel('Category')
plt.title('Frequency of Crime Categories')

# Show the resultant graph
plt.show()
```
