```python
# Importing libraries
import numpy as np
import pandas as pd

# Read the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/largest_malls.csv")

# Remove duplicates
df = df.drop_duplicates()
#Filtering only completed malls
df1 = df[df['Year opened'].str.isdigit()]
#Find malls built after 2010
df1 = df1[df1['Year opened'].astype(int) > 1980]
# Extracting sqft area : Remove ',' from the string values
# Splitting to get the second part of string - sqft
df1['Gross leasable area (GLA)'] = df1['Gross leasable area (GLA)'].str.split('(').str[1]
# Splitting to get the first part of string - only number part
df1['Gross leasable area (GLA)'] = df1['Gross leasable area (GLA)'].str.split(' ').str[0]
# Removing ',' to get only digits
df1['Gross leasable area (GLA)'] = df1['Gross leasable area (GLA)'].str.replace(',','')
df1['Gross leasable area (GLA)'] = df1['Gross leasable area (GLA)'].astype(int)
# Find greater than 7,00,000 sqft area
df1 = df1[df1['Gross leasable area (GLA)'] > 700000]
# Filtering only Phillipines
dfPhillipines = df1[df1['Country'] == 'Philippines']
dfPhillipines = dfPhillipines[['Mall','City (metropolitan area)', 'Gross leasable area (GLA)', 'Shops']]
dfPhillipines
```
