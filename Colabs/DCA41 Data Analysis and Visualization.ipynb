```python
# Importing libraries
import pandas as pd

# Reading the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/student_marks.csv")

# Dropping duplicates
df = df.drop_duplicates()

# Add one column of aggregate of each student and add a average (rounded to two decimals) column
df['Average'] = df[df.columns[1:]].mean(axis=1).round(2)

# Find student with highest average
df1 = df.sort_values(by='Average',ascending=False)

# Find the top student ID assuming no tie
max_avg = df1['Average'].max()
stu_ID_Max = df1['Student_ID'][df1['Average'] == max_avg]

df2 = df1[df1['Average'] == max_avg]

# Slicing to get the maximum score excluding ID and average column
scores = df2.iloc[0,1:-1].values
max_Score = max(scores)

# Get tests where score = maximum score
max_columns = df2.columns[df2.iloc[0] == max_Score]

print(max_columns.values)
```
