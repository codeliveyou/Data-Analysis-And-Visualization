```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)
# parsing dates
df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['Year'] = df['date'].dt.year
# Calculate the mean home and away score by year
mean_scores = df.groupby('Year')[['home_score', 'away_score']].mean()

# Plot the data
fig, ax = plt.subplots()
ax.plot(mean_scores.index, mean_scores['home_score'], label='Home')
ax.plot(mean_scores.index, mean_scores['away_score'], label='Away')

# Set the title and axis labels
ax.set_title('Average home and away score by year')
ax.set_xlabel('Year')
ax.set_ylabel('Average score')

# Add a legend
ax.legend()

plt.show()
```
