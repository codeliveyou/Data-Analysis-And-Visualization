Hi Bard. Plot the paraboloid z = x^2 + y^2 with cmap = viridis and with the values of x and y between -2 and 2. Include the code in the answer.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 3D axes object
ax = plt.axes(projection="3d")

# Define the range and step size for x and y
x = np.arange(-2, 2, 0.1)
y = np.arange(-2, 2, 0.1)

# Create a grid of X and Y values using meshgrid
X, Y = np.meshgrid(x, y)

# Calculate the values of Z using the equation Z = X^2 + Y^2
Z = np.power(X, 2) + np.power(Y, 2)

# Plot the 3D surface using the calculated X, Y, and Z values
# and use the "viridis" colormap for color mapping
ax.plot_surface(X, Y, Z, cmap="viridis")

# Display the 3D plot
plt.show()
```
