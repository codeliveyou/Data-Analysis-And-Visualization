From the dataset "student_marks.csv" I want on the same plot the lines with the scores on each test of the 5 students with the highest average and the 5 with the lowest average and that there is a dotted black line representing the average scores per test. I want the legend of the plot to be outside the graph.
Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/student_marks.csv")

# Calculate average scores per student
df['Average'] = df.loc[:, 'Test_1':'Test_12'].mean(axis=1)

# Sort the dataframe by the value of 'Average' in descending order
df = df.sort_values(by='Average', ascending=False)

# Take the 5 students with the highest 'Average' and the 5 with the lowest 'Average'
top_5_students = df.head(5)
bottom_5_students = df.tail(5)

# Make figure
plt.figure(figsize=(12, 6))

# Lines of each student's grades
for index, row in top_5_students.iterrows():
    plt.plot(range(1, 13), row['Test_1':'Test_12'], marker='o', label=f'Student {row["Student_ID"]}')
for index, row in bottom_5_students.iterrows():
    plt.plot(range(1, 13), row['Test_1':'Test_12'], marker='o', label=f'Student {row["Student_ID"]}')


# Dotted black line for average scores per test
avg_test = df.loc[:, 'Test_1':'Test_12'].mean(axis=0).tolist()
plt.plot(range(1, 13), avg_test, linestyle='--', marker='o', color='black', label='Average')

# Labels, title, legend
plt.xticks(range(1, 13))
plt.xlabel('Test')
plt.ylabel('Grade')
plt.title('The 5 students with the highest average and the 5 with the lowest average')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.grid()

# Show plot
plt.show()
```
