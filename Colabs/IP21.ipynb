```python
# Code for app.py
# Import necessary libraries
import pandas as pd
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource, RadioButtonGroup
from bokeh.transform import factor_cmap
from bokeh.layouts import column
from bokeh.palettes import Category20, Category10, Category20b

# Load the dataset
df = pd.read_csv("Automobile.csv")

color_maps = [Category20, Category10, Category20b]

# Create the scatter plot
plot_title = "weight vs acceleration (colored by Origin)"
p = figure(width=800, height=600, title=plot_title)

# Get the unique values in the 'origin' column
unique_origins = df['origin'].unique()

# Convert the unique_origins to strings
unique_origins_str = unique_origins.astype(str)

# Determine the number of unique values in the 'origin' column
num_unique_origins = len(unique_origins_str)

# Define the color map for the categories
color_map = factor_cmap(field_name='origin',
                            palette=color_maps[0][num_unique_origins],
                            factors=unique_origins)

# Define the data to be used in the scatter plot
source = ColumnDataSource(dict(
    x=df['weight'],
    y=df['acceleration'],
    origin=df['origin']
    ))

# Create a scatter plot with the color map for 'origin'
p.scatter('x', 'y', source=source, size=10,
          color=color_map,
          legend_field='origin')

# Define the labels
p.xaxis.axis_label = "Weight"
p.yaxis.axis_label = "Acceleration"

# Radio button group for selecting the variable of interest
select_var = RadioButtonGroup(labels=['Palette 1','Palette 2','Palette 3'],
                              active=0)

# Define a function to update the plot when the active value changes
def update_plot(attrname, old, new):
    """
    This function updates the Bokeh bar plot with data based on the
    selected variable.
    """
    selected = select_var.active
    print(selected)

    # Define the color map for the categories
    color_map = factor_cmap(field_name='origin',
                            palette=color_maps[selected][num_unique_origins],
                            factors=unique_origins)

    # Create a scatter plot with the color map for 'origin'
    p.scatter('x', 'y', source=source, size=10,
              color=color_map,
              legend_field='origin')

# Call the update_plot function initially to populate the plot
update_plot(None, None, 0)

# Update the plot when a radio button is selected
select_var.on_change("active", update_plot)

# Arrange the plot
layout = column(select_var,p)

# Add the layout to the current Bokeh document
curdoc().add_root(layout)
```
