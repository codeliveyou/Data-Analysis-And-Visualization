Please with the dataset daily_delhi_climate.csv, I would need you to provide me graphs of the temperature and humidity in the year 2013. I would like the graphs to be side by side, have the units, legends with red and purple color respectively.



```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert the data to a DataFrame
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/daily_delhi_climate.csv')
newdf=df.set_index('date').copy()
newdf.index=pd.to_datetime(newdf.index)

# set dataframe for time series
newdf=df.set_index('date').copy()
newdf.index=pd.to_datetime(newdf.index)

# Set time interval endpoints
start_time='2013-01-01 00:00:00'
end_time='2014-01-01 00:00:00'

# source data for graph
source=newdf[start_time:end_time]

# Create subplots for scatter plots
fig, ax = plt.subplots(nrows=1, ncols=2 , figsize=(16,8))

# Set lists colors and features
colors_lst=['red' , 'purple']
fets_lst=["meantemp", "humidity"]
labels_lst=["Mean Temperature", "Humidity"]
units= iter(['°C' , '%'])

# Create axes elements
for i in range(len(colors_lst)):
  ax[i].plot(fets_lst[i], data=source, color=colors_lst[i], label=labels_lst[i])
  ax[i].tick_params(axis='x', labelrotation = 90)

# Create Labels for axes elements
labels_lst=iter(["Mean Temperature", "Humidity"])
for axis in ax:
    axis.set_ylabel(next(units))
    axis.legend()

#Set layouts and leyend
fig.tight_layout()
plt.legend(loc='upper right')

#Show Plot
plt.show()
```
