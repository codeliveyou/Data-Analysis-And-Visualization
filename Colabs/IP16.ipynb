```python
# Code for app.py
# Import necessary libraries
import pandas as pd
from bokeh.models import  TabPanel, Tabs
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource, Scatter
from bokeh.layouts import column

# Load the dataset
df = pd.read_csv("cancer_diagnosis_data.csv")

# Create a list of variable names for each tab
tabs = ["smoothness_mean", "texture_mean", "perimeter_mean"]

# Create labels for each tab and X label
tabs_names = ["Smoothness Mean", "Texture Mean", "Perimeter Mean"]

# Create the first plot
p1 = figure(width=800, height=600,
           title="Radius vs Smoothness")
source = ColumnDataSource(dict(x=df['radius_mean'], y=df[tabs[0]]))
glyph= Scatter(x='x', y='y')
p1.add_glyph(source, glyph)
p1.xaxis.axis_label = "Radius Mean"
p1.yaxis.axis_label = tabs_names[0]
tab1 = TabPanel(child=p1, title=tabs_names[0])

# Create the second plot
p2 = figure(width=800, height=600,
           title="Radius vs Smoothness")
source = ColumnDataSource(dict(x=df['radius_mean'], y=df[tabs[1]]))
glyph= Scatter(x='x', y='y')
p2.add_glyph(source, glyph)
p2.xaxis.axis_label = "Radius Mean"
p2.yaxis.axis_label = tabs_names[1]
tab2 = TabPanel(child=p2, title=tabs_names[1])

# Create the third plot
p3 = figure(width=800, height=600, title="Radius vs Smoothness")
source = ColumnDataSource(dict(x=df['radius_mean'], y=df[tabs[2]]))
glyph= Scatter(x='x', y='y')
p3.add_glyph(source, glyph)
p3.xaxis.axis_label = "Radius Mean"
p3.yaxis.axis_label = tabs_names[2]
tab3 = TabPanel(child=p3, title=tabs_names[2])

# Arrange the plot
layout = column(Tabs(tabs=[tab1, tab2, tab3]))

# Add the layout to the current document (Bokeh server)
curdoc().add_root(layout)
```

This is running with Bokeh server version **3.2.1** (running on Tornado 6.1)

You can save this code to a file named `app.py`, and then run it from the command-line like this:

`bokeh server --show app.py`
