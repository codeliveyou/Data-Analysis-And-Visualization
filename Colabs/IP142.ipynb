```python
# Import libraries
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.io import output_notebook

# Load and preprocess data
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/house_property_raw_sales.csv")

df['datesold'] = pd.to_datetime(df['datesold'])

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create the plot
p = figure(title="Bedrooms vs Price", x_axis_label="Bedrooms", y_axis_label="Price")

# Adding scatter plot
p.scatter(x='bedrooms', y='price', source=source, size=10, color='blue', legend_field='propertyType')

# Adding hover tool
hover = HoverTool()
hover.tooltips = [("Price", "@price"), ("Bedrooms", "@bedrooms"), ("Property Type", "@propertyType")]
p.add_tools(hover)

# Display the plot
output_notebook()
show(p)
```
