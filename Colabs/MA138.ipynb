**Prompt:**
Hello, with the dataset sanfrancisco_crimes_data_2016.csv I need you to make a pie diagram that shows me the distribution by day of the week of juvenile and arrest booked. The borders of the diagram should be dark and with a 5% explode. I need the code in python.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

plt.style.use("bmh")

# Load data from into a pandas dataframe and select data for query
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv")

# Create filter for weapon law
filt_wl = (df['Resolution'] == 'JUVENILE BOOKED') | (df['Resolution'] == 'ARREST, BOOKED')

# create new dataset
newdf = df[filt_wl].copy()

# Select data
newdf = df['DayOfWeek'].value_counts()
labels = newdf.index

# Create explode list with required ascending fashion
myexplode = [0.05 for i in range(len(labels))]

# Set graph size and format
fig, ax = plt.subplots(figsize=(10, 10))

# Create pie graph
ax.pie(newdf, labels=labels,
       labeldistance = 1.2,
       explode = myexplode,
       autopct='%1.1f%%',
       wedgeprops = {"linewidth": 1, "edgecolor": "black"},
       startangle=90, )

# Define title Graph
plt.title("Juvenile and arrest booked distribution by day of week")

# Show graph
plt.show()
```
