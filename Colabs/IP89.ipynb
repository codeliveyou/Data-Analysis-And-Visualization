**Prompt:**

Bard, with this dataset asteroid_all.csv I want you to make an interactive scatterplot that has a color bar in intensity by relative speed of approach of the asteroids, it should be the apparent magnitude versus the relative speed. I need that when passing the pointer over the value of the graph it shows me the reference number and also to be able to scroll interactively.

```python
#Import libraries
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
import pandas as pd
from bokeh.models import ColumnDataSource, ColorBar, HoverTool
from bokeh.palettes import Spectral6
from bokeh.transform import linear_cmap

# Creating pandas dataframe from csv dataset
df=pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/asteroid_all.csv')

# Set mapping colors for relative velocity like continuos variable
mapper = linear_cmap(field_name='Vrelative(km/s)', palette=Spectral6 ,low=min(df['Vrelative(km/s)']) ,high=max(df['Vrelative(km/s)']))

#Create tooltips list
TOOLTIPS = [
    ('Close Aproche', '@ref')

]

# Create figure with hover tool for interactvity and scatter plot
p = figure(width=600, height=600,toolbar_location='right' , tooltips=TOOLTIPS)
p.scatter('H(mag)', 'Vrelative(km/s)', source=df,fill_alpha=0.4, line_color=mapper,color=mapper,size=10)

# Set values for axis labels
p.xaxis.axis_label = 'Apparent Magnitute'

# Create color bar for relative velocity and add layout with this bar to the left
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, label_standoff = 14,  location=(0,0),title="Relative Velocity Km/s" )
p.add_layout(color_bar, 'left')

# Show plot
output_notebook()
show(p)
```
