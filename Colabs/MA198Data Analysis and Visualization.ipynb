Create a line chart showing the percentage change in the number of malls opened each year.Provide the code for the same.Use the provided csv file.Provide the code for the same.Use the provided
csv file.




```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/largest_malls.csv")

# Extract year opened data
years_opened = data['Year opened']

# Calculate the number of malls opened each year
year_counts = years_opened.value_counts().sort_index()

# Calculate the percentage change in the number of malls opened each year
percentage_changes = year_counts.pct_change() * 100

# Create a line chart
plt.figure(figsize=(20, 12))
plt.plot(percentage_changes.index, percentage_changes.values, marker='o')
plt.title('Percentage Change in Number of Malls Opened Each Year')
plt.xlabel('Year')
plt.ylabel('Percentage Change (%)')
plt.xticks(percentage_changes.index,rotation=45)
plt.grid(True)


# Show the line chart
plt.show()
```

Sure, here is the line chart showing the percentage change in the number of malls opened each year.
