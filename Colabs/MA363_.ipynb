```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Reading the dataset
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/pokemon_dataset.csv")

# Counting the total occurence for each of the types in 'Type1' column
count = data['Type1'].value_counts()

# plotting the pie chart
plt.figure(figsize=(10,7))
count.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors) # Configuration
plt.title('Distribution of Pokemon by Type1') # Set title
plt.ylabel('') # Remove y-axis label

plt.show()
```
