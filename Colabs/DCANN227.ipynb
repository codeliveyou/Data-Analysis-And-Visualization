```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

try:
    # Read the workout data from the CSV file
    data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv", encoding='ISO-8859-1')
    data.head()
    print("CSV file loaded successfully, you can start working right now")
except Exception as e:
    print(f"An error occurred while loading the CSV file: {e}, please check the correct mount of the shared folder")

# Scatter plot to show the relationship between experience level and salary in USD
plt.figure(figsize=(10, 6))
sns.scatterplot(x='experience_level', y='salary_in_usd', data=data)
plt.title('Experience Level vs Salary in USD')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()

# Heatmap to show salary distribution across different locations
location_salary = data.groupby('employee_residence')['salary_in_usd'].mean().reset_index()
location_salary_pivot = location_salary.pivot_table(index='employee_residence', values='salary_in_usd', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(location_salary_pivot, cmap='YlGnBu', annot=True)
plt.title('Heatmap of Salary Distribution Across Locations')
plt.xlabel('Location')
plt.ylabel('Salary in USD')
plt.show()

# Box plot to show salary distribution by different employment types
plt.figure(figsize=(12, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=data)
plt.title('Salary Distribution by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary in USD')
plt.show()
```
