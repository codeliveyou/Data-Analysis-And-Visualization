```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256
from bokeh.layouts import column, row
from bokeh.transform import linear_cmap
import warnings
warnings.filterwarnings('ignore')

# Read the data from CSV
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/startup_data.csv")
df.drop(columns = ['Unnamed: 6',"Unnamed: 0"],inplace=True)

# Bar chart for companies vs categories

counts = df['category_code'].value_counts().sort_values(ascending=False)
categories = list(counts.index)
source1 = ColumnDataSource(data=dict(categories=categories, counts=counts[categories]))

# Create color mapper
color_mapper = linear_cmap(field_name='counts', palette=Viridis256, low=min(counts), high=max(counts))

# source1 = ColumnDataSource(data=dict(continents=continents, counts=counts[continents]))

# Create a bar chart
hover = HoverTool(tooltips=[("Categories", "@categories"), ("Categories Count", "@counts")])
plot = figure(x_range=categories, title="Company Distribution by Category", tools=[hover], height=800, width=800)
plot.vbar(x='categories', top='counts', width=0.9, source=source1, legend_label="Categories", color=color_mapper)
plot.xaxis.major_label_orientation = "vertical"

output_notebook()

show(column(row(plot)))
```
