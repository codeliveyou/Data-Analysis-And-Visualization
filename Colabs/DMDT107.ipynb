````python
#```python
# from google.colab import drive
# drive.mount("/content/gdrive", force_remount=True)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/avocado.csv")
    df['Date'] = pd.to_datetime(df['Date'])
    # Group by year and calculate the average price
    average_prices_by_year = df.groupby(df['Date'].dt.year)['AveragePrice'].mean()
    # Plotting the trend in average prices over the years
    plt.figure(figsize=(10, 6))
    plt.plot(average_prices_by_year.index, average_prices_by_year.values, marker='o')
    plt.xlabel('Year')
    plt.ylabel('Average Price')
    plt.title('Average Avocado Price Trend Over Years')
    plt.grid(True)
    plt.show()

except Exception as e:
    print(f"An error occurred while loading the CSV file: {e}, please check the correct path to the CSV file.")
#```
````
