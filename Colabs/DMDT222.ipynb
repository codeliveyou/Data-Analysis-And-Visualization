```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)
# Calculate the mean away score by date
mean_away_score = df.groupby('date')['away_score'].mean()

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(mean_away_score.index, mean_away_score.values, color='red', linewidth=2)

# Set the title and axis labels
ax.set_title('Average away score over time')
ax.set_xlabel('Date')
ax.set_ylabel('Average away score')

plt.show()
```
