```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.plotting import output_notebook
from bokeh.transform import factor_cmap, jitter
from bokeh.palettes import Category20c
import warnings
warnings.filterwarnings('ignore')

# Read the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/laptops_price_dataset.csv")
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.sort_values(by='Storage', inplace = True)
df = df[df['Brand'] == 'Acer']

# Add colors to our dataset based on the destination port
scatter_data = df.copy()
RAM = scatter_data['RAM'].unique()
palette = Category20c[len(RAM)+1]
colormap = {RAM[i]: palette[i] for i in range(len(RAM))}
scatter_data['colors'] = [colormap[x] for x in scatter_data['RAM']]

# Create a figure and add a number of tools
p = figure(height=800, width=800, title = "RAM vs Storage",
           tools="pan,wheel_zoom,box_zoom,reset,hover,save",
           tooltips="RAM = @RAM - Storage = @Storage",
           toolbar_location="below")


# format our data in a ColumnDataSource format
data = ColumnDataSource(scatter_data)
# Add scatter object
# Use jitter to spread the overlapping points apart
p.scatter(x=jitter("RAM", width=5, range=p.x_range), y=jitter('Storage', width=50),
          source=data, legend_field="RAM", fill_alpha=0.8, size=7,
          fill_color='colors', line_color='colors')
# Format axes
p.xaxis.axis_label = 'RAM'
p.yaxis.axis_label = 'Storage'

output_notebook()
show(p)
```
