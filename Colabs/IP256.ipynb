* Prompt: Write a Python code to create an interactive plot where number of crimes in different District in each weekday is shown in bar chart. There should be horizontal slider functionality. When sliding one plot, other plots should slide accordingly. There should be a box zoom functionality. When zooming one plot, other plots should zoom accordingly There should also be a reset button. When hovering on any point, the value of the point must be shown. Finally show the output.

```python
# Import Libraries
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.models import (
    ColumnDataSource,
    Tabs,
    TabPanel,
    HoverTool,
    CDSView,
    GroupFilter,
    FactorRange,
)
from bokeh.layouts import gridplot, column
from bokeh.palettes import Viridis256
import random

# Droping rows that has Null Category Value
df = df.dropna(subset=["PdDistrict"])

# Droping rows that has Null PdDistrict Value
df = df.dropna(subset=["DayOfWeek"])

# Group by "PdDistrict" and "DayOfWeek", then unstack to create separate columns for Days of Week
new_df = df.groupby(["PdDistrict", "DayOfWeek"]).size().unstack(fill_value=0)

# Reset the index to make 'PdDistrict' a column again
new_df.reset_index(inplace=True)

# PdDistrict Names
PdDistrict = new_df["PdDistrict"]

# Saturday Crime
Saturday = new_df["Saturday"].values

# Creating Figure
p1 = figure(
    height=400,
    width=700,
    x_range=PdDistrict,
    title="Crime in Saturday",
    tools=["xpan", "hover", "box_zoom", "reset", "save"],
)
p1.vbar(
    x=PdDistrict,
    top=Saturday,
    width=0.9,
    color=random.sample(Viridis256, 10),
    fill_alpha=0.75,
)

# Sunday Crime
Sunday = new_df["Sunday"].values

p2 = figure(
    height=400,
    width=700,
    x_range=p1.x_range,
    title="Crime in Sunday",
    tools=["xpan", "hover", "box_zoom", "reset", "save"],
)
p2.vbar(
    x=PdDistrict,
    top=Sunday,
    width=0.9,
    color=random.sample(Viridis256, 10),
    fill_alpha=0.75,
)

# Monday Crime
Monday = new_df["Monday"].values

p3 = figure(
    height=400,
    width=700,
    x_range=p1.x_range,
    title="Crime in Monday",
    tools=["xpan", "hover", "box_zoom", "reset", "save"],
)
p3.vbar(
    x=PdDistrict,
    top=Monday,
    width=0.9,
    color=random.sample(Viridis256, 10),
    fill_alpha=0.75,
)

# Tuesday Crime
Tuesday = new_df["Tuesday"].values

p4 = figure(
    height=400,
    width=700,
    x_range=p1.x_range,
    title="Crime in Tuesday",
    tools=["xpan", "hover", "box_zoom", "reset", "save"],
)
p4.vbar(
    x=PdDistrict,
    top=Tuesday,
    width=0.9,
    color=random.sample(Viridis256, 10),
    fill_alpha=0.75,
)

# Wednesday Crime
Wednesday = new_df["Wednesday"].values

p5 = figure(
    height=400,
    width=700,
    x_range=p1.x_range,
    title="Crime in Wednesday",
    tools=["xpan", "hover", "box_zoom", "reset", "save"],
)
p5.vbar(
    x=PdDistrict,
    top=Wednesday,
    width=0.9,
    color=random.sample(Viridis256, 10),
    fill_alpha=0.75,
)

# Thursday Crime
Thursday = new_df["Thursday"].values

p6 = figure(
    height=400,
    width=700,
    x_range=p1.x_range,
    title="Crime in Thursday",
    tools=["xpan", "hover", "box_zoom", "reset", "save"],
)
p6.vbar(
    x=PdDistrict,
    top=Thursday,
    width=0.9,
    color=random.sample(Viridis256, 10),
    fill_alpha=0.75,
)

# Friday Crime
Friday = new_df["Friday"].values

p7 = figure(
    height=400,
    width=700,
    x_range=p1.x_range,
    title="Crime in Friday",
    tools=["xpan", "hover", "box_zoom", "reset", "save"],
)
p7.vbar(
    x=PdDistrict,
    top=Friday,
    width=0.9,
    color=random.sample(Viridis256, 10),
    fill_alpha=0.75,
)

# Set some properties to make the plot look better
p7.yaxis.axis_label = (
    p6.yaxis.axis_label
) = (
    p5.yaxis.axis_label
) = (
    p4.yaxis.axis_label
) = p3.yaxis.axis_label = p2.yaxis.axis_label = p1.yaxis.axis_label = "Number of Crime"

p7.xgrid.grid_line_color = (
    p6.xgrid.grid_line_color
) = (
    p5.xgrid.grid_line_color
) = (
    p4.xgrid.grid_line_color
) = (
    p3.xgrid.grid_line_color
) = p2.xgrid.grid_line_color = p1.xgrid.grid_line_color = "gray"

p7.xgrid.grid_line_alpha = (
    p6.xgrid.grid_line_alpha
) = (
    p5.xgrid.grid_line_alpha
) = (
    p4.xgrid.grid_line_alpha
) = (
    p3.xgrid.grid_line_alpha
) = p2.xgrid.grid_line_alpha = p1.xgrid.grid_line_alpha = 0.75

p7.xgrid.grid_line_dash = (
    p6.xgrid.grid_line_dash
) = (
    p5.xgrid.grid_line_dash
) = (
    p4.xgrid.grid_line_dash
) = (
    p3.xgrid.grid_line_dash
) = p2.xgrid.grid_line_dash = p1.xgrid.grid_line_dash = "dashed"

p7.ygrid.grid_line_color = (
    p6.ygrid.grid_line_color
) = (
    p5.ygrid.grid_line_color
) = (
    p4.ygrid.grid_line_color
) = (
    p3.ygrid.grid_line_color
) = p2.ygrid.grid_line_color = p1.ygrid.grid_line_color = "blue"

p7.ygrid.grid_line_alpha = (
    p6.ygrid.grid_line_alpha
) = (
    p5.ygrid.grid_line_alpha
) = (
    p4.ygrid.grid_line_alpha
) = (
    p3.ygrid.grid_line_alpha
) = p2.ygrid.grid_line_alpha = p1.ygrid.grid_line_alpha = 0.55

p7.ygrid.grid_line_dash = (
    p6.ygrid.grid_line_dash
) = (
    p5.ygrid.grid_line_dash
) = (
    p4.ygrid.grid_line_dash
) = (
    p3.ygrid.grid_line_dash
) = p2.ygrid.grid_line_dash = p1.ygrid.grid_line_dash = "dotted"

p7.xaxis.major_label_orientation = (
    p6.xaxis.major_label_orientation
) = (
    p5.xaxis.major_label_orientation
) = (
    p4.xaxis.major_label_orientation
) = (
    p3.xaxis.major_label_orientation
) = p2.xaxis.major_label_orientation = p1.xaxis.major_label_orientation = 0.9

p = gridplot(
    [[p1, p2, p3], [p4, p5, p6], [None, p7, None]],
    toolbar_location="left",
)
output_notebook()
show(p)
```

```python
# Import Libraries
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.models import (
    ColumnDataSource,
    Tabs,
    TabPanel,
    HoverTool,
    CDSView,
    GroupFilter,
)
from bokeh.layouts import gridplot, column
import warnings

# Ignoring Warnings
warnings.filterwarnings("ignore")

# Droping rows that has Null Category Value
df = df.dropna(subset=["Category"])

# Droping rows that has Null Latitude or Longitude Value
df = df.dropna(subset=["X", "Y"])
cols = ["X", "Y"]

# Converting String to Numeric Data
df[cols] = df[cols].apply(pd.to_numeric, errors="coerce", axis=1)

# Selecting Data based on Catagory
STOLEN_PROPERTY = df[df["Category"] == "STOLEN PROPERTY"]
STOLEN_PROPERTY_cds = ColumnDataSource(STOLEN_PROPERTY)
RECOVERED_VEHICLE = df[df["Category"] == "RECOVERED VEHICLE"]
RECOVERED_VEHICLE_cds = ColumnDataSource(RECOVERED_VEHICLE)
DISORDERLY_CONDUCT = df[df["Category"] == "DISORDERLY CONDUCT"]
DISORDERLY_CONDUCT_cds = ColumnDataSource(DISORDERLY_CONDUCT)
SUICIDE = df[df["Category"] == "SUICIDE"]
SUICIDE_cds = ColumnDataSource(SUICIDE)
DRUNKENNESS = df[df["Category"] == "DRUNKENNESS"]
DRUNKENNESS_cds = ColumnDataSource(DRUNKENNESS)

output_notebook()

# Selection Tools
select_tools = ["hover", "lasso_select", "box_zoom", "reset", "save"]

# Defining Figure
DRUNKENNESS_fig = figure(
    height=600,
    width=600,
    title="DRUNKENNESS",
    x_axis_label="Latitude",
    y_axis_label="Longitude",
    toolbar_location="below",
    tools=select_tools,
)
DRUNKENNESS_fig.circle(
    x="X",
    y="Y",
    source=DRUNKENNESS_cds,
    color="royalblue",
    size=5,
    alpha=0.5,
    selection_color="deepskyblue",
    nonselection_color="lightgray",
    nonselection_alpha=0.3,
)

# Configure render to be used as hover
hover_glyph = DRUNKENNESS_fig.circle(
    x="X",
    y="Y",
    source=DRUNKENNESS_cds,
    size=10,
    alpha=0,
    hover_fill_color="black",
    hover_alpha=0.3,
)
DRUNKENNESS_fig.add_tools(HoverTool(renderers=[hover_glyph]))


STOLEN_PROPERTY_fig = figure(
    height=600,
    width=600,
    title="STOLEN PROPERTY",
    x_axis_label="Latitude",
    y_axis_label="Longitude",
    toolbar_location="below",
    tools=select_tools,
)
STOLEN_PROPERTY_fig.circle(
    x="X",
    y="Y",
    source=STOLEN_PROPERTY_cds,
    color="brown",
    size=5,
    alpha=0.5,
    selection_color="sandybrown",
    nonselection_color="lightgray",
    nonselection_alpha=0.3,
)
hover_glyph = STOLEN_PROPERTY_fig.circle(
    x="X",
    y="Y",
    source=STOLEN_PROPERTY_cds,
    size=10,
    alpha=0,
    hover_fill_color="black",
    hover_alpha=0.3,
)
STOLEN_PROPERTY_fig.add_tools(HoverTool(renderers=[hover_glyph]))

RECOVERED_VEHICLE_fig = figure(
    height=600,
    width=600,
    title="RECOVERED VEHICLE",
    x_axis_label="Latitude",
    y_axis_label="Longitude",
    toolbar_location="below",
    tools=select_tools,
)
RECOVERED_VEHICLE_fig.circle(
    x="X",
    y="Y",
    source=RECOVERED_VEHICLE_cds,
    color="darkturquoise",
    size=5,
    alpha=0.5,
    selection_color="cyan",
    nonselection_color="lightgray",
    nonselection_alpha=0.3,
)
hover_glyph = RECOVERED_VEHICLE_fig.circle(
    x="X",
    y="Y",
    source=RECOVERED_VEHICLE_cds,
    size=10,
    alpha=0,
    hover_fill_color="black",
    hover_alpha=0.3,
)
RECOVERED_VEHICLE_fig.add_tools(HoverTool(renderers=[hover_glyph]))

DISORDERLY_CONDUCT_fig = figure(
    height=600,
    width=600,
    title="DISORDERLY CONDUCT",
    x_axis_label="Latitude",
    y_axis_label="Longitude",
    toolbar_location="below",
    tools=select_tools,
)
DISORDERLY_CONDUCT_fig.circle(
    x="X",
    y="Y",
    source=DISORDERLY_CONDUCT_cds,
    color="green",
    size=5,
    alpha=0.5,
    selection_color="lime",
    nonselection_color="lightgray",
    nonselection_alpha=0.3,
)
hover_glyph = DISORDERLY_CONDUCT_fig.circle(
    x="X",
    y="Y",
    source=DISORDERLY_CONDUCT_cds,
    size=10,
    alpha=0,
    hover_fill_color="black",
    hover_alpha=0.3,
)
DISORDERLY_CONDUCT_fig.add_tools(HoverTool(renderers=[hover_glyph]))

SUICIDE_fig = figure(
    height=600,
    width=600,
    title="SUICIDE",
    x_axis_label="Latitude",
    y_axis_label="Longitude",
    toolbar_location="below",
    tools=select_tools,
)
SUICIDE_fig.circle(
    x="X",
    y="Y",
    source=SUICIDE_cds,
    color="orange",
    size=10,
    alpha=0.5,
    selection_color="orangered",
    nonselection_color="lightgray",
    nonselection_alpha=0.3,
)
hover_glyph = SUICIDE_fig.circle(
    x="X",
    y="Y",
    source=SUICIDE_cds,
    size=10,
    alpha=0,
    hover_fill_color="black",
    hover_alpha=0.3,
)
SUICIDE_fig.add_tools(HoverTool(renderers=[hover_glyph]))

# Creating Tabs
DRUNKENNESS_tab = TabPanel(child=DRUNKENNESS_fig, title="DRUNKENNESS")
STOLEN_PROPERTY_tab = TabPanel(child=STOLEN_PROPERTY_fig, title="STOLEN PROPERTY")
RECOVERED_VEHICLE_tab = TabPanel(child=RECOVERED_VEHICLE_fig, title="RECOVERED VEHICLE")
DISORDERLY_CONDUCT_tab = TabPanel(
    child=DISORDERLY_CONDUCT_fig, title="DISORDERLY CONDUCT"
)
SUICIDE_tab = TabPanel(child=SUICIDE_fig, title="SUICIDE")

tabs = Tabs(
    tabs=[
        DRUNKENNESS_tab,
        STOLEN_PROPERTY_tab,
        RECOVERED_VEHICLE_tab,
        DISORDERLY_CONDUCT_tab,
        SUICIDE_tab,
    ]
)

show(tabs)
```
