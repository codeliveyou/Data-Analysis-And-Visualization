```python
# Primary data structure library
import pandas as pd

# Primary plotting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Set the fig size
plt.figure(figsize=(20, 3))

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv')

# Group the data by job_title and experience_level, compute the average salary
grouped = df.groupby(['job_title', 'experience_level'])['salary_in_usd'].mean().reset_index()

# Create a numerical column for experience_level for color-coding
grouped['exp_level_num'] = grouped['experience_level'].astype('category').cat.codes

# Create the bubble chart (scatter plot)
scatter = plt.scatter(grouped['job_title'], grouped['salary_in_usd'], s=grouped['salary_in_usd']/1000, c=grouped['exp_level_num'], alpha=0.5)

# Add a color bar
plt.colorbar(scatter, label='Experience Level')

# Add title and labels
plt.title('Average Salary in USD by Job Title and Experience Level')
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')

# Rotate x-axis ticks for better readability
plt.xticks(rotation=90)

# Output the resultant plot
plt.show()
```
