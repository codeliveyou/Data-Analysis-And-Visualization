Hello Bard, from the data set "IMDb_Top_250_Movies.csv" I want 4 horizontal bar charts in the same plot. At the top I want one that shows the top 10 most frequent actors and next to it the most frequent directors of the most frequent actor. Below I want one that shows the top 10 most frequent directors and next to it the most frequent actors of the most frequent director. Let all the bars be sorted in descending order according to their length.Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/imdb_top_250_movies.csv", encoding='unicode_escape')

# Get only the top 10 actors
actors = [actor.strip() for star_list in df['Stars'] for actor in star_list.split(',')]
actor_counts = pd.Series(actors).value_counts()[:10]

# Get only the top 10 directors
director_counts = df['Director'].value_counts()[:10]
most_common_actor = actor_counts.idxmax()

# Filter the dataframe to get rows where the most common actor appears
common_actor_rows = df[df['Stars'].str.contains(most_common_actor)]

# Find directors the most common actor has worked with
common_directors = common_actor_rows['Director'].value_counts()

# Find the most common director
most_common_director = director_counts.idxmax()

# Filter the dataframe to get rows of the most common director
common_director_rows = df[df['Director'] == most_common_director]

# Find actors that most frequently worked with the most common director
common_actors_for_director = [actor.strip() for star_list in common_director_rows['Stars'] for actor in star_list.split(',')]
actor_counts_for_director = pd.Series(common_actors_for_director).value_counts()[:10]

# Create bar plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Bar plot for actors
actor_counts.sort_values(ascending=True).plot(kind='barh', ax=axes[0, 0])
axes[0, 0].set_title('Actor Frequency (Top 10)')
axes[0, 0].set_xlabel('Count')
axes[0, 0].set_ylabel('Actors')

# Bar plot for directors
director_counts.sort_values(ascending=True).plot(kind='barh', ax=axes[1, 0], color='orange')
axes[1, 0].set_title('Director Frequency (Top 10)')
axes[1, 0].set_xlabel('Count')
axes[1, 0].set_ylabel('Directors')

# Bar plot for directors of the most common actor
common_directors.sort_values(ascending=True).plot(kind='barh', ax=axes[0, 1], color='green')
axes[0, 1].set_title(f'Directors of {most_common_actor}')
axes[0, 1].set_xlabel('Count')
axes[0, 1].set_ylabel('Directors')

# Bar plot for actors of the most common director
actor_counts_for_director.sort_values(ascending=True).plot(kind='barh', ax=axes[1, 1], color='purple')
axes[1, 1].set_title(f'Actors for {most_common_director}')
axes[1, 1].set_xlabel('Count')
axes[1, 1].set_ylabel('Actors')

# Show the plot
plt.tight_layout()
plt.show()
```
