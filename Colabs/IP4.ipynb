```python
# Code for app.py
# Import necessary libraries
import pandas as pd
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource, RadioButtonGroup
from bokeh.transform import factor_cmap
from bokeh.layouts import column

# Load the dataset
df = pd.read_csv("cancer_diagnosis_data.csv")

# Create the scatter plot
plot_title = "Radius Mean vs Texture Mean"
p = figure(width=800, height=600, title=plot_title)

# Get the unique values in the 'diagnosis' column
unique_diagnosis = df['diagnosis'].unique()

# Convert to strings
unique_diagnosis_str = unique_diagnosis.astype(str)

# Determine the number of unique values in the 'diagnosis' column
num_unique_diagnosis = len(unique_diagnosis_str)

# In this case the most simple palettes are color pairs since there are only 2 categories
colors1 = ['#1f78b4', '#33a02c']
colors2 = ['#e41a1c', '#377eb8']
colors3 = ['#984ea3', '#ff7f00']
colors = [colors1, colors2, colors3]

# Define the initial color map for the categories
color_map = factor_cmap(field_name='diagnosis',
                            palette=colors[0],
                            factors=unique_diagnosis)

# Create the initial data
source = ColumnDataSource(dict(
    x=df['radius_mean'],
    y=df['texture_mean'],
    diagnosis=df['diagnosis']))

# Create a scatter plot with the color map for 'diagnosis'
p.scatter('x', 'y', source=source, size=10,
          color=color_map,
          legend_field='diagnosis')

# Create the labels
p.xaxis.axis_label = "Radius Mean"
p.yaxis.axis_label = "Texture Mean"

# Radio button group for selecting the palette
select_var = RadioButtonGroup(labels=['Palette 1','Palette 2','Palette 3'],
                              active=0)

# Define a function to update the plot when the active value changes
def update_plot(attrname, old, new):
    """
    This function updates the Bokeh bar plot with a different palette based on the
    selected value.
    """
    selected = select_var.active

    # Define the color map for the categories
    color_map = factor_cmap(field_name='diagnosis',
                            palette=colors[selected],
                            factors=unique_diagnosis)

    # Create a scatter plot with the color map for 'diagnosis'
    p.scatter('x', 'y', source=source, size=10,
              color=color_map,
              legend_field='diagnosis')

# Call the update_plot function initially to populate the plot
update_plot(None, None, 0)

# Update the plot when a radio button is selected
select_var.on_change("active", update_plot)

# Arrange the plot
layout = column(select_var,p)

# Add the layout to the current document (Bokeh server)
curdoc().add_root(layout)
```

You can save this code to a file named `app.py`, and then run it from the command-line like this:

`bokeh server --show app.py`
