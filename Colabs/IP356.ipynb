```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool
from bokeh.plotting import output_notebook
import warnings
warnings.filterwarnings('ignore')

# Read data
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/startup_data.csv")

# Dropping irrelevant columns
df.drop(columns = ['Unnamed: 6',"Unnamed: 0"],inplace=True)

# Create Bokeh figure
p = figure(title="Number of Milestones Achieved by Companies", x_axis_label="Number of Milestones", y_axis_label="Count")

# Calculate histogram
hist, edges = np.histogram(df["age_first_milestone_year"].dropna(), bins=10)

# Add histogram plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color="white")

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Milestones Range", "@left - @right"), ("Count", "@top")]
p.add_tools(hover)

# Output the plot to an HTML file
output_file("histogram_chart.html")

output_notebook()
# Show the plot
show(p)
```
