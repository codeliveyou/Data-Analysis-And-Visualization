Hi Bard, I have a dataset 'ufo_sighting_data.csv' that has a 'country' field with values 'us', nan, 'gb', 'ca', 'au', 'de'. I need a line chart showing the number of ufos by date, without 'nan' or 'us'.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset loading
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/ufo_sighting_data.csv')

# Filter records with valid values in the field 'country'
valid_countries = ['gb', 'ca', 'au', 'de']
filtered_data = df[df['country'].isin(valid_countries)]

# Field format is changed to datetime
filtered_data['date_documented'] = pd.to_datetime(df["date_documented"], format="%m/%d/%Y")

# Counting the number of sightings by date
ufo_counts_by_date = filtered_data.groupby('date_documented').size()

# Create a line chart for each country
plt.figure(figsize=(10, 6))
for country in valid_countries:
    country_data = filtered_data[filtered_data['country'] == country]
    ufo_counts_by_date = country_data.groupby('date_documented').size()
    plt.plot(ufo_counts_by_date.index, ufo_counts_by_date.values, label=country)

plt.xlabel('Date')
plt.ylabel('Number of Sightings')
plt.title('Number of UFO Sightings by Date and Country')
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the graph
plt.show()
```
