```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into data frame
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/world_cups_data.csv')

# Create a dictionary to store evaluation of each country at world cup
country_points = dict()

# Evaluate each country by their first, second, and third place performances
for index, row in df.iterrows():
  champion = row['Champion']
  runner_up = row['Runner-up']
  third_place = row['3rd Place']

  if champion not in country_points:
    country_points[champion] = 0
  if runner_up not in country_points:
    country_points[runner_up] = 0
  if third_place not in country_points:
    country_points[third_place] = 0

  country_points[champion] += 3
  country_points[runner_up] += 2
  country_points[third_place] += 1

# Sort countries by their total score
sorted_countries = sorted(country_points.items(), key = lambda x:x[1], reverse = True)

# Create lists of countries and their scores for graph
ranked_countries = [x[0] for x in sorted_countries]
scores = [x[1] for x in sorted_countries]

# Configure plot for countries and their respective scores
plt.figure(figsize = (12, 8))
bars = plt.bar(ranked_countries, scores)

# Set plot titles and labels
plt.title('Performance Metric of Countries at World Cup')
plt.xlabel('Countries')
plt.ylabel('Scores')

# Provide a score inside each bar of bar graph to display score of each country
for bar, score in zip(bars, scores):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, f"{score}", ha='center', va='center', color='black')

# Create annotation explaining the points system
plt.annotate("First Place: +3 Points", xy = (15, 20))
plt.annotate("Runner Up: +2 Points", xy = (15, 19))
plt.annotate("Third Place: +1 Points", xy = (15, 18))

# Display plot
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```
