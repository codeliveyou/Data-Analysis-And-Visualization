**All the code together**

```python
#import the libraries
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

#set color blind friendly style
plt.style.use('tableau-colorblind10')

# load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/Top-Largest-Universities.csv")

# filter by America continent
data = df[df["Continent"].str.contains("America")]

# Group by the data by Location and Affiliation, then sum the enrollment
data = data.groupby(["Location","Affiliation"])["Enrollment"].sum().reset_index()

# pivot the previous df_group, see pandas.DataFrame.pivot for more information
data = data.pivot(index="Location",columns="Affiliation",values="Enrollment").reset_index()

# plot the location in X-axis using a stacked bar chart
data.plot(x="Location", kind='bar', stacked=True, figsize=(12,6),
          title = "Number of Enrollment per type of Affiliation in Locations in America continent")

# Display numbers without scientific notation
plt.ticklabel_format(style='plain', axis='y')

# show the plot
plt.show()
```
