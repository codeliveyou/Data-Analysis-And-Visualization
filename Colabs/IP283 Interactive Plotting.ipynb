```python
# Importing the necessary libraries
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import HoverTool, ColumnDataSource

# Load the dataset file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/6mon_nifty_clean_text.csv")
df['Date'] = pd.to_datetime(df['Date'])

# Creata a Bokeh ColumDataSource
source = ColumnDataSource(df)

# Creata a Bokeh figure
fig = figure(title="Interactive Line Chart: Volume Over Time", x_axis_label="datetime")

# Create the line chart for the Volume column
fig.line(x='Date', y='Volume', source=source, line_width=2, color='green', legend_label='Volume')

# Add hover tool to display exact volume and date
hover = HoverTool(tooltips=[("Date", "@Date{%F}"), ("Volume", "@Volume")], formatters={"@Date": "datetime"})
fig.add_tools(hover)

# Customize x-axis and y-axis labels
fig.xaxis.axis_label = 'Date'
fig.yaxis.axis_label = 'Volume'

# Show legend
fig.legend.location = "top_left"

# Output the plot
output_notebook()
show(fig)
```
