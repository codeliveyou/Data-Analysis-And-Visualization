**All the code together**

```python
#import the libraries
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

#set a gray color style
plt.style.use('grayscale')

# load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/Top-Largest-Universities.csv")

#filter the dataset by Continent = North America,
# then we keep the first 10 rows, and the
# columns "Rank","Institution","Enrollment"
data = df[(df.Continent == "North America")].head(10)[["Rank","Institution","Enrollment"]]

# create a new column called Institutions, where store the concatenation of column Institution and Rank
data["Institutions"] = data['Institution'] +" - Position: "+ data["Rank"].astype(str)

# plot a bar plot horizontal graph
ax2 = data.plot.barh(x='Institutions',y='Enrollment')

# Show top values first in the Y axis
ax2.invert_yaxis()
```
