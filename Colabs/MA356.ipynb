```python
# Primary data structure libraries
import pandas as pd
import numpy as np

# Primary plotting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv')

# Split the words in 'Descript' on the basis of commas and count the frequency of each word
words = np.concatenate(df['Descript'].str.split(',').values)
unique, counts = np.unique(words, return_counts=True)

# Prepare data for bubble plot
# Sort the sizes and labels so that the largest bubbles are plotted last
sorted_indices = np.argsort(counts)
sizes = counts[sorted_indices]
labels = unique[sorted_indices]

# Increase the figure size
fig, ax = plt.subplots(figsize=(10, 10))

# Create a scatter plot with random positions and sizes corresponding to word frequencies
x, y = np.random.rand(2, len(sizes))
scatter = ax.scatter(x, y, s=sizes, color='lightskyblue', edgecolor='black', alpha=0.5)

# Annotate each bubble with the respective word
for i in range(len(sizes)):
    # Adjust the font size based on the bubble size and filter out smaller bubbles annotaions
    fontsize = 0 if sizes[i]<1000 else (sizes[i] / np.max(sizes) *10 + 5)
    plt.text(x[i], y[i], labels[i], fontsize=fontsize, ha='center', va='center')

# Turn off the axis
plt.axis('off')

# Show the resultant graph
plt.show()
```
