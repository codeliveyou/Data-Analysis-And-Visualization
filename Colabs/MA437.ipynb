As a law enforcement officer, i would like to know where theft crimes cases happened according to sanfrancisco_crimes_data_2016.csv dataset. Plot a scatter plot to visualize.

```python
# import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# import the dataset
crimes = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv')

# retrieve theft cases
theft_cases = crimes[(crimes['Category'] == 'LARCENY/THEFT')]

# list variables to store the x and y coordinates
x=[]
y=[]

# retrieve x and y coordinates for the theft cases
for theft in theft_cases.values:
  x.append(theft[9])
  y.append(theft[10])

# Creating a figure
plt.figure(figsize=(20,20))

# first subplot
plt.subplot(2,2,1)

# scatter plot
plt.scatter(x,y)

# x label
plt.xlabel('X coords')

# y label
plt.ylabel('Y coords')

# title
plt.title('X and Y coordinates Theft Map')

plt.show()
```
