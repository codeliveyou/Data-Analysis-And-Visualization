```python
# Importing the libraries
import numpy as np
import pandas as pd

# Read the workout data from the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv")

# We filter by Data Scientist
df_DataScientist = df[df['job_title'] == ('Data Scientist')]
# We calculate the mean absolute deviation
Mean_absolute_deviation = df_DataScientist.salary.mad()
# We round up to 2 decimals the result
Mean_absolute_deviation = (np.round(Mean_absolute_deviation, decimals=2))
# We print the mean absolute deviation
print('The mean absolute deviation for the Data Scientist Salary is:', Mean_absolute_deviation)
```
