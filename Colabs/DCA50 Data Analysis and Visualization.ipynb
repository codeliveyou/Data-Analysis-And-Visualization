```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/spotify.csv")

df.hist(figsize = (10,10))
plt.title("Histogram Plot", size=20, weight='bold')
plt.show()

# Get count of artists songs to find top 10 artists
df1 = df.groupby('artist').count().sort_values(by='song_title', ascending = False)
number_of_songs = df1['song_title'][:10]
number_of_songs.plot.barh()
```
