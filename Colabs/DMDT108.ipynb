```python
# Import necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt

# Dataset path.
path = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/bigfoot_locations.csv'

# Load the dataset.
df = pd.read_csv(path)


# Timestamp column to DateTime.
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Filter for dates between 1992 and 2023.
df = df[(df['timestamp'].dt.year >= 1992) & (df['timestamp'].dt.year <= 2023)]

# Filter latitude and longitude values.
filtered_data = df[(df['latitude'] > -90) & (df['latitude'] < 90) &
                             (df['longitude'] > -180) & (df['longitude'] < 180)]
# Set background to black.
plt.style.use('dark_background')

# Create the scatter plot.
plt.figure(figsize=(12, 5))

# Set style.
plt.scatter(filtered_data['longitude'], filtered_data['latitude'], alpha=0.8, s=30,
            c='lime', linewidth=0.5, marker='x')

# Add title.
plt.title('Bigfoot sightings by location', fontsize=10, color='lime')

# Add labels.
plt.xlabel('Longitude', fontsize=10, color='lime')
plt.ylabel('Latitude', fontsize=10, color='lime')
plt.xticks(color='lime')
plt.yticks(color='lime')

# Show plot.
plt.tight_layout()
plt.grid(color='lime')
plt.show()
```
