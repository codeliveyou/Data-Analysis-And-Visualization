## 1. Data Manipulation

**Prompt**:

If I provide you with the data, can you plot a scattering distribution for the UFO sightings between 2000 and 2007, using the latitude and longitude for the US? Use distinct color for each year.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Sample data
data_file = 'ufo_sighting_data.csv'

# Convert the data into a DataFrame
df = pd.read_csv(data_file, low_memory=False)

# Convert the "Date_time" column to a proper datetime format with 'coerce' option
df['Date_time'] = pd.to_datetime(df['Date_time'], format='%m/%d/%Y %H:%M', errors='coerce')

# Filter the data for UFO sightings in the United States (US) between 2000 and 2007
us_sightings = df[(df['country'] == 'us') & (df['Date_time'].dt.year >= 2000) & (df['Date_time'].dt.year <= 2007)].copy()

# Convert 'longitude' and 'latitude' columns to numeric using .loc
us_sightings.loc[:, 'longitude'] = pd.to_numeric(us_sightings['longitude'], errors='coerce')
us_sightings.loc[:, 'latitude'] = pd.to_numeric(us_sightings['latitude'], errors='coerce')

# Create the scatter plot with equal aspect ratio
plt.figure(figsize=(12, 8))

# Get a list of distinct years
years = us_sightings['Date_time'].dt.year.unique()

# Create a color map with distinct colors for each year
color_map = cm.tab10

# Iterate through each year, extract the data, and plot as a separate color group
for i, year in enumerate(years):
    data_by_year = us_sightings[us_sightings['Date_time'].dt.year == year]
    plt.scatter(data_by_year['longitude'], data_by_year['latitude'], color=color_map(i), alpha=0.5, s=15, label=str(year))

# Set the name of axis and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('UFO Sightings in the United States (2000 - 2007)')

# Enable grid lines on the scatter plot
plt.grid(True)

# Set the axis limits for latitude and longitude
plt.xlim(-180, 180)  # Longitude range is -180 to 180
plt.ylim(-90, 90)    # Latitude range is -90 to 90

# Add the labels for each year at the top right
for i, year in enumerate(years):
    data_by_year = us_sightings[us_sightings['Date_time'].dt.year == year]

plt.legend(title='Year', loc='upper left', bbox_to_anchor=(1, 1))

# Set the aspect ratio to equal
plt.axis('equal')

# Adjust the positions of subplots or plot elements within the figure
plt.tight_layout()

plt.show()
```
