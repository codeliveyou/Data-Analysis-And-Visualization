```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the csv file.
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv")

# Store the data
type_crime_count = df['Category'].value_counts()

# Create the plot and set the plot size
plt.figure(figsize=(17, 6))

# Plot the values
crime_plot = type_crime_count.plot.bar(type_crime_count.index)

# For loop to display the value over the bars
for i, v in enumerate(type_crime_count):
    crime_plot.text(i, v + 0.20, str(v), ha='center', rotation = 45)

# Set the title and labels
plt.title('Crime count by type of crime')
plt.xlabel('Type of crime')
plt.ylabel('Count')

# Display
plt.plot()
```
