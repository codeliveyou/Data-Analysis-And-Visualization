```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

df = pd.read_csv("/content/HRDataset_v14.csv")

df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'])

# Identify terminated employees
terminated_df = df[df['Termd'] == 1]

# Calculate tenure for terminated employees
terminated_df['TenureAtTermination'] = (terminated_df['DateofTermination'] - terminated_df['DateofHire']).dt.days

# Create a histogram to visualize tenure distribution for terminated employees
plt.figure(figsize=(10, 6))
plt.hist(terminated_df['TenureAtTermination'], bins=20, color='red', edgecolor='black')
plt.title('Tenure Distribution of Terminated Employees')
plt.xlabel('Tenure (days)')
plt.ylabel('Number of Terminated Employees')
plt.grid(True)
plt.show()

# Print summary statistics of tenure for terminated employees
print("Summary Statistics of Tenure for Terminated Employees:")
print("Mean Tenure:", terminated_df['TenureAtTermination'].mean())
print("Median Tenure:", terminated_df['TenureAtTermination'].median())
print("Minimum Tenure:", terminated_df['TenureAtTermination'].min())
print("Maximum Tenure:", terminated_df['TenureAtTermination'].max())
```
