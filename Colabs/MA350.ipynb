```python
# Primary data structure libraries
import pandas as pd
import numpy as np

# Primary plotting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv')

# Count the number of crimes for each district
district_counts = df['PdDistrict'].value_counts()

# Create a circle for the center of the plot to make the pie look like a donut
center_circle = plt.Circle((0,0),0.70,fc='white')

# Explode the slice with the highest crime rate
explode = [0.1 if i == district_counts.idxmax() else 0 for i in district_counts.index]

# Define colors
colors = plt.cm.viridis(np.linspace(0, 1, len(district_counts.index)))

# Create the pie chart
fig, ax = plt.subplots(figsize=(10,10))
ax.pie(
    district_counts,
    labels=district_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode,
    colors=colors,
    pctdistance=0.85,
    shadow=True
)

# Draw the donut chart
fig.gca().add_artist(center_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Add a title
plt.title('Proportion of crimes by district', fontsize=20)

# Show the resultant graph
plt.show()
```
