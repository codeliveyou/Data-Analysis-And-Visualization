Hello Bard, from the data set "Global_YouTube_Statistics.csv" I want to make a new column ('created_date') that joins the columns 'created_year' and 'created_month' and then I want 3 bar charts in the same plot. One bar chart next to the other. One that has the years when more youtube channels were created. Another one that has the months in which more youtube channels were created and the last one that has the top 10 of the 'created_date'. All of them with the bars in decreasing order. The bars should be in gold color and the background of all graphs should be in silver color. Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/global_youtube_statistics.csv", encoding='latin-1')

# Calculate the year and month in which most YouTube channels were created
most_common_years = df['created_year'].dropna().astype(int).value_counts().index
most_common_months = df['created_month'].dropna().value_counts().index

# Drop rows with NaN values in 'created_year' or 'created_month'
df = df.dropna(subset=['created_year', 'created_month'])

# Create a new column that combines 'created_year' and 'created_month'
df['creation_date'] = df['created_year'].astype(int).astype(str) + '-' + df['created_month']

# Calculate the top 10 'creation_date'
top_10_creation_dates = df['creation_date'].value_counts().head(10).index

# Set up the size of the plot
plt.figure(figsize=(15, 5))

# Set the background color of the whole plot to blue
plt.rcParams['axes.facecolor'] = 'silver'

# Bar plot for the years with the most YouTube channels
plt.subplot(1, 3, 1)
plt.bar(np.arange(len(most_common_years)), df['created_year'].dropna().astype(int).value_counts(), color = 'gold')
plt.xticks(np.arange(len(most_common_years)), most_common_years, rotation='vertical')
plt.title('Top Years for YouTube Channel Creation')
plt.xlabel('Year')
plt.ylabel('Number of Channels')
plt.grid()

# Bar plot for the months with the most YouTube channels
plt.subplot(1, 3, 2)
plt.bar(np.arange(len(most_common_months)), df['created_month'].dropna().value_counts(), color = 'gold')
plt.xticks(np.arange(len(most_common_months)), most_common_months)
plt.title('Top Months for YouTube Channel Creation')
plt.xlabel('Month')
plt.ylabel('Number of Channels')
plt.grid()

# Bar plot for the top 10 'creation_date'
plt.subplot(1, 3, 3)
plt.bar(np.arange(len(top_10_creation_dates)), df['creation_date'].value_counts().head(10), color = 'gold')
plt.xticks(np.arange(len(top_10_creation_dates)), top_10_creation_dates, rotation='vertical')
plt.title('Top 10 Creation Dates for YouTube Channels')
plt.xlabel('Creation Date')
plt.ylabel('Number of Channels')
plt.grid()

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()
```
