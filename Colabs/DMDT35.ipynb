```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/Largest-Malls.csv")
#Excluding under-construction malls
df1 = df[df['Year opened'].str.isdigit()]
#Find malls built after 2010
df1 = df1[df1['Year opened'].astype(int) > 2010]
#Remove ',' from the string values
df1['Shops'] = df1['Shops'].str.replace(',','').astype(int)
#Find greater than 300 shops and filter top 10
df1 = df1[df1['Shops'] > 300].sort_values(by=['Shops'], ascending=False)
df1 = df1[['Mall','Country','Shops']].head(10)
print(df1)
```
