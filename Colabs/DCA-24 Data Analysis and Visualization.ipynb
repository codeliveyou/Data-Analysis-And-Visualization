```python
# Importing the libraries
import pandas as pd

# Read the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv")

# Filtering the dataframe for having European residence
dfEUResidents = df[df['employee_residence'] == ('ES')]
# Filtering the dataframe for only Machine Learning Engineer positions
dfEuropeML = dfEUResidents[dfEUResidents['job_title'] == ('Machine Learning Engineer')]
# Calculating the average salary
avg_salary = dfEuropeML.salary.mean().round(0).astype(int)
# Printing the result
print('The average salary for a Machine Learning engineer in Europe is: $', avg_salary)
```
