```python
# Primary data handling libraries
import pandas as pd
import numpy as np

# Primary ploting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/vehicle_insurance_claim_fraud_detection.csv')

def handle_value(value: any) -> int:
  """
  This function handles a variety of input value types and tries to convert them into an integer.

  Parameters: value (any): The input value to be handled. It can be of any data type.

  Returns: int: The converted integer value. If conversion is not possible, returns 0.
  """

  # Check if the input value is of type string
  if isinstance(value, str):
      # If the string is 'none', 'None', or empty, return 0
      if value in ['none', 'None', '']:
          return 0
      else:
          # Split the string into individual words (string may contain the int value)
          for val in value.split():
              # Try to convert each word into an integer
              try:
                  return int(val)
              # If conversion fails, continue to the next word
              except:
                  pass
  else:
      # If the input value is not a string, it's already a number so try to convert it directly into an integer
      return int(value)
  # If all attempts to convert the input value into an integer fail, return 0
  return 0

# Convert 'PastNumberOfClaims' to numerical if 'PastNumberOfClaims' contains any non-numerical values (like 'none')
df['PastNumberOfClaims'] = df['PastNumberOfClaims'].map(handle_value)

# Get columns to plot
x = df['Age']
y = df['DriverRating']
z = df['PastNumberOfClaims']

# Set the fig size
fig = plt.figure(figsize=(10, 8))

# Create a colormap based on the y values
colors = plt.cm.viridis(y / y.max())

# Create a 3D scatter plot
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c = colors)

# Set the labels
ax.set_xlabel('Age')
ax.set_ylabel('Driver Rating')
ax.set_zlabel('Past Number Of Claims')

# Set the Title
plt.title('3D Scatter Plot of Age, Driver Rating, and Past Number Of Claims')

# Show the resultant chart
plt.show()
```
