**Prompt:**

I would like a dispersion graph of the vehicles that exceed 250 g/Km of CO2 emission where we have Fuel Consumption Comb (L/100 km) versus Fuel Consumption Comb (mpg), I need the points to be red and the grid to be present dotted in blue, the graph must have a square format. For this dataset co2_emissions.csv.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Creating dataset
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/co2_emissions.csv")

# Set Filter and select sub dataset
filter_emissions= df['CO2 Emissions(g/km)'] > 250

# Define subplot objects and type scatter plot
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x=df['Fuel Consumption Comb (L/100 km)'], y=df['Fuel Consumption Hwy (L/100 km)'], c='red')

# Make grid visible
ax.grid(linestyle='dotted', linewidth=2,color='#4260f5')

# Define title and axes labels
plt.title("Scatter Plot of Fuel Consumption Comb (L/100 km) vs Fuel Consumption Hwy (L/100 km)" )
plt.xlabel("Fuel Consumption Combined'")
plt.ylabel("Fuel Consumption Highway")

# Show graph
plt.show()
```
