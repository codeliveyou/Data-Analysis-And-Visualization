```python
# Filter startups that have crossed their first milestone
# Assuming if 'age_first_milestone_year' is not NaN, then they have crossed their first milestone
milestone_data = data[data['age_first_milestone_year'].notna()]

# Group by 'category_code' and count the number of startups in each category
category_counts = milestone_data['category_code'].value_counts()

# Plot the pie chart
plt.figure(figsize=(10, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Startups Crossing their First Milestone by Category")
plt.show()
```
