Hello Bard, from the dataset 'heart.csv' I want a scatter plot of RestingBP vs Cholesterol. I want the points to be diamond shaped, different color according to their RestingECG type and with alpha = 0.5. The color gamma of the dots obtained from 'viridis'. I want the axes and the title to have the style {fontsize=16, fontweight='bold', fontname='Liberation Serif', color='pink'}. I want the x and y axes to be salmon colored and with linewidth = 1.8 and the axis ticks with style {fontsize=12, fontweight='bold', fontname='Liberation Serif', color='white'}. That the plot has grid with style {color='salmon', linestyle='--', linewidth=0.5}.  And that the figure has background color dimgray. Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/heart.csv")

# Get unique values in the 'RestingECG' column and assign viridis colors
unique_ecg_values = df['RestingECG'].unique()
color_map = plt.get_cmap('viridis', len(unique_ecg_values))

# Create a scatter plot
fig = plt.figure(figsize=(10, 6))
for i, value in enumerate(unique_ecg_values):
    subset = df[df['RestingECG'] == value]
    plt.scatter(subset['RestingBP'], subset['Cholesterol'],
                marker='D', color=color_map(i), label=value, alpha=0.5)

# Title, legend and grid
plt.title('Scatter Plot of RestingBP vs. Cholesterol according to RestingECG', fontsize=16,
          fontweight='bold', fontname='Liberation Serif', color='pink')
plt.legend()
plt.grid(color='salmon', linestyle='--', linewidth=0.5)

# Y-axis style
plt.ylabel('Cholesterol', fontsize=16, fontweight='bold', fontname='Liberation Serif', color='pink')
plt.axvline(0, color='salmon', linewidth=1.8, )
plt.yticks(fontsize=12, fontweight='bold',
                   fontname='Liberation Serif',
                   color='white')

# X-axis style
plt.xlabel('RestingBP', fontsize=16, fontweight='bold', fontname= 'Liberation Serif', color='pink')
plt.axhline(0, color='salmon', linewidth=1.8, )
plt.xticks(fontsize=12, fontweight='bold',
                   fontname='Liberation Serif',
                   color='white')

# Set the background color of the figure to dimgray
fig.set_facecolor('dimgray')

# Show the plot
plt.show()
```
