```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/yahoo_stock.csv")

# Convert the 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Calculate Moving Average (MA-26) and Exponential Moving Average (EMA-13)
data['MA-50'] = data['Adj Close'].rolling(window=50).mean()
data['MA-200'] = data['Adj Close'].rolling(window=200).mean()

# Create a new column 'status' for candlestick fill color
data['status'] = np.where(data['Close'] > data['Open'], 'increase', 'decrease')

# Create a Bokeh figure
p = figure(x_axis_type="datetime", title="Interactive Candlestick Chart of Yahoo with MA-50 and MA-200")

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Plot the candlestick chart
inc = data.Close > data.Open
dec = data.Open > data.Close
w = 12 * 60 * 60 * 1000  # width of each candlestick

p.segment('Date', 'High', 'Date', 'Low', source=source, color="black")
p.vbar('Date', w, 'Open', 'Close', source=source, fill_color=factor_cmap('status', palette=Spectral6, factors=['increase', 'decrease']), line_color="black")

# Add Moving Average (MA-26) and Exponential Moving Average (EMA-13) lines
p.line(x='Date', y='MA-50', source=source, line_width=2, color='orange', legend_label='MA-50')
p.line(x='Date', y='MA-200', source=source, line_width=2, color='green', legend_label='MA-200')

# Customize the hover tool
hover = HoverTool()
hover.tooltips = [("Date", "@Date{%F}"),
                  ("Open", "@Open{0.00}"),
                  ("Close", "@Close{0.00}"),
                  ("High", "@High{0.00}"),
                  ("Low", "@Low{0.00}"),
                  ("Adj Close", "@{Adj Close}{0.00}"),
                  ("Volume", "@Volume{0,0}")]
hover.formatters = {'@Date': 'datetime'}
p.add_tools(hover)

# Styling
p.xaxis.major_label_orientation = np.pi / 4
p.grid.grid_line_alpha = 0.3

# Show the interactive plot
output_notebook()
show(p)
```
