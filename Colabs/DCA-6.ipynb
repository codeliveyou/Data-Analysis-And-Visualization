```python
#Importing the libraries
import pandas as pd

# Reading the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv")
# We filter the dataframe and only retrieve the ones being data scientist
dfDataScientist = df[df['job_title'] == ('Data Scientist')]
# We calculate the mean salary
average_salary = dfDataScientist.salary.mean()
# We round the average salary so it is not with decimals
average_salary = int(np.round(average_salary, decimals=0))
# We print the average salary
print('The average salary of a Data Scientist is: ', average_salary)
```
