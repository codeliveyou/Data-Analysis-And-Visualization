```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure

# Read csv
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/spotify.csv")

# Dropping duplicates
df = df.drop_duplicates()
# Dropping NaN
df = df.dropna()
# Dropping columns
df = df.drop(columns = 'Unnamed: 0')

# Create Bokeh figure
p = figure(title = "Danceability vs. Tempo", tools="pan,wheel_zoom,box_zoom,reset,hover,save",
           tooltips="Danceability = @danceability - Tempo = @tempo, Port = @tempo",
           toolbar_location="below")

p.circle('danceability','tempo',source=df,fill_alpha=0.2, size=10)

# Format axes
p.xaxis.axis_label = 'Danceability'
p.yaxis.axis_label = 'Tempo'

output_notebook()
show(p)
```
