Hi Bard. From the dataset "Heart_Disease_Classification_Dataset.csv" I want a 3d scatter plot of size (12, 12) having as x-axis 'kcm', as y-axis 'troponin' and as z-axis 'glucose'. I want the points that are negative 'class' to be red, with triangles looking down and alpha 0.5. Positive class points should be green, diamond-shaped and alpha 0.2.
Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/heart_disease_classification_dataset.csv")

# Create figure
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111, projection = '3d')

# Data
neg = df[df['class'] == 'negative']
pos = df[df['class'] == 'positive']

x1 = neg['kcm']
y1 = neg['troponin']
z1 = neg['glucose']

x2 = pos['kcm']
y2 = pos['troponin']
z2 = pos['glucose']

# Points with form and style requirements
ax.scatter(x1, y1, z1, c='r', marker='v', label='negative', alpha = 0.5)
ax.scatter(x2, y2, z2, c='g', marker='D', label='positive', alpha = 0.2)

# Lables, title and legend
ax.set_xlabel('kcm')
ax.set_ylabel('troponin')
ax.set_zlabel('glucose')
plt.title("3D Scatter Plot relating kcm, proponin and glucose")
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
