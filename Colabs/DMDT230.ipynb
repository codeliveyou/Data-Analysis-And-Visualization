```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)

df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['month'] = df['date'].dt.month
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
df['weekday'] = df['date'].map(lambda x: day_order[x.weekday()] )


# Count the number of matches played by month and weekday
month_weekday_counts = df.groupby(['month', 'weekday']).size().unstack(fill_value=0).reindex(columns=day_order)

# Plot the data as a stacked bar chart
fig, ax = plt.subplots()
month_weekday_counts.plot.bar(stacked=True, ax=ax)

# Set the title and axis labels
ax.set_title('Number of matches played by month and weekday')
ax.set_xlabel('Month')
ax.set_ylabel('Matches played')

plt.show()
```
