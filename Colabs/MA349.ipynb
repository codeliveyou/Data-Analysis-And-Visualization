```python
# Primary data structure libraries
import pandas as pd
import numpy as np

# Primary plotting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv')

# Convert the 'Date' column into datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group by 'Date' and 'Category' and count the number of incidents
grouped = df.groupby(['Date', 'Category']).size().reset_index(name='counts')

# Pivot the data so that each category is a separate column
pivot = grouped.pivot(index='Date', columns='Category', values='counts')

# Fill NA values with 0
pivot = pivot.fillna(0)

# Get the top 5 categories with the highest total crime counts
top_categories = pivot.sum().nlargest(5).index

# Select only the top categories
pivot = pivot[top_categories]

# Plot the data
pivot.plot(kind='area', stacked=False, figsize=(10,6))

# Set the title and labels
plt.title('Trend of top 5 crime categories over time')
plt.xlabel('Date')
plt.ylabel('Number of crimes')

# Add a legend
plt.legend()

# Show the resultant graph
plt.show()
```
