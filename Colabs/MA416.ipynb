```python
df = data.copy()
# Filter data based on loan_status
approved = df[df[' loan_status'] == ' Approved']
rejected = df[df[' loan_status'] == ' Rejected']

# Create a scatter plot
plt.figure(figsize=(10, 6))

# Plot approved loans in one color
plt.scatter(approved[' income_annum'], approved[' loan_amount'], color='green', label='Approved', alpha=0.6)
# Plot rejected loans in another color
plt.scatter(rejected[' income_annum'], rejected[' loan_amount'], color='red', label='Rejected', alpha=0.6)

# Setting labels, title, and legend
plt.xlabel('Annual Income')
plt.ylabel('Loan Amount')
plt.title('Relationship between Annual Income and Loan Amount')
plt.legend()

# Display the plot
plt.show()
```
