```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool
from bokeh.io import push_notebook
from bokeh.layouts import column
from bokeh.transform import linear_cmap
from bokeh.palettes import Spectral11

# Convert the data to a DataFrame
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/walmart_sales_data.csv")

# Convert the 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

# Filter data for Store 1
store_data = data[data['Store'] == 1].copy()

# Convert 'Date' column to datetime type
store_data['Date'] = pd.to_datetime(store_data['Date'])

# Calculate relative changes
store_data['Relative_Sales'] = store_data['Weekly_Sales'] / store_data['Weekly_Sales'].iloc[0]
store_data['Relative_Temperature'] = store_data['Temperature'] / store_data['Temperature'].iloc[0]

# Create a Bokeh figure
p = figure(title="Relative Change in Weekly Sales vs Temperature for Store 1",
           x_axis_label='Date', y_axis_label='Relative Change', x_axis_type='datetime')

# Add line glyphs for relative sales and temperature
sales_line = p.line(x='Date', y='Relative_Sales', source=store_data, line_width=2, line_color=Spectral11[0], legend_label='Relative Sales')
temp_line = p.line(x='Date', y='Relative_Temperature', source=store_data, line_width=2, line_color=Spectral11[1], legend_label='Relative Temperature')

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Date", "@Date{%F}"), ("Relative Sales", "@Relative_Sales"), ("Relative Temperature", "@Relative_Temperature")]
hover.formatters = {'@Date': 'datetime'}
p.add_tools(hover)

# Add legend
p.legend.title = 'Legend'
p.legend.label_text_font_size = '10pt'
p.legend.location = "top_left"

# Show the plot inline in Jupyter Notebook
output_notebook()
show(p)
```
