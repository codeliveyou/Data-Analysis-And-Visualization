From the dataset "digital_currency_time_series.csv" I want an interactive line graph made with the bokeh library on the 'close_USD' over time. I want it to have hoover data 'date', 'volume' and 'close_USD'.

Include the code in the answer.

```python
# Import libraries
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
output_notebook()

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/digital_currency_time_series.csv")

# Prepare dataframe
df.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df = df.sort_values('date', ascending=True)

# Source data
source = ColumnDataSource(df)

# Hoover data
hover=HoverTool(tooltips=[
    ( 'Date', '@date'),
    ( 'Volume', '@volume' ),
    ( 'close usd','@close_USD'),
])

# Create figure
p = figure(tools=[hover],
           title="Digital currency price time series",toolbar_location="below",
           x_range=df['date'])

# Line, axis
p.line(x='date', y='close_USD', line_color='green', line_width=1.3, source=source)
p.xaxis.axis_label = "Date"
p.yaxis.axis_label = "Price"

# Ocultar los ticks del eje x
p.xaxis.major_label_text_font_size = "0pt"

# Show plot
show(p)
```
