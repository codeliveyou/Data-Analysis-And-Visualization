Bart, I need to compare a 3D helix with variable radius in simple and cubic growth ratio with a uniform spiral also in 3D, I would like them to have a title and a legend.

```python
# Import libraries
from sympy import *
from sympy.plotting import plot3d_parametric_line

# Setting symbolic variables
r, t, x, y = symbols('r t x y')
r=3

# Creating parametrizations
alpha1 = [t* cos(t), t*sin(t), t]
alpha2 = [r*cos(t), (r**3)*sin(t), t]

# Plotting parametrizations
plot3d_parametric_line((*alpha1,  (t, 0, 12*pi)) ,
                       (*alpha2,  (t, 0, 12*pi)),
                       legend=True,
                       title="comparing parameterization")
```
