```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file
# df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/mall_customers.csv")

# Dropping duplicates
df = df.drop_duplicates()
# Dropping NaN
df = df.dropna()
# Dropping columns
df = df.drop(columns = 'Unnamed: 0')

# Plotting scatter plot
plt.figure(figsize = (14,8))
plt.scatter(x='danceability', y='tempo', data=df)
plt.xlabel('Danceability')
plt.ylabel('Tempo')

plt.title('Scatter Plot: Danceability vs. Tempo')

plt.show()
```
