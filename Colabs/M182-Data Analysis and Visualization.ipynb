**All the code together**

Show me a density estimate plot for the Enrollment column in the dataset Top-Largest-Universities.csv for each range of Rank as follow: 0-10, 10-30, 30-50, and rank major than 50.


```python
#import the libraries
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/top_largest_universities.csv")

# creates the column "Rank Interval" to filter and set up the different ranges of Rank
df.loc[df["Rank"] <= 10,"Rank Interval"] = "[0 - 10]"
df.loc[((df["Rank"] > 10) & (df["Rank"] <= 30)),"Rank Interval"] = "(10 - 30]"
df.loc[((df["Rank"] > 30) & (df["Rank"] <= 50)),"Rank Interval"] = "(30 - 50]"
df.loc[df["Rank"] > 50,"Rank Interval"] = "> 50"

# converting data into wide-format
data_wide = df.pivot(columns='Rank Interval', values='Enrollment')

# calling density() to make
# multiple density plot
data_wide.plot.density(figsize = (12, 7),linewidth = 1)

# show the plot
plt.show()
```
