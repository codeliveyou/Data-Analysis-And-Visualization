```python
# Primary data structure libraries
import pandas as pd
import numpy as np

# Primary plotting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv')

# Splitting the descriptions to extract crimes
crimes = np.concatenate(df['Descript'].str.split(',').values)

# Calculating the frequencies
crimes, counts = np.unique(crimes, return_counts=True)

# Sorting by frequency
sorted_indices = np.argsort(counts)[::-1]
crimes = crimes[sorted_indices]
counts = counts[sorted_indices]

# Getting the top 30 crimes
top_30_crimes = crimes[:30]
top_30_counts = counts[:30]

# Creating the bar chart
fig, ax = plt.subplots(figsize=(10, 10))

# Plot the data
bars = ax.barh(range(30), top_30_counts, color='lightblue')

# Add the labels inside the bars on the right end
for i, bar in enumerate(bars):
    x = bar.get_width()
    y = bar.get_y()
    ax.text(x-1000, y+0.4, str(x), va='center')

# Set labels and title
ax.set_xlabel('Frequency')
ax.set_yticks([i for i in range(30)])
ax.set_yticklabels(top_30_crimes)
plt.title('Top 30 Crimes from Descriptions')

# Reverse the order of bars to be top down
plt.gca().invert_yaxis()

# Show the resultant graph
plt.show()
```
