```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)
# parsing dates
df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['Year'] = df['date'].dt.year
# Filter the DataFrame to include only the last 5 years of data
last_year = df['Year'].max()
df_last_5_years = df[df['Year'] >= last_year - 4]

# Count the number of neutral and non-neutral matches
neutral_counts = df_last_5_years['neutral'].value_counts()

# Plot the data as a pie chart
fig, ax = plt.subplots()
ax.pie(neutral_counts.values, labels=['Neutral', 'Non-neutral'],autopct='%1.1f%%')

# Set the title
ax.set_title('Number of neutral and non-neutral matches (last 5 years)')

plt.show()
```
