Based on the following file loan_approval_dataset.csv, I want you to calculate the sum of the different assets values plus the annual income and plot them versus the loan amount, using different colors for approved and rejected loans

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Read the dataset from the csv file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/loan_approval_dataset.csv")

# Calculate the total value of the assets
df['total_assets_values']=df[' income_annum']+df[' residential_assets_value']+df[' commercial_assets_value']+df[' luxury_assets_value']+df[' bank_asset_value']

# Group the data by the loan status
grouped_data=df.groupby(' loan_status')

# Create colors and markers lists for using for each value of loan status
colors=['g','r']
markers=['v','P']

# Create the plot
fig, ax = plt.subplots(figsize=(6, 6))

# Iterate over the loan statuses to generate each plot
for (key, group),index in zip(grouped_data,range(len(grouped_data))):
    ax.scatter(group['total_assets_values'],group[' loan_amount'], color=colors[index], marker=markers[index], label=key)

# Set the plot title
plt.title("Total assests value vs Loan amount")

# Set the axis label
ax.set_xlabel('total assets values')
ax.set_ylabel('loan amount')

# Add the legends
plt.legend()
```
