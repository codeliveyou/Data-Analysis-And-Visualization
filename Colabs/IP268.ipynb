```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Load your dataset
data = pd.read_csv("/content/dataset_olympics.csv")

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=data['Height'], y=data['Weight'], name=data['Name'], city=data['City']))

# Create a Bokeh figure
p = figure(title="Scatter Plot of Height vs. Weight", x_axis_label="Height", y_axis_label="Weight", height=400, width=600, tools="pan,box_zoom,reset")

# Add scatter plot glyphs
p.circle(x='x', y='y', source=source, size=8, color='blue', alpha=0.6)

# Add HoverTool
hover = HoverTool()
hover.tooltips = [("Name", "@name"), ("City", "@city"), ("Height", "@x"), ("Weight", "@y")]
p.add_tools(hover)

# Customize the plot
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Show the plot
output_notebook()
show(p)
```
