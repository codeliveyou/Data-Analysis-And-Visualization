```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/spotify.csv")
# Drop duplicates
df = df.drop_duplicates()
# Drop NaN
df = df.dropna()
# Plotting heatmap without values
plt.figure(figsize = (14,8))
correlation_matrix = np.array(df.corr())
plt.imshow(correlation_matrix, cmap='autumn', interpolation='nearest')
# Add colorbar, add title and display map
plt.colorbar()
plt.title("Heatmap with color bar")
plt.show()
```
