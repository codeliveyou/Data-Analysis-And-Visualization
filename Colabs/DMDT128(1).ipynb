```python
# Import necesary libraries
import pandas as pd
import matplotlib.pyplot as plt




# Dataset path.
path = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/dataset_olympics.csv'

# Load the dataset.
df = pd.read_csv(path)

# Set time intervals.
intervals = [(start, start + 12) for start in range(1896, 2016, 12)]

# Set background to black.
plt.style.use('dark_background')

# Set up Subplots.
fig, axes = plt.subplots(nrows=3,ncols=10,figsize=(12, 6))

# Select features.
features = ['Age', 'Height', 'Weight']

# Set plot colors.
colors = ['blue', 'red', 'green']

# Create subplots.
for row, feature, color in zip(axes, features, colors):
    for ax, (start, end) in zip(row, intervals):
        subset = df[(df['Year'] >= start) & (df['Year'] < end)]
        ax.hist(subset[feature].dropna(),bins=30,color=color,edgecolor='black')
        ax.set_title(f'{start}-{end-1}')
        ax.set_xlabel(feature)
        ax.set_ylabel('Number of Athletes')
        ax.set_xticks([])
        ax.set_yticks([])

# Set titles.
axes[0,0].set_ylabel('Age')
axes[1,0].set_ylabel('Height (cm)')
axes[2,0].set_ylabel('Weight (kg)')

# Show plot.
plt.grid()
plt.tight_layout()
plt.show()
```
