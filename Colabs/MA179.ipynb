**Prompt:**

Bard, with the following dataset heart.csv I would like you to make me a pie chart of the number of cases studied in the dataset, grouped by age. I need the label to be inside the graphic.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

# Load data from csv into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/heart.csv")

# Count Heart Disease cases group by Age and set names like labels
df = df.groupby('Age')['HeartDisease'].count()
labels = df.index

# Define subplot objects and type pie diagram, with label in the graph
fig, ax = plt.subplots(figsize=(12, 12))
ax.pie(df, labels=labels, labeldistance = 0.95)

# Define title
plt.title("Distribution of heart disease studied by age")

# Show graph
plt.show()
```
