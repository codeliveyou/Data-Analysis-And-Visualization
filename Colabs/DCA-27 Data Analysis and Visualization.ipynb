```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv")

# Calculating how many are for each job
df.groupby('job_title').size()
# Filtering only those belonging to the US
df1 = df[df['employee_residence'] == ('US')]
# Calculating the mean salary of each of them
df_job_title = df1.groupby('job_title')['salary'].mean().round(0).astype(int).reset_index()
# Sorting the values by descending
df_job_title = df_job_title.sort_values(by=('salary'),ascending=False)
# Taking the first ten
df_job_title = df_job_title.head(10)
# Printing the values
print(df_job_title)
```
