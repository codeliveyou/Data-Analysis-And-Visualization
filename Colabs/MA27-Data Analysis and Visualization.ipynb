````python
# import necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt



# dataset path.
path = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/spotify.csv'

# Loading the dataset to a pandas dataframe.
df = pd.read_csv(path)

# Splits numerical variables into discrete and continuous variables.
discrete_variables = [column for column in df.columns if df[column].dtype == 'int64']
continuous_variables = [column for column in df.columns if df[column].dtype == 'float64']

# Split categorical variables.
categorical_variables = df.select_dtypes(include=['object']).columns.tolist()

# print output.
print(f"categorial variables: {categorical_variables}, discrete variables: {discrete_variables}, continuos variables: {continuous_variables}")

# Create subplots for each of the continuous variables.
len_continuous_vars = len(continuous_variables)
fig, axes = plt.subplots(nrows=len_continuous_vars, ncols=1, figsize=(8, 20))

# Density plot for each continuous variable
for i, col in enumerate(continuous_variables):
    ax = axes[i]
    df[col].plot(kind='density', ax=ax)
    ax.set_xlabel(col)
    ax.set_ylabel("Density")
    ax.set_title(f"Density Plot for {col}")

# Show plot.
plt.tight_layout()
plt.show()
````
