For the data set 'digital_currency_time_series.csv' I want a line graph showing the volume traded over time. I want the volume line to be blue with width = 0.3. I also want a gold horizontal line with the average volume. I want to paint green the area between the daily volume and the average volume when the daily volume is greater than average and red when the daily volume is less than average. I want the background of the chart to be gold.

Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/digital_currency_time_series.csv")

# Prepare dataframe
df.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df = df.sort_values('date', ascending=True)

# calculate average volume
average_volume = df['volume'].mean()

# Create figure
fig = plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['volume'], label='Volume', color='blue', linewidth = 0.3)
plt.axhline(y=average_volume, color='gold', label='Average Volume')

# Fill the area between the volume average and the volume line
plt.fill_between(df['date'], df['volume'], average_volume, where=(df['volume'] >= average_volume), interpolate=True, color='green', alpha=0.4)
plt.fill_between(df['date'], df['volume'], average_volume, where=(df['volume'] < average_volume), interpolate=True, color='red', alpha=0.4)

# Legend, labels and title
plt.legend()
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Volume traded over time')

# Set the ticks on the x-axis only for the beginning of each year.
df['year'] = pd.to_datetime(df['date']).dt.strftime('%Y')
unique_years = df['year'].unique()
first_dates_by_year = []
for year in unique_years:
    first_date = df[df['year'] == year]['date'].iloc[0]
    first_dates_by_year.append(first_date)

plt.xticks(first_dates_by_year, rotation=45, ha='right')

# Set the background color
fig.patch.set_facecolor('gold')

# Show the plot
plt.show()
```
