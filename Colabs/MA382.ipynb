For the dataset "digital_currency_time_series.csv" I want in the same plot pie charts showing the total percentage of days that the price in usd closed above and below the average of 'close_USD'. And the same for each year. I want the days above the average to be 'mediumturquoise' and the days below the average to be 'lightcoral'. I want the slices to have wedgeprops = {'linewidth': 0.3, 'edgecolor': 'black'} and the background of the figure to be gold.

Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/digital_currency_time_series.csv")

# Prepare dataframe
df.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year

# Calculate the average of 'close_USD'
average_close = df['close_USD'].mean()

# Calculate how many days closed above and below the average
df['days_above_average'] = df['close_USD'] > average_close
above_count = df['days_above_average'].sum()
below_count = len(df) - above_count

# Calculate how many days per year closed above the average
days_above_by_year = pd.DataFrame(df.groupby('year')['days_above_average'].sum().reset_index())

# Create figure and axes
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

labels = ['days above avg', 'days below avg']
colors = ['mediumturquoise', 'lightcoral']

axes[0].pie([above_count, below_count], labels = labels, autopct='%1.1f%%',
            startangle=90, colors=colors,
            wedgeprops={'linewidth': 0.3, 'edgecolor': 'black'})
axes[0].set_title('Total days')

for index, row in days_above_by_year.iterrows():
    year = row['year']
    above = row['days_above_average']
    below = len(df[df['year'] == year]) - above
    axes[index +1].pie([above, below], labels=labels, autopct='%1.1f%%',
                       startangle=90, colors=colors,
                       wedgeprops={'linewidth': 0.3, 'edgecolor': 'black'})
    axes[index +1].set_title(f'Year {year}')

# Set title to the figure
fig.suptitle('Days closed above and below to close_USD average ', fontsize=16)

# Set the background color
fig.patch.set_facecolor('gold')

# Adjust spacing between graphics
plt.tight_layout()

# Show plot
plt.show()
```
