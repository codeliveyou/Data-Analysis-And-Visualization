Please make me a nice plot of the number of matches played by year  with this dataset: fifa_results.csv

```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)
# parsing dates
df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['Year'] = df['date'].dt.year
# Count the number of matches played by year
year_counts = df.groupby('Year').size()

# Plot the data
fig, ax = plt.subplots()
ax.plot(year_counts.index, year_counts.values)

# Set the title and axis labels
ax.set_title('Total number of matches played by year')
ax.set_xlabel('Year')
ax.set_ylabel('Matches played')

plt.show()
```
