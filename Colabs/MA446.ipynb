```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/yahoo_stock.csv')

# Convert the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract year from the date
df['Year'] = df['Date'].dt.year

# Create a line plot for each year
years = df['Year'].unique()

for year in years:
    yearly_data = df[df['Year'] == year]

    plt.figure(figsize=(10, 6))
    plt.plot(yearly_data['Date'], yearly_data['Open'], color='green', label='Opening Price')
    plt.plot(yearly_data['Date'], yearly_data['Close'], color='red', label='Closing Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'Trend of Opening and Closing Prices for Year {year}')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Display the plot
    plt.show()
```
