```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into data frame
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/yahoo_stock.csv')

# Configure scatter plot for market high, low, open, and close of Yahoo stocks
plt.figure(figsize = (12, 8))
plt.scatter(df['Date'], df['High'], label = 'High', color = 'red')
plt.scatter(df['Date'], df['Low'], label = 'Low', color = 'orange')
plt.scatter(df['Date'], df['Open'], label = 'Open', color = 'yellow')
plt.scatter(df['Date'], df['Close'], label = 'Close', color = 'green')

# Set title and labels for plot
plt.title('Yahoo Stock Date Over Time')
plt.xlabel('Date')
plt.ylabel('Price')

# Set x labels to be every 30 to avoid overcrowding
dates = df['Date']
plt.xticks(dates.values[::30], rotation = 90)

# Display graph
plt.legend()
plt.tight_layout()
plt.show()
```
