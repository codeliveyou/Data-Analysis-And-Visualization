```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)
# parsing dates
df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['Year'] = df['date'].dt.year

# Filter the DataFrame to include only data from the FIFA World Cup tournament without using .loc
df_world_cup = df[df['tournament'] == 'FIFA World Cup']

# Calculate the average number of goals per game for each World Cup
df_world_cup['total_goals'] = df_world_cup['home_score'] + df_world_cup['away_score']
df_world_cup['year'] = pd.DatetimeIndex(df_world_cup['date']).year
goals_per_game = df_world_cup.groupby('year')['total_goals'].mean().reset_index()

# Plot the data
fig, ax = plt.subplots()
ax.plot(goals_per_game['year'], goals_per_game['total_goals'])

# Set the title and axis labels
ax.set_title('Average number of goals per game at the FIFA World Cup')
ax.set_xlabel('Year')
ax.set_ylabel('Average goals per game')

plt.show()
```
