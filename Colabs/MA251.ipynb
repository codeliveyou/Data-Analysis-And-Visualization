**Prompt:**

Hi Bard, with the following dataset world_data_2023.csv, I need you to make three different line graphs, one next to the other, to show me the fertility ratio, the birth ratio, and the population of the 30 countries with the largest population. Please avoid empty data. The colors should be red, purple and lime respectively, I also want the names of these countries respectively.

```python
import matplotlib.pyplot as plt
import pandas as pd
import re

# Function for data non numeric transformation
def transfto_num(sentence:str)->int:
    """
    # Transforms strings containing commas into float numbers.

    # :param str: A string containing digits and commas.

    # Output: float number
    """

    #Replace comma by dot
    sentence=sentence.replace(',', '')

    # Find numbers in string and take value in fts
    num=re.findall(r"[-+]?(?:\d*\.*\d+)", sentence)[-1]

    return float(num)


# Create dataform new without nan values
df=pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/world_data_2023.csv')
newdf= df.dropna().reset_index(drop=True)


# Apply transformation
newdf['Population']= newdf['Population'].apply(transfto_num)
most_population=newdf.nlargest(30, 'Birth Rate')

# Create subplots for scatter plots
fig, ax = plt.subplots(nrows=1, ncols=3 , figsize=(16,8))

# Set lists colors and features
colors_list=['red' , 'purple' ,  'lime' ]
features_list=["Birth Rate", "Fertility Rate", "Population"]

# Create axes elements
for i in range(len(colors_list)):
  ax[i].plot("Country",features_list[i], data=most_population, color=colors_list[i])
  ax[i].tick_params(axis='x', labelrotation = 90)

# Create Labels for axes elements
for axis in ax:
    axis.set_xlabel("Countrys")
    axis.set_ylabel(r"$\psi$")
    axis.legend()

#Set layouts and leyend
fig.tight_layout()
plt.legend(loc='upper right')

#Show Plot
plt.show()
```
