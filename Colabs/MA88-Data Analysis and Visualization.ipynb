Hi Bard. From the dataset "data_science_salaries_2023.csv" make 2 bar charts on the same plot. One at the top and one at the bottom. The top one should be of the 10 'job_title' that have the highest average usd salary. Let the bars be forestgreen and the height of the bars be 0.7. The bottom graph should be of the 10 'job_title' with the lowest average usd salary. Let the bars be orangered and the height of the bars be 0.3. Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv")

# Get average salary_in_usd grouped by job_title
avg_salary = df.groupby('job_title')['salary_in_usd'].mean().reset_index()

# Get data
top_10_salary = avg_salary.sort_values(by='salary_in_usd', ascending=True).tail(10)
bottom_10_salary = avg_salary.sort_values(by='salary_in_usd', ascending=True).head(10)

# Create the first subplot
plt.subplot(2, 1, 1)

# Plot the first bar chart (the top one)
plt.barh(top_10_salary['job_title'], top_10_salary['salary_in_usd'],color='forestgreen',  height=0.7)
plt.xlabel('Average salary')
plt.ylabel('Job')
plt.title('Ten highest paid roles')
plt.grid()

# Create the second subplot
plt.subplot(2, 1, 2)

# Plot the second bar graph (the bottom one)
plt.barh(bottom_10_salary['job_title'], bottom_10_salary['salary_in_usd'], color='orangered', height=0.3)
plt.xlabel('Average salary')
plt.ylabel('Job')
plt.title('Ten least paid roles')
plt.grid()

# Adjust spacing to avoid overlapping
plt.tight_layout()

# Show the plot
plt.show()
```
