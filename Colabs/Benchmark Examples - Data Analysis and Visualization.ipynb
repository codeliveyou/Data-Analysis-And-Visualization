## 1. Data Manipulation

**Prompt 1**:

If I provide you data about Automobile models and some characteristics, can you tell me what is the model with the highest horsepower manufactured in the year 78?

```python
import pandas as pd

# Create a DataFrame representing automobile models and their horsepower
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/Automobile.csv")

df = pd.DataFrame(data)

# Filter the DataFrame to get car models produced in the year 1978 (model_year == 78)
df_1978 = df[df['model_year'] == 78]

# Find the model with the highest horsepower among car models produced in 1978
best_model_1978 = df_1978.loc[df_1978['horsepower'].idxmax()]
highest_horsepower_1978 = df_1978['horsepower'].max()

print(f"The vehicle with the highest horsepower manufactured in 1978 is {best_model_1978['name']} with {highest_horsepower_1978} horsepower.")
```

**Response (model)**:

Sure, the vehicle with the highest horsepower manufactured in 1978 is buick regal sport coupe (turbo) with 165.0 horsepower.

**Prompt 2**:

If I provide you data about Automobile models and some characteristics, can you plot the top 10 models showing the horsepower in descending order from year '78?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame representing automobile models and their horsepower
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/Automobile.csv")

df1 = pd.DataFrame(data)

# Filter the DataFrame to get car models produced in the year 1978 (model_year == 78)
df1_1978 = df1[df1['model_year'] == 78]

# Sort the DataFrame by horsepower in descending order and take the top 10 rows
top_10_models = df1_1978.sort_values(by='horsepower', ascending=False).head(10)

# Plot the horsepower for each model
plt.bar(top_10_models['name'], top_10_models['horsepower'], color='b')
plt.xlabel('Automobile Model')
plt.ylabel('Horsepower')
plt.title('Top 10 Automobile Models by Horsepower')
plt.xticks(rotation=90)
plt.grid(True, axis='y')
plt.show()
```

**Prompt 1:**

Hello Bard. From the World Cups data, make a bar plot of the total goals scored over world cups. I want the columns to be green.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt

# access the data through Pandas
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/world_cups_data.csv")

# Sort the DataFrame by 'Year'
df_sorted = df.sort_values(by='Year')

# Bar plot for the number of goals scored in each tournament
plt.figure(figsize=(10, 6))
plt.bar(df_sorted['Year'], df_sorted['total_goals_scored'], color="g")

plt.xlabel('Year')
plt.ylabel('Total Goals Scored')

plt.title('Total Goals Scored in Each World Cup')
plt.xticks(rotation=45)

# show the image
plt.show()
```

**Prompt 2**

Please from the laptops_price_dataset.csv , make a pie chart of laptop Brands, but including only the top 10 most common.

```python
# import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data from the CSV file into a pandas DataFrame
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/laptops_price_dataset.csv")

# Visualization: Pie chart of genre distribution
genre_counts = df['Brand'].value_counts()[:10]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%')

# Add title
plt.title('Status of Laptops')

# Display the pie chart
plt.show()
```

**Prompt 3**

Hello Bard. From the World Cup data, make a line plot the total goals scored over world cups. I want the line to be orange, and I want each world cup year to be specified below the x axis. Add next to each dot the number of goals scored, centered around the dot.

Include the code in the answer.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt

# access the data through Pandas
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/world_cups_data.csv")

# Line plot of total goals scored over World Cups
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['total_goals_scored'], marker='o', color='orange')

# Add labels
plt.xlabel('Year')
plt.ylabel('Total Goals Scored')

# Add title
plt.title('Total Goals Scored in Each World Cup')
plt.xticks(df['Year'], rotation=45)

# Make it a grid
plt.grid(True)

# Add number of goals scored next to each dot, centered around the dot, color orange
for x, y in zip(df['Year'], df['total_goals_scored']):
    plt.text(x, y, str(y), ha='center', va='bottom')

plt.show()
```
