```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Load the data from the CSV file into a Pandas DataFrame
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv")

# Filter for residence, job title and employment type
dfEUResidents = df[(df['employee_residence'] == ('ES')) & (df['job_title'] == ('Data Analyst')) & (df['employment_type'] == ('FT'))]

# Filter for work years between 2022 and 2023
dfEuropeFullTime = dfEUResidents[(dfEUResidents['work_year'] >= 2022) & (dfEUResidents['work_year'] <= 2023)]

# Sort by salary for the filtered years
dfEuropeFullTime = dfEuropeFullTime.sort_values(by='salary', ascending = False)
dfEuropeFullTime = dfEuropeFullTime[['work_year','salary']].head(10)

# Calculate the mean salary for the filtered years
mean_salary = dfEuropeFullTime.salary.mean().round(0).astype(int)
print(mean_salary)
```
