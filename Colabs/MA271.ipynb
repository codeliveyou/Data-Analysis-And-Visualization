```python
# Primary data structure libraries
import pandas as pd
import numpy as np

# Primary plotting library
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/bank.csv', sep=';')

# Compute the cross tabulation matrix
crosstab = pd.crosstab(df['job'], df['education'])

# Create a custom diverging colormap
cmap = mcolors.LinearSegmentedColormap.from_list("", ["red","yellow","green"])

# Create the matplotlib figure
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Generate the heatmap using imshow
im = axs[0].imshow(
    # Using percentage instead of count for better visualisation
    crosstab.div(crosstab.sum(axis=1), axis=0)*100,
    cmap=cmap,
    aspect='auto'
)
fig.colorbar(im, ax=axs[0], shrink=.5)

# Set the title
axs[0].set_title('Job and Education Level Heatmap')

# Set the ticks labels
axs[0].set_xticks(np.arange(len(crosstab.columns)))
axs[0].set_yticks(np.arange(len(crosstab.index)))
axs[0].set_xticklabels(crosstab.columns)
axs[0].set_yticklabels(crosstab.index)

# Set ylabel
axs[0].set_ylabel('Percentage')
axs[0].yaxis.set_label_position("right")

# Create a stacked bar chart
crosstab.plot(kind='bar', stacked=True, ax=axs[1])

# Add Labels
plt.xlabel('Job Categories')
plt.ylabel('Count')

# Add Title
axs[1].set_title('Job categories broken down by education level')

# Output the resultant plot
plt.tight_layout()
plt.show()
```
