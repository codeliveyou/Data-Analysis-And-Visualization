```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool
from bokeh.plotting import output_notebook
import warnings
warnings.filterwarnings('ignore')

# Read data
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/startup_data.csv")

# Create Bokeh figure
p = figure(title="Scatter Plot of First vs Last Funding Year", x_axis_label="Age at First Funding Year", y_axis_label="Age at Last Funding Year")

# Add scatter plot
p.circle(df["age_first_funding_year"], df["age_last_funding_year"], size=10, fill_alpha=0.6)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Age at first funding year", "@x"), ("Age at last funding year", "@y")]
p.add_tools(hover)

# Output the plot to an HTML file
output_file("scatter_plot.html")

# Show the plot
output_notebook()
# Show the plot
show(p)
```
