```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool
from bokeh.plotting import output_notebook
import warnings
warnings.filterwarnings('ignore')

# Read the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/laptops_price_dataset.csv")
df = df[['Brand','Final Price','RAM','Storage']]
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Get top 10 expensive brands
df = df.sort_values(by="Final Price", ascending=False)
df.rename(columns= {"Final Price": "Price"}, inplace=True)
df = df[0:10]

# # Create Bokeh figure
p = figure(x_range=df['Brand'].unique(), title="Brand vs. Price", x_axis_label="Brand", y_axis_label="Price")

# Add bar plot
p.vbar(x="Brand", top="Price", source=df, width=0.5)

# Customize x-axis labels
p.xaxis.major_label_orientation = "vertical"

# Add hover tool
hover = HoverTool()
hover.tooltips = [
    ("Brand", "@Brand"),
    ("Price", "@Price"),
    ("RAM", "@RAM"),
    ("Storage", "@Storage")]
p.add_tools(hover)

# Output the plot to HTML file
output_file("bar_plot.html")

# Show the plot
output_notebook()
# Show the plot
show(p)
```
