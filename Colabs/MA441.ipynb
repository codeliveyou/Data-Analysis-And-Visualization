```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a Pandas DataFrame
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/ip19.csv")

# Get unique regions from the dataset
regions = data['Region'].unique()

# Create a 2x3 grid for the pie charts (assuming you have 5 regions)
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Flatten the axs array for easier indexing
axs = axs.flatten()

# Loop through each region and create a pie chart
for i, region in enumerate(regions):
    # Extract data for the current region
    region_data = data[data['Region'] == region]

    # Aggregate sales for each product within the region
    aggregated_sales = region_data.groupby('Product')['Sales'].sum()

    # Create a pie chart for the current region
    axs[i].pie(aggregated_sales, labels=aggregated_sales.index, autopct='%1.1f%%', startangle=90)
    axs[i].set_title(f"Sales distribution in {region}")
    axs[i].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Remove the empty subplot position in the lower-right part
fig.delaxes(axs[-1])

# Adjust spacing
plt.tight_layout()

# Show the figure
plt.show()
```
