Based on the following file Heart_Disease_Classification_Dataset.csv, I want you to give me the code and the plot showing the higher pressure and the glucose in a 3d plot, using different color red for positive diagnosis and green for negative.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset from the csv file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/heart_disease_classification_dataset.csv")

# Create a new column with patient index
df['patient']=df.index

# Create a list of colors  and markers with the same length as the diagnosis
colors=['g','r']
markers=['v','P']

# Group the dataset by the diagnosis
grouped = df.groupby('class')

# Create a 3d plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection = '3d')

# Iterate over the diagnosis to generate each plot
for (key, group),index in zip(grouped,range(len(grouped))):
    ax.scatter(group['patient'],group['pressurehight'], group['glucose'], color=colors[index], marker=markers[index], label=key)

# Set the title
plt.title('High pressure vs glucose in heart disease')

# Set the axis label
ax.set_xlabel('patient number')
ax.set_ylabel('high pressure')
ax.set_zlabel('glucose')

# Add the legends
plt.legend()
# Show the plot
plt.show()
```
