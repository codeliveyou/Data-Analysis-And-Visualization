```python
# Code for app.py
# Import necessary libraries
import pandas as pd
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource, RadioButtonGroup
from bokeh.layouts import column

# Load the dataset
df = pd.read_csv("Automobile.csv")

# Variables that the user will be able to select for the sizes
variables = ['mpg','cylinders','displacement','horsepower']
var_names = ['MPG (Miles per Gallon)','Cylinders','Displacement','Horsepower']

# The scale of the points, you can adjust this according to your preferences
scale = 1

# Create the scatter plot
plot_title = "weight vs acceleration"
p = figure(width=800, height=600, title=plot_title)

# Define the data to be used in the scatter plot
source = ColumnDataSource(dict(
    x=df['weight'],
    y=df['acceleration'],
    sizes=(df[variables[0]]) * scale
    ))

# Create a scatter plot with the initial sizes
p.scatter('x', 'y', source=source, size='sizes' * scale)

# Define the labels
p.xaxis.axis_label = "Weight"
p.yaxis.axis_label = "Acceleration"

# Radio button group for selecting the variable to use for point sizes
select_var = RadioButtonGroup(labels=var_names,
                              active=0)

# Define a function to update the plot when the active value changes
def update_plot(attrname, old, new):
    """
    This function updates the Bokeh bar plot with a different variable defining
    the sizes of the points based on the selected value.
    """
    selected = select_var.active

    # Update the data
    source = ColumnDataSource(dict(
                                   x=df['weight'],
                                   y=df['acceleration'],
                                   sizes=(df[variables[selected]] * scale)
                                  )
                              )

    # Clear the previous plot by removing all renderers
    p.renderers = []

    # Create a scatter plot with the sizes according to the selected variable
    p.scatter('x', 'y', source=source, size='sizes' * scale)

# Call the update_plot function initially to populate the plot
update_plot(None, None, 0)

# Update the plot when a radio button is selected
select_var.on_change("active", update_plot)

# Arrange the plot
layout = column(select_var,p)

# Add the layout to the current Bokeh document
curdoc().add_root(layout)
```

You can save this code to a file named `app.py`, and then run it from the command-line like this:

`bokeh server --show app.py`
