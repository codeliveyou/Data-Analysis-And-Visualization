Bard, I need to draw a torus in such a way that the distance from the center of the circle to the axis and the radius of the circle come from the roots of the polynomial x**2-4*x+3, please the torus must have its hole oriented in the direction of the x axis.

```python
# Import libraries
from sympy import *
from sympy.plotting import plot3d_parametric_surface

# Setting symbolic variables
x, u ,v  = symbols('x u v')

# Solve equation required and generate solution list
solution_list=solve(Eq(x**2-4*x+3, 0), x)
solution_list.sort()

# Create Torus with conditions required
# Create Torus with conditions required
r, R = solution_list[0] , solution_list[1]
plot3d_parametric_surface((R+r*sin(u))*cos(v), r*cos(u), (R+r*sin(u))*sin(v),(u, -5, 5), (v, -5, 5));
```
