```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

# Read the supermarket sales data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/supermarket_sales.csv")

# Store information about products and payment methods
products = df['Product line'].unique()
payment_methods = df['Payment'].unique()

data = []
colors = ['r', 'c', 'g']

# Create 2 D array where row is product and columns is payment method
for x in payment_methods:
  count = []
  for y in products:
    count.append(len(df[(df['Product line'] == y) & (df['Payment'] == x)]))
  data.append(count)

# Set the plot size
plt.figure(figsize=(10, 10))

# Plot the stacked bar for each product
plt.bar(products, data[0], color=colors[0])
for i in range(1, len(data)):
  plt.bar(products, data[i], bottom=data[0], color=colors[i])
  for j in range(0, len(data[0])):
    data[0][j] += data[i][j]

# Create patch to store colour coding
patches = []
for i in range(0, len(data)):
  patches.append(mpatches.Patch(color = colors[i], label=payment_methods[i]))

plt.xticks(rotation=45)
plt.legend(handles=patches, title = "Modes of payment")

# Define title, label of x and y axis
plt.xlabel("Product Categories")
plt.ylabel('Number of payment transaction')
plt.title("Number payment transaction done via different modes")

plt.show()

```
