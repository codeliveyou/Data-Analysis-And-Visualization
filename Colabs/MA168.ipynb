```python
# primary data structure library
import pandas as pd

# primary plotting library
import matplotlib.pyplot as plt

# for ggplot-like style of plots
plt.style.use('ggplot')

# reading the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv')
df = df[['job_title', 'salary_in_usd']]

# extracting the minimun, maximum and average values for all job titles
y1 = df.groupby(by='job_title')['salary_in_usd'].min()
y2 = df.groupby(by='job_title')['salary_in_usd'].mean()
y3 = df.groupby(by='job_title')['salary_in_usd'].max()

# taking all job titles
x = y1.index

# since there are a lot of job titles, adjust the chart dimensions and font size accordingly
plt.rcParams.update({'font.size': 30})
fig, ax = plt.subplots(figsize=(50, 10))

# ploting the bar graphs stacked on each other
ax.bar(x, y3, color='r')
ax.bar(x, y2, color='b')
ax.bar(x, y1, color='g')

# add the legend
plt.legend(["Minimum", "Average", "Maximum"])

# Rotating X-axis labels for better view
plt.xticks(rotation = 90)

# add the title
plt.title('Minimum, Maximum and Average Salary by Job Title')

# label the axes
plt.xlabel("Job Title")
plt.ylabel("Salary in usd")

# output the resultant graph
plt.show()
```
