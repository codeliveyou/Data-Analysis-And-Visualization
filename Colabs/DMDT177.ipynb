```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/HRDataset_v14.csv")
df['DateofHire'] = pd.to_datetime(df['DateofHire'])

# Extract year from DateofHire column
df['HireYear'] = df['DateofHire'].dt.year

# Group employees by hire year and count the number of employees in each year
recruitment_trends = df['HireYear'].value_counts().sort_index()

# Create a bar plot to visualize recruitment trends
plt.figure(figsize=(10, 6))
recruitment_trends.plot(kind='bar', color='blue')
plt.title('Recruitment Trends Over Years')
plt.xlabel('Hire Year')
plt.ylabel('Number of Employees Hired')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

print("Recruitment Trends Over Years:")
print(recruitment_trends)
```
