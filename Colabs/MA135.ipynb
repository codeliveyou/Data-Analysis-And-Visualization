Hi Bard, I need a graph showing the amount of travel time per PROPOSE in hours. The dataset is 'UberDataset.csv' consists of the fields 'START_DATE', 'END_DATE' and 'PROPOSE', the date format is '%m-%d-%Y %H:%M' or '%m/%d/%Y %H:%M'.

```python
# Import librarieas
import pandas as pd

# Read dataset
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/uber_dataset.csv")

# Replace
df["START_DATE"] = df["START_DATE"].str.replace(r"[/|-]", " ")
df["END_DATE"] = df["END_DATE"].str.replace(r"[/|-]", " ")

# Cast to datetime
df["START_DATE"] = pd.to_datetime(
    df["START_DATE"], format="%m %d %Y %H:%M", errors="coerce"
)
df["END_DATE"] = pd.to_datetime(
    df["END_DATE"], format="%m %d %Y %H:%M", errors="coerce"
)

# Calculating the trip in hours
df["travel_time_hours"] = (
    (df["END_DATE"] - df["START_DATE"]).dt.total_seconds() / 60 / 60
)

# Sum of hours
df = df.groupby(by="PURPOSE")["travel_time_hours"].sum().reset_index(name="Hours")

# Sort Hours
df = df.sort_values(by="Hours", ascending=True)

# Graph
plt.figure(figsize=(12, 12))
courses = list(df["PURPOSE"].astype(str))
values = list(df["Hours"])
plt.title("Hours per purpose")
plt.pie(values, labels=courses, shadow=True, autopct="%1.1f%%", startangle=180)
```
