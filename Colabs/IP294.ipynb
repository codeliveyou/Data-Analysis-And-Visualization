```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.palettes import Category20c
from bokeh.transform import cumsum
from math import pi

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/startup_data.csv")
df.drop(columns = ['Unnamed: 6',"Unnamed: 0"],inplace=True)

# Prepare data for pie chart
counts = df['category_code'].value_counts().sort_values(ascending=False)
categories = counts.head(10)
categories_data = categories.reset_index(name='counts').rename(columns={'index': 'category'})

# Find angle for each category
categories_data['angle'] = categories_data['counts'] / categories_data['counts'].sum() * 2 * pi

# Give different colors to each catagory
df1 = pd.DataFrame()
colors = {}
palette = Category20c[10]
for i, category in enumerate(categories_data['category'].values):
    colors[category] = palette[i]

categories_data = categories_data[0:10]
categories_data['color'] = Category20c[len(categories_data)]

# Create a Bokeh figure
p = figure(width=600, height=400, title="Company Distribution by Category", toolbar_location=None,
        tools="hover", tooltips=[("Categories", "@category"), ("Categories Count", "@counts")])

# Add pie chart wedges
p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='category', source=categories_data)

output_notebook()

show(p)
```
