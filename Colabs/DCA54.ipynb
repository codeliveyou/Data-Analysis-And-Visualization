```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/world_cups_data.csv")
# Dropping duplicates
df = df.drop_duplicates()
# Dropping NaN
df = df.dropna()
# Creating separate lists for Champion and Runner-up
list1 = df['Champion'].value_counts()
list2 = df['Runner-up'].value_counts()
# Concaternating both lists to a single dataframe and creating 3 columns including country
df3 = pd.concat([list1, list2],axis=1)
df3 = df3.dropna()
df3 = df3.reset_index()
df3.rename(columns = {'index':'Country'}, inplace=True)
df3
```
