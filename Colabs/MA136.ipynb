```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount("/content/gdrive", force_remount=True)

# Load the CSV data into a DataFrame
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/Amazon_Customer_Behavior_Survey.csv")

# Filter data based on Gender and Reiview Reliability
filtered_data = df[
    (df['Gender'] == "Female") &
    (df['Review_Reliability'] == "Heavily")
]

# Group data by Age and Purchase Categories, and count occurrences
grouped_data = filtered_data.groupby(['age', 'Purchase_Categories']).size().reset_index(name='Count')

# Create a scatter plot with specific figure size
plt.figure(figsize=(18, 12))
plt.scatter( filtered_data['age'],filtered_data['Purchase_Categories'], color='red')

#Set x and y label and title
plt.xlabel('Age')
plt.ylabel('Purchase Categories')
plt.title('Purchase Categories vs Age chart for Females with Heavy review reliability')

# Set age interval to 5
plt.xticks(range(15, filtered_data['age'].max()+1, 5))

# Set left padding using subplots_adjust
plt.subplots_adjust(left=0.45, right=0.9, top=0.9, bottom=0.1)

# Show the plot
plt.tight_layout()
plt.show()
```
