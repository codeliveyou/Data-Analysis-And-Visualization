````python
#```python
from google.colab import drive
drive.mount("/content/gdrive", force_remount=True)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/bank_direct_marketing_campaigns.csv")
    marital_counts = df['marital'].value_counts()
    plt.figure(figsize=(8, 8))
    plt.pie(marital_counts, labels=marital_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
    plt.axis('equal')
    plt.title('Distribution of Marital Status')
    plt.show()

except Exception as e:
    print(f"An error occurred while loading the CSV file: {e}, please check the correct path to the CSV file.")
#```
````
