Hello Bard, from the dataset "heart.csv" I want a 3d scatter plot having as axes 'RestingBP', 'Cholesterol' and 'MaxHR'. I want the points with Sex = M to be lime colored, marker = 'D' and alpha = 0.35. Let the points having Sex = F be fuchsia colored, marker = 'o' and alpha = 0.35. Let the figure be of size (12,12) and background color grey. That the axes have the following style: {fontsize=12, fontweight='bold', fontname='Liberation Sans Narrow'}. Let one axis be red, one cyan and one yellow. That the plot has a black background and that the title has the following style: {fontsize=16, fontweight='bold', fontname='Liberation Sans Narrow', color='gold'}. Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/heart.csv")

# Create a figure and a 3D subplot
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')

# Set the background color of the entire figure
fig.patch.set_facecolor('grey')

# Separate data for male ('M') and female ('F')
x1 = df[df['Sex'] == 'M']['RestingBP']
y1 = df[df['Sex'] == 'M']['Cholesterol']
z1 = df[df['Sex'] == 'M']['MaxHR']

x2 = df[df['Sex'] == 'F']['RestingBP']
y2 = df[df['Sex'] == 'F']['Cholesterol']
z2 = df[df['Sex'] == 'F']['MaxHR']

# Scatter plot for males and females
ax.scatter(x1, y1, z1, c='lime', marker='D', label='M', alpha=0.35)
ax.scatter(x2, y2, z2, c='fuchsia', marker='o', label='F', alpha=0.35)

# Customize the appearance of the plot
ax.set_xlabel('RestingBP', fontsize=12, fontweight='bold', fontname='Liberation Sans Narrow', color='red')
ax.set_ylabel('Cholesterol', fontsize=12, fontweight='bold', fontname='Liberation Sans Narrow', color='cyan')
ax.set_zlabel('MaxHR', fontsize=12, fontweight='bold', fontname='Liberation Sans Narrow', color='yellow')

ax.tick_params(axis='x', colors='red')
ax.tick_params(axis='y', colors='cyan')
ax.tick_params(axis='z', colors='yellow')

# Change the background color of the plot
ax.set_facecolor('black')

# Change the font style
plt.title("RestingBP, Cholesterol and MaxHR according to sex", fontsize=16, fontweight='bold', fontname='Liberation Sans Narrow', color='gold')

# Display legend and adjust layout
ax.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
