```python
# Import necessary libraries
import pandas as pd

# Load CSV file into data frame
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/hrdataset_v14.csv')

# Group data frame by positions using the average of the salaries
avg_salary_by_pos = df.groupby('Position')['Salary'].mean()

# Sort the positions by average salary
avg_salary_by_pos = avg_salary_by_pos.sort_values(ascending = False)

# Create input and output data for graph
positions = avg_salary_by_pos.index
avg_salaries = avg_salary_by_pos.values

# Configure plot for positions and their respective average salary
plt.figure(figsize = (12, 8))
bars = plt.bar(positions, avg_salaries)

# Set plot titles and labels
plt.title('Average Salaries by Position')
plt.xlabel('Position')
plt.ylabel('Average Salary')

# Provide a salary inside each bar of bar graph to display specific salaries
for bar, salary in zip(bars, avg_salaries):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, f"${salary:.2f}", ha='center', va='center', rotation = 90, color='black')

# Display plot
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```
