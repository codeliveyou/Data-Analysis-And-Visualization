## 1. Data Manipulation

**Prompt**:

Can you plot the distribution of traffic for Junction 2 along the dates at 8:00 AM?

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_traffic_distribution(data_file, junction_number):
    # Read the CSV data into a pandas DataFrame
    df = pd.read_csv(data_file)

    # Convert the 'DateTime' column to datetime format
    df['DateTime'] = pd.to_datetime(df['DateTime'])

    # Filter data for the specified junction number and time (8:00 AM)
    junction_data_8am = df[(df['Junction'] == junction_number) & (df['DateTime'].dt.hour == 8)]

    # Set the 'DateTime' column as the index for the time series plot
    junction_data_8am.set_index('DateTime', inplace=True)


    # Create a time series plot
    plt.figure(figsize=(12, 6))
    plt.plot(junction_data_8am['Vehicles'], marker='o', markersize=4, color='b')
    plt.xlabel('Date and Time')
    plt.ylabel('Number of Vehicles')
    plt.title(f'Traffic Distribution for Junction {junction_number} at 8:00 AM')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/traffic_congestion.csv'
    junction_number = 2
    plot_traffic_distribution(data_file, junction_number)
```
