```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
url = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/automobile.csv'
data = pd.read_csv(url)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(data['weight'], data['mpg'], color="green")
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.title('MPG by Weight')
plt.grid(True)
plt.show()
```
