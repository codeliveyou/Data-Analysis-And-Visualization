Hello Bard. From the world data 2023 I want 4 pie charts on 'Population', 'Co2-Emissions', 'Land Area(Km2)', 'Armed Forces size' that sum the countries by hemisphere ('Latitude'<0 and 'Latitude'>0 countries) in one plot.

Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/world_data_2023.csv")

# Convert 'Population', 'Land Area(Km2)', 'Co2-Emissions', and 'Armed Forces size' columns to numeric
def convert_to_integer_without_comma(data_value):
    """
    Function to remove comma and convert to integer.

    Args:
        data_value (str or any): The input value, can be a string with a comma or any other type.

    Returns:
        int or any: If the input is a string with a comma, it returns the integer value after removing the comma.
                    If the input is of any other type, it returns the same value without any modification.
    """
    if isinstance(data_value, str):
        return int(data_value.replace(',', ''))
    else:
        return data_value

# Apply the function to the desired columns using applymap and the defined function
desired_columns = ['Population','Land Area(Km2)', 'Armed Forces size', 'Co2-Emissions']

df[desired_columns] = df[desired_columns].applymap(convert_to_integer_without_comma)

# Group by Latitude
latitude_groups = df.groupby(df['Latitude'].apply(lambda x: 'Latitude > 0' if x > 0 else 'Latitude < 0'))

# Calculate the sum for each group
sum_df = latitude_groups[['Population', 'Land Area(Km2)', 'Armed Forces size', 'Co2-Emissions']].sum()

index = ['Latitude < 0', 'Latitude > 0']
df = pd.DataFrame(sum_df, index=index)

# Plotting pie charts
labels = df.index
colors = ['gold', 'lightblue']
explode = (0.1, 0)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

for i, ax in enumerate(axs.flat):
    ax.pie(df.iloc[:, i], explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
    ax.set_title(df.columns[i])

plt.tight_layout()
plt.show()
```
