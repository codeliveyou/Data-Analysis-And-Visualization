I want to find the relationships between all the different features.Draw a heatmap of the correlation matrix to visualize the same. Please provide the code for the same.Use the provided csv file.




```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/winequality_red.csv")

# Compute the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
heatmap = plt.pcolor(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)

# Display the correlation values on the heatmap
for i in range(len(correlation_matrix)):
    for j in range(len(correlation_matrix)):
        plt.text(i + 0.5, j + 0.5, f'{correlation_matrix.iloc[i, j]:.2f}',
                 ha='center', va='center', color='black', fontsize=10)

plt.colorbar(heatmap)

# Display feature names as labels
feature_names = data.columns
plt.xticks(np.arange(0.5, len(feature_names), 1), feature_names, rotation='vertical')
plt.yticks(np.arange(0.5, len(feature_names), 1), feature_names)

plt.tight_layout()
plt.show()
```
