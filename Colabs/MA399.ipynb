**Prompt:**
Hello, for the following dataset sanfrancisco_crimes_data_2016.csv I would need you to create diagrams showing the locations of crimes such as assault in diferent colors, vandalism and car theft, with grids in red dotted lines. They must be next to each other and with appropriate labels. I would like the code in python.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm

# Load data from csv into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/sanfrancisco_crimes_data_2016.csv")

# Create Filters
filt_assault = df['Category'] == 'ASSAULT'
filt_vandalism = df['Category'] == 'VANDALISM'
filt_vehiclet = df['Category'] == 'VEHICLE THEFT'

# Create new datasets
df_assault = df.loc[filt_assault].copy()
df_vandalism = df.loc[filt_vandalism].copy()
df_vehiclet = df.loc[filt_vehiclet].copy()

# Make lists datasets and labels
list_labels=iter(['Assault', 'Vandalis', 'Vehicle Theft'])
list_df=[df_assault, df_vandalism , df_vehiclet ]

# Create color list
n=len(list_df)
color = iter(cm.cividis(np.linspace(0, 1, n)))
colors=[next(color) for i in range(n) ]

# Set subplot objects
fig, axs = plt.subplots(1,3, figsize=(12, 6))

# Create Scatters in axis
for i in range(n):
    axs[i].scatter(x=list_df[i]['X'], y=list_df[i]['Y'], color=colors[i])
    axs[i].grid(color='r', linestyle='dotted', linewidth=1)

# Put labels by axis
for axis in axs:
    axis.set_xlabel(next(list_labels))
    axis.set_ylabel('Counts')

# Show graph
plt.show()
```
