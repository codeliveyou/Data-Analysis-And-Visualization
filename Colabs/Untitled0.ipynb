````python

# Import necessary libraries
import pandas as pd

def read_data(file_path):
    """
    Read data from a CSV file and return a DataFrame.

    Parameters:
        file_path (str): Path to the CSV file.

    Returns:
        pandas.DataFrame: DataFrame containing the data from the CSV file.
    """
    return pd.read_csv(file_path)

# Read data from the CSV file "startup_data.csv"
data = read_data("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/startup_data.csv")

def get_word_matches(str1: str, str2: str) -> int:
    """
    Count the number of matching words between two strings.

    Args:
        str1 (str): The first input string.
        str2 (str): The second input string.

    Returns:
        int: The number of matching words between the two strings.
    """
    # Split the input strings into lists of words
    str1_words = str1.split()
    str2_words = str2.split()

    # Initialize the count of matching words
    matching_word_count = 0

    # Iterate through the words in str1_words and check if they exist in str2_words
    for word in str1_words:
        if word in str2_words:
            matching_word_count += 1

    return matching_word_count

# Set the minimun common words
commom_words = 1

# Calculate the count of common words between the name and city of start ups and add it as a new column
data['common_words'] = data.apply(lambda df: get_word_matches(df['name'], df['city']), axis=1)

# Filter startups with at least on common words
data_filtered = data[data['common_words'] >= commom_words]

# Chech if filtered data is empty or not
if data_filtered.empty:
  # Prepare response
  response = "There are no startups with at least %i common words in both tehir names and cities of origin" % (commom_words)
else:
  # Get the number of rows in the filtered data frame
  matched_startups = data_filtered.shape[0]

  # Prepare response
  response = "There are %i startups with at least %i common words in both tehir names and cities of origin" % (matched_startups, commom_words)

# Print response
print(response)
````

From the given csv file give me a Python code that creates a line plot that shows the number of soghtings of Bigfoot by year from 1920 to the current year

````python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

def read_data(file_path):
    """
    Read data from a CSV file and return a DataFrame.

    Parameters:
        file_path (str): Path to the CSV file.

    Returns:
        pandas.DataFrame: DataFrame containing the data from the CSV file.
    """
    return pd.read_csv(file_path)

# Read data from the CSV file "bigfoot_locations.csv"
data = read_data("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/bigfoot_locations.csv")

# Change the dtype of column timestamp to datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Get the year in a new column
data['year'] = data['timestamp'].dt.year

print(data.dtypes)
# Get the count of the number of sightings for each year
sightings_by_year = data['year'].value_counts()

# Sort the counts in ascending order by year
sightings_by_year = sightings_by_year.sort_index()

# Get the current year
current_year = pd.to_datetime('today').year

# Filter counts from 1920 to current year
sightings_by_year = sightings_by_year[(sightings_by_year.index >= 1920) & (sightings_by_year.index <= current_year)]

# Create a figure
plt.figure(figsize=(10, 6))

# Create a line plot
plt.plot(sightings_by_year, marker='o')

# Set the title and labels for the plot
plt.title('Number of Bigfoot Sightings by Year (1920 to present)')
plt.xlabel('Year')
plt.ylabel('Number of Sightings')

# Show the plot
plt.show()
````
