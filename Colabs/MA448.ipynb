```python
# Primary data handling libraries
import pandas as pd
import numpy as np

# Primary ploting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/vehicle_insurance_claim_fraud_detection.csv')

# Selecting only the numerical columns for pairplot
df_numerical = df[['Age', 'Deductible', 'DriverRating', 'WeekOfMonthClaimed']]

# Drop rows with any empty cells
df_numerical.dropna(
  axis=0,
  how='any',
  subset=None,
  inplace=True
)

# Number of variables to plot
num_vars = len(df_numerical.columns)

# Create a figure and a grid of subplots
fig, ax = plt.subplots(
  nrows=num_vars,
  ncols=num_vars,
  figsize=(12,12)
)

# Nested loop to create a scatter plot for each pair of features
for i, x in enumerate(df_numerical.columns):
  for j, y in enumerate(df_numerical.columns):
    # create scatter plot
    ax[i, j].scatter(
      df_numerical[x],
      df_numerical[y],
      edgecolors = 'white',
    )
    # if this is the last row, add x label
    if i == num_vars - 1:
      ax[i, j].set_xlabel(y)
    # if this is the first column, add y label
    if j == 0:
      ax[i, j].set_ylabel(x)

# Set Title
plt.suptitle('Pair Plot of Numerical features', fontsize=20)

# Show the resultant chart
plt.show()
```
