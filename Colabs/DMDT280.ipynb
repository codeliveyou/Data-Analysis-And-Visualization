Hey, I just uploaded a csv file. Please show me in a bar plot the durartion of each president's tearm in days, use blue bar please.

````python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

def read_data(file_path):
    """
    Read data from a CSV file and return a DataFrame.

    Parameters:
        file_path (str): Path to the CSV file.

    Returns:
        pandas.DataFrame: DataFrame containing the data from the CSV file.
    """
    return pd.read_csv(file_path)

# Read data from the CSV file "us_presidents_2.csv"
data = read_data("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/us_presidents_2.csv")

# Change the dtype of the columns start and end to datetime
data['start'] = pd.to_datetime(data['start'], format="%B %d, %Y", errors='coerce')
data['end'] = pd.to_datetime(data['end'], format="%B %d, %Y", errors='coerce')

# Get the term duration in days in a new colum
data['duration'] = (data['end'] - data['start']).dt.days

# Keep only name and duration for each president
data = data.set_index('president')
data = data[['duration']]

# Create a figure
plt.figure(figsize=(10, 6))

# Create a bar plot
data.plot(kind='bar', color='blue')

# Set the title and labels for the plot
plt.title('Number of days by President')
plt.xlabel('President')
plt.ylabel('Number of days')

# Show the plot
plt.show()
````
