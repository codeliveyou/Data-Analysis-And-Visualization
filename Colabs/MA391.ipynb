```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/pokemon_dataset.csv", index_col="Name")

# Create a list with the names of the PokÃ©mons
starters = ["Bulbasaur", "Charmander", "Squirtle"]

# Columns to select from the dataframe
columns  = ["HP", "Attack", "Defense", "SpAtk", "SpDef", "Speed"]

# Initialize an empty dictionary to store the PokÃ©mons attributes
attrs = {}
for name in starters:
    arr = df.loc[name, columns].values
    # You will need to repeat the first value at the end of the array
    attrs[name] = np.append(arr, arr[0])

# Set figure size and polar projection
fig = plt.figure(figsize =(16, 6))
ax = plt.subplot(polar = True)

# Use the Matplotlib's `thetagrid` function
lines, labels = plt.thetagrids(range(0, 360, int(360/len(columns))), (columns))

# Plot the attributes for every PokÃ©mon
theta = np.linspace(0, 2*np.pi, len(columns)+1)
for name in starters:
    ax.plot(theta, attrs[name])
    ax.fill(theta, attrs[name], alpha=0.1, label=name)

# Place the legend outside the plot
ax.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

# Display the plot
plt.show()
```
