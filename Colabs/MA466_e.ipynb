```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
url = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/automobile.csv'
data = pd.read_csv(url)

# Counting the number of cars for each number of cylinders
cylinders_count = data['cylinders'].value_counts()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.bar(cylinders_count.index, cylinders_count.values, color=["r","g","b","y","orange",])

# Set title, labels
plt.xlabel('Number of Cylinders')
plt.ylabel('Number of Cars')
plt.title('Distribution of Number of Cylinders')

# Add grid
plt.grid(True)

# Display
plt.show()
```
