```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/dataset_olympics.csv")

# Filter the DataFrame to remove rows without medals
df_with_medals = df.dropna(subset=['Medal'])

# Group data by Sport and count medals in each sport
sport_medal_counts = df_with_medals['Sport'].value_counts()

# Select the top N sports with the highest medal counts
top_n = 10 # Change this value to select a different number of top sports
top_sports = sport_medal_counts.head(top_n)

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_sports, labels=top_sports.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Equal aspect ratio ensures the pie chart is circular
plt.axis('equal')

plt.title('Distribution of Medals by Sport')
plt.legend(sport_medal_counts.index, loc='upper right')
plt.show()
```
