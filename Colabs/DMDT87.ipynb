## 1. Data Manipulation

**Prompt**:

I'll provide a CSV file containing financial time series data. I need you to plot a line plot showing the evolution in the Close price, but showing those Close prices where "Mercury" is mentioned in the "Text" column. Also, share with me the code in Python.

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/6mon_Nifty_Clean_text - 6mon_Nifty_Clean_text.csv'
df = pd.read_csv(data_file)

# Convert the "Date" column to a proper datetime format
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")

# Function to extract the keyword from the text
def extract_keyword(text, keyword):
    match = re.search(r'\b' + re.escape(keyword) + r'\b', text, re.IGNORECASE)
    return match.group(0) if match else None

# Determine the keyword to search for
keyword_to_search = "Mercury"

# Create a new column to store the isolated keyword from Text
df["Isolated_Keyword"] = df["Text"].apply(lambda x: extract_keyword(x, keyword_to_search))

# Filter rows where the keyword is found
df_with_keyword = df[df["Isolated_Keyword"].notnull()]

# Plot the evolution of "Close" price with a marker for the keyword
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Close"], label="Close Price")
plt.scatter(df_with_keyword["Date"], df_with_keyword["Close"],
            color="red", marker="o", label=f"Keyword '{keyword_to_search}' Found")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.title(f"Evolution of Close Price with Isolated Keyword '{keyword_to_search}' Marker")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
