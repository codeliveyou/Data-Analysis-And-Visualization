Hi Bard, I have two datasets 'FIFA_results' and 'FIFA_goalscorers.csv' with the fields 'date', 'tournament' and 'scorer'. I want a pie chart that shows the percentage of goals per player per FIFA World Cup match in the last 10 years where Argentina played, only the top 10 players are shown.

```python
# Imports
import pandas as pd

# Read dataset
try:
    df_1 = pd.read_csv(
        "/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/FIFA_results.csv"
    )
    df_2 = pd.read_csv(
        "/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/FIFA_goalscorers.csv"
    )
    df = pd.merge(df_1, df_2, on=["date", "home_team", "away_team"], how="inner")
except Exception as e:
    print(e)
    pass

# Cast to datatime
df["date"] = pd.to_datetime(df["date"])

# Max Year
max_year = df["date"].dt.year.max()

# Filter years.
df = df[(df["date"].dt.year >= max_year - 10)]

# Filter tournament
df = df[df["tournament"] == "FIFA World Cup"]

# Filter team Argentina
df = df[(df["home_team"] == "Argentina") | (df["away_team"] == "Argentina")]

# Grouped by
df_group_by = df.groupby(by=["scorer"])["scorer"].agg(["count"]).reset_index()

# Column renaming
df_group_by = df_group_by.rename(columns={"count": "total_goals_scorer"})

# Ordered
df_group_by = df_group_by.sort_values(by=["total_goals_scorer"], ascending=False)

# TOP 10
df_group_by = df_group_by.head(10)
# Plot
plt.figure(figsize=(10, 6))
plt.pie(
    df_group_by["total_goals_scorer"],
    labels=df_group_by["scorer"],
    autopct="%1.1f%%",
    startangle=140,
)
plt.title("Percentage of goals per player per FIFA World Cup match in the last 10 years in which Argentina played")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
