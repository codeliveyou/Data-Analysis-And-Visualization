Having the dataset 'yahoo_stock.csv', I want a graph showing the Closing price and the exponential moving averages of 3 and 7, in monthly period, taking only the last 24 months.

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta

# Read dataset
try:
    df = pd.read_csv(
        "yahoo_stock.csv"
    )

except Exception as e:
    print(e)
    pass

# Cast to datatime
df["Date"] = pd.to_datetime(df["Date"])

# Create the monthly period
df["Month"] = df["Date"].dt.to_period("M")

# Group by month
df = df.groupby("Month")["Close"].mean().reset_index(name="Close")

# Max date
max_date = df["Month"].max()

# Filter date.
min_date = max_date - 12*2
df = df[ (df["Month"] >= min_date)]

# Ordered
df = df.sort_values(by=["Month"], ascending=True)

# Calculating the exponential moving average
df['EMA7'] = df['Close'].ewm(span=7, adjust=False).mean()
df['EMA3'] = df['Close'].ewm(span=3, adjust=False).mean()

# Create Month_str field
df["Month"] = df["Month"].dt.strftime("%Y-%m")

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['EMA7'],linestyle='--', color='green', label='Close')
plt.plot(df['Month'], df['EMA3'], linestyle='--', color='red', label='Close')
plt.plot(df['Month'], df['Close'], marker='o', linestyle='-', color='blue', label='Close')

plt.title(f"{min_date} to {max_date}")
plt.xlabel('Month')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```
