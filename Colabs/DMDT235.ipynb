```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)
# parsing dates
df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['Year'] = df['date'].dt.year

# Filter the data to only include matches played during the 1990s
df_90s = df[(df['Year'] >= 1990) & (df['Year'] < 2000)]

# Count the number of matches played in each tournament and select the top 20
top_tournaments = df_90s['tournament'].value_counts().nlargest(20)

# Plot the data
fig, ax = plt.subplots()
ax.bar(top_tournaments.index, top_tournaments)

# Set the title and axis labels
ax.set_title('Number of matches played in different types of games during the 1990s')
ax.set_xlabel('Tournament')
ax.set_ylabel('Matches played')

plt.xticks(rotation=90)

plt.show()
```
