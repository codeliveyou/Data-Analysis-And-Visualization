```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/automobile.csv")

# Dropping duplicates
df = df.drop_duplicates()
# Dropping NaN
df = df.dropna()
# Filtering the makes that give high mileage
df1 = df[df['mpg'] > df['mpg'].mean()]
#Sort values in descenging order
df1 = df1.sort_values(by=['mpg'], ascending=False)
# Top 10 vehicles that give highest mileage
dfTop10 = df1[:10]
# Plotting the miles per gallon and horsepower as group bars
x = np.arange(len(dfTop10))
y1 = dfTop10['mpg'].tolist()
y2 = dfTop10['horsepower'].tolist()
width = 0.4
# Set figure size
plt.figure(figsize=(10, 6))
plt.bar(x-0.2, y1, width, color='magenta')
plt.bar(x, y2, width, color='purple')
plt.xticks(x, dfTop10['name'].tolist())
plt.xticks(rotation=90)
# Setting labels for the bars
plt.xlabel("Make")
plt.ylabel("Measure")
#Setting legends
plt.legend(["Miles per gallon", "HorsePower"])
plt.show()
```
