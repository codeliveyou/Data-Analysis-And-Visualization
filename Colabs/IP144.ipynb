```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool, Legend
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Read the dataset
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/walmart_sales_data.csv")

# Convert the 'Date' column to datetime with the appropriate format
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

# Filter data for Store 1
store_1_data = data[data['Store'] == 1]

# Sort the filtered data by date
store_1_data = store_1_data.sort_values('Date')

# Create a Bokeh ColumnDataSource
source = ColumnDataSource(store_1_data)

# Create a color map for holidays
holiday_cmap = linear_cmap(field_name='Holiday_Flag', palette=Viridis256, low=0, high=1)

# Create the figure
p = figure(title='Weekly Sales Over Time (Store 1)', x_axis_label='Date', y_axis_label='Weekly Sales',
           x_axis_type='datetime', sizing_mode='stretch_width')

# Add a line for weekly sales
line = p.line(x='Date', y='Weekly_Sales', source=source, line_width=2, line_color='gray')

# Add hover tool for displaying information on hover
hover = HoverTool()
hover.tooltips = [("Date", "@Date{%d-%m-%Y}"), ("Weekly Sales", "@Weekly_Sales{$0,0}")]
hover.formatters = {'@Date': 'datetime',}
p.add_tools(hover)

# Add a legend
legend = Legend(title="Holiday")
p.add_layout(legend, 'right')

# Show the holiday legend using a custom legend field
p.circle(x='Date', y='Weekly_Sales', source=source, size=8, color=holiday_cmap,
         legend_field='Holiday_Flag', fill_alpha=0.6)

# Show the plot
show(p)
output_notebook()  # For displaying the plot in a Notebook
```
