**Prompt:**
Bard, I need a pie chart with the dataset "CO2 Emissions.csv" that shows me the average number of cylinders per make of vehicle, on a light but not white background. I need the first element to be highlighted going one tenth out of the graph.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

# Change chart style to get dark background
style.use('Solarize_Light2')

# Load data into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/co2_emissions.csv")

# Count absences group by department and set names like labels
df1 = df.groupby('Make')['Cylinders'].mean()
labels = df1.index

# Create list for explode
list_explod=[0 for i in range(len(labels))]
list_explod[0]=0.1

# Define subplot objects and type pie diagram
fig, ax = plt.subplots(figsize=(18,12))

# Generate Pie Graph with explode 0.1 for first element
df1.plot.pie(y=df1.index,
           shadow=False,
           explode=list_explod,
           startangle=90,
           autopct='%1.1f%%')

# Define title
plt.title("Cylinders Mean by Make")

# Show graph
plt.show()
```
