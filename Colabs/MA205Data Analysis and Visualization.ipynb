Create a horizontal bar chart showing the top cities with largest malls.Use cyan color for the graph.Provide the code for the same.Use the provided csv file.




```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/largest_malls.csv")

# Extract numeric value from Gross leasable area (GLA)
data['GLA Numeric'] = data['Gross leasable area (GLA)'].str.replace(',', '').str.extract('(\d+)').astype(int)

# Group data by city and calculate the sum of Gross Leasable Area (GLA) for each city
city_gla_sum = data.groupby('City (metropolitan area)')['GLA Numeric'].sum()

# Sort cities by GLA in descending order
sorted_cities = city_gla_sum.sort_values(ascending=False)

# Create a horizontal bar chart
plt.figure(figsize=(10, 8))
sorted_cities.plot(kind='barh', color='cyan')
plt.title('Top Cities with Largest Malls')
plt.xlabel('Total Gross Leasable Area (GLA)')
plt.ylabel('City')
plt.tight_layout()

# Show the horizontal bar chart
plt.show()
```

Sure, here is the horizontal bar chart showing the top cities with largest malls.
