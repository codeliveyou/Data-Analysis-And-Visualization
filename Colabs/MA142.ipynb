Hi Bard, I have a dataset 'ufo_sighting_data.csv' which has a date field 'date_documented' with the format MM/DD/YYYY and the field 'UFO_shape', I need 6 histogram graphs of the most viewed UFO_shape of ufos to be made.

```python
import pandas as pd

# Dataset loading
df = pd.read_csv(
    "/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/ufo_sighting_data.csv"
)
df.head()

# Field format is changed to datetime
df["date_documented"] = pd.to_datetime(df["date_documented"], format="%m/%d/%Y")
# unicos
df_UFO_shape = df.groupby("UFO_shape").size().reset_index(name="counts")
df_UFO_shape = df_UFO_shape.sort_values(by="counts", ascending=False)
df_unique_UFO_shape = list(df_UFO_shape["UFO_shape"].unique())[:6]

# A subplot is created for all the charts
fig, axes = plt.subplots(nrows=len(df_unique_UFO_shape), figsize=(12, 12))

for i, item in enumerate(df_unique_UFO_shape):
    df_UFO_shape = df[df["UFO_shape"] == item]
    # Plot is created
    axes[i].hist(df_UFO_shape["date_documented"], bins=10, edgecolor="black")
    axes[i].grid(True)
    # Title is added and displayed
    axes[i].set_title(f" Histogram of ufo {item} type by date")

# Adjust and display plot
plt.tight_layout()
plt.show()
```
