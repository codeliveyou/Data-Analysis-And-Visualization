```python
from google.colab import drive
drive.mount("/content/gdrive", force_remount=True)
import pandas as pd
from scipy.stats import ttest_ind

try:
    # Read the workout data from the CSV file
    df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/loan_approval_dataset.csv")
    df.head()
    self_employed_scores = df[df[' self_employed'] == ' Yes'][' cibil_score']
    non_self_employed_scores = df[df[' self_employed'] == ' No'][' cibil_score']

    mean_self_employed = self_employed_scores.mean()
    mean_non_self_employed = non_self_employed_scores.mean()

    t_statistic, p_value = ttest_ind(self_employed_scores, non_self_employed_scores, equal_var=False)

    print(f"Average cibil_score for self-employed applicants: {mean_self_employed:.2f}")
    print(f"Average cibil_score for non-self-employed applicants: {mean_non_self_employed:.2f}")
    print(f"T-statistic: {t_statistic:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < 0.05:
      print("There is a significant difference between the average cibil_scores of self-employed and non-self-employed applicants.")
    else:
      print("There is no significant difference between the average cibil_scores of self-employed and non-self-employed applicants.")
except Exception as e:
    print(f"An error occurred while loading the CSV file: {e}, please check the correct mount of the shared folder")
```
