```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def remove_transform_column(feature: str , symn: str, error_rise: str):
    """
    # Transforms string object from pandas series containing digits and symbols into float numbers.

    # :param feature: A string with the column to transform.
    # :param symn: A string with symbolt to remove.
    # :param error_rise: String with option por action after invalid raise error.

    # Output: float number
    """
    newdf[feature] = pd.to_numeric(newdf[feature].str.replace(symn, ''),
                                errors=error_rise)


# Create dataform new without nan values
df=pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/world_data_2023.csv')
newdf= df.dropna().reset_index(drop=True)

# Select features and create leabels
features_list=['Population', 'Physicians per thousand', 'Urban_population']
features_labels=['Population', 'Physicians', 'Urban_population']

# Apply transformation, remove comma and transform numeric values columns
remove_transform_column(features_list[0], ',' ,'coerce' )
remove_transform_column(features_list[2], '%' ,'coerce' )

# Create new dataset with 25 Countrys with smallest urban population
lest_landarea=newdf.nsmallest(25, 'Urban_population')

# Create new column 'Physicians_pop' in dataset with number of Physicians
lest_landarea['Physicians_pop'] =(round(lest_landarea[features_list[0]]*lest_landarea['Physicians per thousand']/1000))

# Set figure size
plt.figure(figsize=(16,10))

# setting positions for graphic requirements
n = len(lest_landarea['Country'])
x = np.arange(n)

# Creating lineplot and activation colors
plt.plot(x, lest_landarea['Physicians_pop'], color = "g",
                    marker = '^',
                    markerfacecolor = 'red',
                    markeredgecolor = 'black')

# Set labels in rotation required
plt.xticks(x, lest_landarea['Country'], rotation=45)

# Set title
plt.title('Physicians by country for the 25 least urban populated')

# Set Grid
plt.grid()

# Show Plot
plt.show()
```
