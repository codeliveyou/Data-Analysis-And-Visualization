Hi Bard, I have a dataset 'ufo_sighting_data.csv' that has a date field 'date_documented' with the format MM/DD/YYYY, I need you to show me those between the years 2008 to 2023, group it by UFO_shape and show the amount per UFO_shape in a pie chart.

```python
import pandas as pd

# Dataset loading
df = pd.read_csv(
    "/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/ufo_sighting_data.csv"
)
df.head()

# Field format is changed to datetime
df["date_documented"] = pd.to_datetime(df["date_documented"], format="%m/%d/%Y")

# Filtering by date
df = df[
    (df["date_documented"].dt.year >= 2008) & (df["date_documented"].dt.year <= 2023)
]

# Grouped by UFO_shape
df = df.groupby("UFO_shape").size().reset_index(name="counts")

# Sort
df = df.sort_values(by="counts", ascending=False)

# Plot is created
plt.figure(figsize=(11, 11))
plt.pie(
    df["counts"], labels=df["UFO_shape"], shadow=True, autopct="%1.1f%%", startangle=180
)

# Title is added and displayed
plt.title("Number of UFO shape")
plt.legend()
plt.show()
```
