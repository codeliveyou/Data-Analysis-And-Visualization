```python
import pandas as pd

try:
    df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/content/avocado.csv")
    df['Date'] = pd.to_datetime(df['Date'])
    seasons = {
    'Winter': [12, 1, 2],   # December, January, February
    'Spring': [3, 4, 5],    # March, April, May
    'Summer': [6, 7, 8],    # June, July, August
    'Fall': [9, 10, 11]     # September, October, November
    }
    def assign_season(month):
      for season, months in seasons.items():
        if month in months:
          return season

    df['Season'] = df['Date'].dt.month.apply(assign_season)
    average_prices_by_season = df.groupby('Season')['AveragePrice'].mean()
    print(average_prices_by_season)


except Exception as e:
    print(f"An error occurred while loading the CSV file: {e}, please check the correct path to the CSV file.")
```
