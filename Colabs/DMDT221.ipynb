```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)

# Calculate the mean home score by date
mean_home_score = df.groupby('date')['home_score'].mean()

# Plot the data
fig, ax = plt.subplots()
ax.plot(mean_home_score.index, mean_home_score.values)

# Set the title and axis labels
ax.set_title('Average home score over time')
ax.set_xlabel('Date')
ax.set_ylabel('Average home score')

plt.show()
```
