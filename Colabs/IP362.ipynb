```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool
from bokeh.plotting import output_notebook
import warnings
warnings.filterwarnings('ignore')

# Read data
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/startup_data.csv")

# Calculate the number of funding rounds
df["num_funding_rounds"] = df["funding_rounds"]

# Group by the number of funding rounds and status and count occurrences
grouped = df.groupby(["num_funding_rounds", "status"]).size().reset_index(name="count")

# Remove duplicates in the "num_funding_rounds" column
grouped = grouped.drop_duplicates(subset=["num_funding_rounds"])

# Create Bokeh figure
p = figure(x_range=grouped["num_funding_rounds"].astype(str), title="Number of Funding Rounds vs. Status", x_axis_label="Number of Funding Rounds", y_axis_label="Count")

# Add bar plot
p.vbar(x="num_funding_rounds", top="count", source=grouped, width=0.5)

# Customize x-axis labels
p.xaxis.major_label_orientation = "vertical"

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Number of funding rounds", "@num_funding_rounds"), ("Count", "@count")]
p.add_tools(hover)

# Output the plot to HTML file
output_file("bar_plot.html")

# Show the plot
output_notebook()
# Show the plot
show(p)
```
