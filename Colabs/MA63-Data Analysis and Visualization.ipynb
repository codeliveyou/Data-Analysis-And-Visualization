Hi Bard. From the loan_approval_dataset.csv make a 3d scatter plot having as x-axis 'loan_term', y-axis 'cibil_score' and z-axis 'loan_amount'. Let the points with loan_amount greater than the mean be blue and triangle-shaped upwards. Let points with loan_amount less than the mean be red and triangle-shaped downward.
Include the code in the answer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

# Read the workout data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/loan_approval_dataset.csv")

# Separate data
avg_loan_amount = df[' loan_amount'].mean()

x1 = df[df[' loan_amount']> avg_loan_amount][' loan_term']
y1 = df[df[' loan_amount']> avg_loan_amount][' cibil_score']
z1 = df[df[' loan_amount']> avg_loan_amount][' loan_amount']

x2 = df[df[' loan_amount']< avg_loan_amount][' loan_term']
y2 = df[df[' loan_amount']< avg_loan_amount][' cibil_score']
z2 = df[df[' loan_amount']< avg_loan_amount][' loan_amount']

# Create 3D Scatter Plot
fig = plt.figure(figsize=(11,11))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(x1, y1, z1, c='b', marker='^', label='upper')
ax.scatter(x2, y2, z2, c='r', marker='v', label='lower')

# Labels, title and legends.
ax.set_xlabel('loan_term')
ax.set_ylabel('cibil_score')
ax.set_zlabel('loan_ammount')
plt.title("Loan amount according to loan term and cibil score")
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
