```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into data frame and use 'latin1' encoding since there is a UnicodeDecodeError
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/co2_emissions.csv')

# Isolate data frame to make of vehicles and their CO2 emissions, then group them by make
emission_by_make = df[['Make','CO2 Emissions(g/km)']].groupby('Make')

# Create a list for input data of boxplot
data_for_boxplots = []

# Append each data set of a car make to the input data list
for make, emissions in emission_by_make:
  data_for_boxplots.append(emissions['CO2 Emissions(g/km)'])

# Configure boxplot with CO2 emissions of each make
plt.figure(figsize = (12, 8))
plt.boxplot(data_for_boxplots, vert = False)

# Set labels for box plot
plt.title('CO2 Emissions by Make of Vehicle')
plt.ylabel('Vehicle Makes')
plt.xlabel('CO2 Emissions')

# Set y ticks for each make and their respective CO2 emissions
plt.yticks(range(1, len(emission_by_make) + 1), emission_by_make.groups.keys())

# Display plot
plt.tight_layout()
plt.show()
```
