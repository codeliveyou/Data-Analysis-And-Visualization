Hi Bard, I have two datasets 'FIFA_results' and 'FIFA_goalscorers.csv' with the fields 'date', 'tournament' and 'scorer'. I want a pie chart that has the percentage of goals per team in the last 'Copa America'.

```python
# Imports
import pandas as pd

# Read dataset
try:
    df_1 = pd.read_csv(
        "/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/FIFA_results.csv"
    )
    df_2 = pd.read_csv(
        "/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/FIFA_goalscorers.csv"
    )
    df = pd.merge(df_1, df_2, on=["date", "home_team", "away_team"], how="inner")
except Exception as e:
    print(e)
    pass

df.head()

# Cast to datatime
df["date"] = pd.to_datetime(df["date"])

# Filter tournament
df = df[df["tournament"] == "Copa América"]

# Max Year
max_year = df["date"].dt.year.max()

# Filter years.
df = df[(df["date"].dt.year == max_year)]

# Grouped by boals
df_group_by = (
    df.groupby(by=["team"])["team"]
    .agg(["count"])
    .reset_index()
)

# Column renaming
df_group_by = df_group_by.rename(
    columns={"count": "count_goals_team"}
)

# Ordered
df_group_by = df_group_by.sort_values(by=["count_goals_team"], ascending=False)

# TOP 10
df_group_by = df_group_by.head(10)

# Plot
plt.figure(figsize=(10, 6))

plt.pie(
    df_group_by["count_goals_team"],
    labels=df_group_by["team"],
    autopct='%1.1f%%',
    startangle=140
)
plt.title(
    "Percentage of goals per team in the last Copa America"
)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
