```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
data = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/loan_approval_dataset.csv')

# Filter data for Approved loan status
approved_data = data[data[' loan_status'] == ' Approved']

# Create a scatter plot
plt.figure(figsize=(14, 10))
plt.scatter(approved_data[' income_annum'], approved_data[' loan_amount'])

#Set Label of x and y axis and title of scatter plot
plt.xlabel('Amount of loan')
plt.ylabel('Annual Income')
plt.title('Annual Income vs. Amount of loan for Approved Loans')

# Show the plot
plt.tight_layout()
plt.show()
```
