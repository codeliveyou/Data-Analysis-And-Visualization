```python
# import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt

# import dataset
vehicles = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/co2_emissions.csv')

# create a color list for every data point as per the 'Make' attribute
c_map = dict(zip(vehicles['Vehicle Class'].unique(), plt.cm.tab10.colors))
colors = vehicles['Vehicle Class'].map(c_map)

# create a figure
plt.figure(figsize=(10,10))

# first subplot plot
plt.subplot(1,2,1)

# engine size and co2 emission scatter plot
plt.scatter(vehicles['Engine Size(L)'],vehicles['CO2 Emissions(g/km)'])

# title
plt.title('Engine Size vs CO2 Emissions')

# x label
plt.xlabel('Engine Size(L)')

# y label
plt.ylabel('CO2 Emissions(g/km)')

# layout adjustment to prevent overlapping
plt.tight_layout()

# display the graph
plt.show()
```
