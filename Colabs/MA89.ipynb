```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load csv file into a dataframe
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/brentoilprices.csv')

# Configure and create scatter plot
plt.figure(figsize = (30,15))
plt.scatter(df.index, df['Price'], color = 'blue', label = 'Oil Price')

# Create a line of best fit
coefficients = np.polyfit(df.index, df['Price'], 1)
polynomials = np.poly1d(coefficients)
x = np.arange(len(df))
plt.plot(x, polynomials(x), color = 'red', label = 'Line of Best Fit')

# Set labels and title of plot
plt.xlabel('Dates')
plt.ylabel('Oil Price')
plt.title('Oil Prices vs. Time')
plt.legend()

# Customize the values on the X-axis to be with their respective dates
# Have it show every 50 since labels will too scrunched up otherwise
plt.xticks(df.index[::50], df['Date'][::50], rotation=90)

# Display plot
plt.tight_layout()
plt.show()
```
