```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the telecom dataset
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/comcast_telecom_complaints_data.csv", parse_dates=['Date'])

# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot 1: Histogram of complaints by city
df['City'].value_counts().head(10).plot(kind='bar', ax=axes[0], color='c')
axes[0].set_title("Top 10 Cities with Highest Complaints")
axes[0].set_xlabel("City")
axes[0].set_ylabel('Number of Complaints')
axes[0].tick_params(axis='x', rotation=45)

# Plot 2: Line graph of Complaints over time
monthly_complaints = df.groupby(df["Date"].dt.to_period('M')).size()
monthly_complaints.plot(ax=axes[1], marker='o')
axes[1].set_title("Customer Complaints Over Time")
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Number of Complaints')
axes[1].tick_params(axis='x', rotation=45)

# Plot 3: Bar plot of complaint statuses
df['Status'].value_counts().plot(kind='bar', ax=axes[2], color='g')
axes[2].set_title('Distribution of Complaints by Status')
axes[2].set_xlabel('Status')
axes[2].set_ylabel('Number of Complaints')
axes[2].tick_params(axis='x', rotation=0)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
