```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
data = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/walmart_sales_data.csv')

# Group the data by 'Store' and calculate the sum of 'Weekly_Sales' for each store
total_sales_per_store = data.groupby('Store')['Weekly_Sales'].sum()

# Create a bar chart
plt.figure(figsize=(10, 6))
total_sales_per_store.plot(kind='bar', color='blue')
plt.title('Total Sales per Store')
plt.xlabel('Store')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```
