I want a pie chart representation for the percentaje of matches played on each month with the information given in this dataset: fifa_results.csv

```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)
# parsing dates
df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['month'] = df['date'].dt.month

# Count the number of matches played by month
month_counts = df['month'].value_counts()

# Plot the data as a pie chart
fig, ax = plt.subplots()
ax.pie(month_counts.values, labels=month_counts.index,autopct='%1.1f%%')

# Set the title
ax.set_title('Percentage of matches played by month')

plt.show()
```
