```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# Load the data from the CSV file into a Pandas DataFrame
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv")

# Filter for residence, job title and employment type
dfEUResidents = df[(df['employee_residence'] == ('ES')) & (df['job_title'] == ('Data Analyst')) & (df['employment_type'] == ('PT'))]

# Filter for work years between 2020 and 2022
dfEuropePartTime = dfEUResidents[dfEUResidents['work_year'].between(2020,2022)]

# Filter for work salary larger than 10000
dfEuropePartTime = dfEuropePartTime[dfEuropePartTime['salary']>10000]

# Calculate the mean salary for the previous conditions
mean_salary = dfEuropePartTime.salary.mean().round(0).astype(int)
print(mean_salary)
```
