Having the dataset 'yahoo_stock.csv', the EMA, SMA, WMA and Close of the last 60 days.

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta
import numpy as np

# Read dataset
try:
    df = pd.read_csv("yahoo_stock.csv")
except Exception as e:
    print(e)
    pass

# Cast to da  tatime
df["Date"] = pd.to_datetime(df["Date"])


# Max date
max_date = df["Date"].max()

# Filter date.
min_date = max_date - timedelta(days=30)
df = df[(df["Date"] >= min_date)]

# Ordered
df = df.sort_values(by=["Date"], ascending=True)

period = 5
# Calculating SMA
df["SMA"] = df["Close"].rolling(window=period).mean()

# Calculating EMA
df["EMA"] = df["Close"].ewm(span=period, adjust=False).mean()

# Calculating WMA
weights = list(range(1, period + 1))
df["WMA"] = (
    df["Close"]
    .rolling(window=period)
    .apply(lambda close: np.dot(close, weights) / sum(weights))
)

# Plot
plt.style.use("dark_background")
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["SMA"], linestyle="--", color="green", label="SMA")
plt.plot(df["Date"], df["EMA"], linestyle="--", color="red", label="EMA")
plt.plot(df["Date"], df["WMA"], linestyle="--", color="yellow", label="WMA")
plt.plot(
    df["Date"], df["Close"], marker="o", linestyle="-", color="blue", label="Close"
)

plt.title(f"{min_date.strftime('%Y-%m-%d')} to {max_date.strftime('%Y-%m-%d')}")
plt.grid(True)
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
