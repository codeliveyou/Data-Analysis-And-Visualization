```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the CO2 Emission data from the CSV file and load it into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/co2_emissions.csv")

# Store unique classes of vehicle
Vehical_class = df['Vehicle Class'].unique()

# Group vehical class and aggregate sum for CO2 Emissions
co2_emissions = df.groupby('Vehicle Class')['CO2 Emissions(g/km)'].sum()

# Sort the values as we need top 10 classes
co2_emissions = co2_emissions.sort_values(ascending=False)

# We pick top emittor
top10_emittor = co2_emissions[:10].copy()

# Remaining classes we will put together in others category
others = co2_emissions[10:].sum()
top10_emittor['Others'] = others

# Plot the pie chart
fig, ax = plt.subplots()
ax.pie(top10_emittor.values, labels=top10_emittor.index, shadow=True, wedgeprops = {"edgecolor" : "white", 'linewidth': 1}, autopct='%1.1f%%')

plt.title("Vehical classes contributing to CO2 Emission")
plt.show()
```
