## 1. Data Manipulation

**Prompt**:

I have a csv file related to traffic congestion for different junctions along the time. Can you bring me the heatmap that visualizes the overall traffic patterns across the junction 1 for all the Januarys along the years at every hour?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/traffic_congestion.csv'

# Read the CSV data into a pandas DataFrame
df = pd.read_csv(data_file)

# Convert 'DateTime' column to datetime format
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Filter the data for junction=1 and January
df_january_junction1 = df[(df['DateTime'].dt.month == 1) & (df['Junction'] == 1)]

# Create a pivot table to get the data in the desired format for the heatmap
heatmap_data = pd.pivot_table(df_january_junction1, values='Vehicles', index=df_january_junction1['DateTime'].dt.date, columns=df_january_junction1['DateTime'].dt.hour, fill_value=0)

# Convert the pivot table to a NumPy array for plotting
heatmap_array = heatmap_data.to_numpy()

# Get the date labels for the y-axis
date_labels = [str(date) for date in heatmap_data.index]

# Create the heatmap using Matplotlib
plt.figure(figsize=(12, 8))
plt.imshow(heatmap_array, cmap='YlGnBu', aspect='auto', origin='lower')
plt.colorbar(label='Number of Vehicles')
plt.title('Traffic Heatmap for January (Junction=1)')
plt.xlabel('Hour of the Day')
plt.ylabel('Date')
plt.xticks(np.arange(24), np.arange(24))
plt.yticks(np.arange(len(date_labels)), date_labels, fontsize=8)
plt.show()
```
