```python
import pandas as pd
from bokeh.palettes import Category20c
from bokeh.plotting import figure, curdoc
from bokeh.layouts import column
from bokeh.transform import factor_cmap

# Load the bank data
data = pd.read_csv("bank.csv", delimiter=";")

# Calculate the distribution of the 'marital' variable
marital_counts = data['marital'].value_counts()

# Create a DataFrame with 'marital' and 'marital_percent' columns
total_counts = sum(marital_counts)
marital_data = pd.DataFrame({'marital': marital_counts.index, 'marital_percent': marital_counts.values / total_counts * 100.0})

# Sort the DataFrame by marital_percent in descending order
marital_data = marital_data.sort_values(by='marital_percent', ascending=False)

# Create a bar plot using Bokeh
# The tooltips parameter is used to format and show a tooltip
p = figure(x_range=marital_data['marital'], min_height=350, title="Marital Status Distribution",
           tools="hover", tooltips="@marital: @marital_percent{0.2f}%")

# Add data to the plot
marital_statuses = marital_data['marital'].tolist()
colors = Category20c[len(marital_data)]
source = {'marital': marital_statuses, 'marital_percent': marital_data['marital_percent']}
p.vbar(x='marital', top='marital_percent', width=0.9, source=source, fill_color=factor_cmap('marital', palette=colors, factors=marital_statuses))

# Arrange the plot and the dropdown menu in a layout
layout = column(p)

# Add the layout to the current document (Bokeh server)
curdoc().add_root(layout)

```

You can save this code to a file named `app.py`, and then run it from the command-line like this:

`bokeh server --show app.py`
