## Prompt:


*  Hi bard! i am using the dataset "CO2 Emissions.csv" i want to make a scatter plot comparing emissions vs fuel consumption and engine size, the columns of the data set are "Fuel Consumption Comb (L/100 km)", "CO2 Emissions(g/km)" and "Engine Size(L)".


```python
# Import necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt



# Dataset path.
path = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/co2_emissions.csv'

# Loading the dataset to a pandas dataframe.
data = pd.read_csv(path)

# Scatter plot to compare fuel consumption, CO2 emissions, and engine size.
plt.figure(figsize=(12, 6))
plt.scatter(data['Fuel Consumption Comb (L/100 km)'], data['CO2 Emissions(g/km)'], s=data['Engine Size(L)']*30, c=data['Engine Size(L)'], cmap='viridis', alpha=0.6)

# Add labels.
plt.xlabel('Fuel consumption (L/100 km)')
plt.ylabel('CO2 emissions (g/km)')

# Add title.
plt.title('Fuel Consumption vs. CO2 Emissions (Engine Size)')

# Assign colors according to motor size
plt.colorbar(label='Engine Size (L)')

# Show plot.
plt.grid()
plt.show()
```
