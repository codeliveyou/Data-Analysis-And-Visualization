Visualise the relationship between Fuel Price and Temperature. Use a suitable chart.Please provide the code for the same.Use the provided csv file.




```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/walmart_sales_data.csv")

# Extract columns for Fuel Price and Temperature
fuel_prices = data['Fuel_Price']
temperatures = data['Temperature']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(fuel_prices, temperatures, color='skyblue', alpha=0.5)
plt.title('Relationship between Fuel Price and Temperature')
plt.xlabel('Fuel Price')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```
