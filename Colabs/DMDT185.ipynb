```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/hrdataset_v14.csv")

df['DOB'] = pd.to_datetime(df['DOB'])
df['DateofHire'] = pd.to_datetime(df['DateofHire'])

# Calculate age and tenure for each employee
current_year = datetime.now().year
df['Age'] = current_year - df['DOB'].dt.year
df['Tenure'] = (datetime.now() - df['DateofHire']).dt.days

# Group employees by recruitment source and calculate average age and tenure
recruitment_age_tenure = df.groupby('RecruitmentSource')[['Age', 'Tenure']].mean()

# Create subplots for age and tenure by recruitment source
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
recruitment_age_tenure['Age'].plot(kind='bar', ax=axes[0], color='blue')
axes[0].set_title('Average Age by Recruitment Source')
axes[0].set_xlabel('Recruitment Source')
axes[0].set_ylabel('Average Age')
axes[0].grid(True)
axes[0].set_xticklabels(recruitment_age_tenure.index, rotation=45, ha="right")

recruitment_age_tenure['Tenure'].plot(kind='bar', ax=axes[1], color='green')
axes[1].set_title('Average Tenure by Recruitment Source')
axes[1].set_xlabel('Recruitment Source')
axes[1].set_ylabel('Average Tenure (days)')
axes[1].grid(True)
axes[1].set_xticklabels(recruitment_age_tenure.index, rotation=45, ha="right")

plt.tight_layout()
plt.show()

# Print average age and tenure by recruitment source
print("Average Age and Tenure by Recruitment Source:")
print(recruitment_age_tenure)
```
