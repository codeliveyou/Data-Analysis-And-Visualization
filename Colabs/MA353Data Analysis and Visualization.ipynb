Create a scatter plot visualising the relation between Density and pH.Please provide the code for the same.Use the provided csv file.




```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/winequality_red.csv")

# Select 'density' and 'pH' columns
density = data['density']
ph = data['pH']

# Create a scatter plot with legends
plt.figure(figsize=(8, 6))
plt.scatter(density, ph)
plt.title('Relationship between Density and pH')
plt.xlabel('Density')
plt.ylabel('pH')
plt.grid(True)
plt.show()
```
