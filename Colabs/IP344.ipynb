```python
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool
from bokeh.plotting import output_notebook

import warnings
warnings.filterwarnings('ignore')

# Read data
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/startup_data.csv")

# Dropping columns
df.drop(columns = ['Unnamed: 6',"Unnamed: 0"],inplace=True)

# Create Bokeh figure
p = figure(title="Average Participants in Funding Rounds Over Years", x_axis_label="Year", y_axis_label="Average Participants")

# Calculate average participants in each year
avg_participants_by_year = df.groupby("age_first_funding_year")["avg_participants"].mean()

# Add line plot
p.line(x=avg_participants_by_year.index, y=avg_participants_by_year.values, line_width=2)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Year", "@x"), ("Avg Participants", "@y")]
p.add_tools(hover)

# Output the plot to an HTML file
output_file("line_chart.html")
output_notebook()
# Show the plot
show(p)
```
