```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

df = pd.read_csv("/content/HRDataset_v14.csv")

df['DOB'] = pd.to_datetime(df['DOB'])
df['DateofHire'] = pd.to_datetime(df['DateofHire'])

# Calculate age and tenure for each employee
current_year = datetime.now().year
df['Age'] = current_year - df['DOB'].dt.year
df['Tenure'] = (datetime.now() - df['DateofHire']).dt.days

# Group by department and calculate average age and tenure
department_age_tenure = df.groupby('Department')[['Age', 'Tenure']].mean()

# Create subplots for age and tenure
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
department_age_tenure['Age'].plot(kind='bar', ax=axes[0], color='blue')
axes[0].set_title('Average Age by Department')
axes[0].set_xlabel('Department')
axes[0].set_ylabel('Average Age')
axes[0].grid(True)

department_age_tenure['Tenure'].plot(kind='bar', ax=axes[1], color='green')
axes[1].set_title('Average Tenure by Department')
axes[1].set_xlabel('Department')
axes[1].set_ylabel('Average Tenure (days)')
axes[1].grid(True)

plt.tight_layout()
plt.show()
print("Average Age and Tenure by Department:")
print(department_age_tenure)
```
