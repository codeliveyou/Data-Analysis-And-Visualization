```python
# Import necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt

# Dataset path.
path = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/imdb_top_250_movies.csv'

# Load the dataset.
df = pd.read_csv(path, encoding='latin-1')


# Create the scatter plot.
plt.figure(figsize=(10, 6))
plt.scatter(df['Release_Year'], df['Duration'], color='green', alpha=0.7)

# Add labels.
plt.xlabel('Year of release')
plt.ylabel('Duration (minutes)')

# Add Title.
plt.title('Relationship between the year of release and the length of the movie')

# Show plot
plt.grid()
plt.tight_layout()
plt.show()
```
