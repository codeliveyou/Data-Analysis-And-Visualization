```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool
from bokeh.plotting import output_notebook
from bokeh.transform import factor_cmap
from bokeh.palettes import Category20c
import warnings
warnings.filterwarnings('ignore')

# Read the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/laptops_price_dataset.csv")
df['Screen'] = df['Screen'].astype(str)
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

dfGrouped = df.groupby('Screen')['Final Price'].mean()
df2 = pd.DataFrame({'Screen': dfGrouped.index, 'Price': dfGrouped.values.flatten()})

group = dfGrouped.groupby('Screen')
Category = Category20c[20] + Category20c[20]
# Apply a categorical color mapping to our datset
cmap = factor_cmap('Screen', palette=Category, factors=sorted(df2['Screen'].unique()))

# Create a Bokeh plot, set size and key elements of the plot
p = figure(height=350, x_range=df2['Screen'], title="Screen Vs Final Price",
           toolbar_location="below", tools="pan,wheel_zoom,box_zoom,reset,hover,save")

# Add hover tool
hover = HoverTool()
hover.tooltips = [
    ("Screen", "@Screen"),
    ("Price", "@Price")]
p.add_tools(hover)

# Add vertical bars to our plot
p.vbar(x='Screen', top='Price', width=0.5, source=df2,
       line_color=cmap, fill_color=cmap)

# Set our axes settings and labels
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Screen"
p.xaxis.major_label_orientation = 1.2
p.outline_line_color = None

output_notebook()
show(p)
```
