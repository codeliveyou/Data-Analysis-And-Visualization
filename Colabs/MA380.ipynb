```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Columns to select from the dataframe
columns = ["HP", "Attack", "Defense", "SpAtk", "SpDef", "Speed"]

# Assuming you have the 'root' variable defined elsewhere
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/pokemon_dataset.csv", index_col="Name")[columns]

# Create a list with the names of the Pokmons
charms = {"Charmander": "darkorange", "Charmeleon": "crimson", "Charizard": "orangered"}

# Initialize an empty dictionary to store the Pokmons attributes
attrs = {}
for name in charms:
    arr = df.loc[name, :].values
    attrs[name] = np.append(arr, arr[0])

# Set figure size and polar projection
fig = plt.figure(figsize=(16, 6))
ax = plt.subplot(polar=True)

# Use Matplotlib's thetagrid function
lines, labels = plt.thetagrids(range(0, 360, int(360 / len(columns))), (columns))

# Plot the attributes for every Pokmon
theta = np.linspace(0, 2 * np.pi, len(columns) + 1)
for name, clr in charms.items():
    ax.plot(theta, attrs[name], color=clr)
    ax.fill(theta, attrs[name], alpha=0.1, color = clr)

# Create a custom legend with colored squares
legend_patches = [mpatches.Patch(color=color, label=name) for name, color in charms.items()]
ax.legend(handles=legend_patches, bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.title("Statistics of the Charmander evolution tree", color=charms["Charmander"])

# Display the plot
plt.show()
```
