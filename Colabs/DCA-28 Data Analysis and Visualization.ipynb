```python
# Importing libraries
import pandas as pd

# Read the workout data from the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_science_salaries_2023.csv")

# Filtering by those having the European residency
dfEUResidents = df[df['employee_residence'] == ('ES')]
# Filtering those who are Data Analyst
dfEUResidents = dfEUResidents[dfEUResidents['job_title'] == ('Data Analyst')]
# Filtering those having a salary between 40000 and 90000
dfEUResidents = dfEUResidents[(dfEUResidents['salary'] > 40000) & (dfEUResidents['salary'] < 90000)]

# We get the median salary
median_salary = 0
if(dfEUResidents.size > 0):
  median_salary = dfEUResidents.salary.median().round(0).astype(int)
  print('The median salary for a Data Analyst in Europe is: $', median_salary)
else:
  print("No data exists in the given query")
```
