```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
url = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/automobile.csv'
data = pd.read_csv(url)

# Grouping the data by model_year and calculating the mean of weight
avg_weight = data.groupby('model_year')['weight'].mean()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(avg_weight.index, avg_weight.values, marker='x', color="orangered")
plt.xlabel('Model Year')
plt.ylabel('Average Weight')
plt.title('Average Weight by Model Year')
plt.grid(True)
plt.show()
```
