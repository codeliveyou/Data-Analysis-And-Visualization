```python
import pandas as pd
from datetime import datetime

df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/hrdataset_v14.csv")

# Assuming your dataset is stored in a DataFrame called 'df'
# Convert DateofHire column to datetime format
df['DateofHire'] = pd.to_datetime(df['DateofHire'])

# Calculate tenure for each employee
current_date = datetime.now()
df['Tenure'] = (current_date - df['DateofHire']).dt.days

# Calculate average tenure
average_tenure = df['Tenure'].mean()

# Print average tenure
print("Average Tenure of Employees:", average_tenure, "days")
```
