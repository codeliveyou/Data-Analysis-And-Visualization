```python
# import form mpmath
from mpmath import fp

# create variable
ctx = fp

def mandelbrot(z):
    c = z
    for i in range(50):
        z = z*z + c
        if abs(z) > 8:
            return ctx.exp(1j*(i + 1 - ctx.log(ctx.log(abs(z)))/ctx.log(2)))
    return 0

# plot for x between -3 and 2, y between -2 and 2
ctx.cplot(mandelbrot, [-3, 2], [-2, 2], points=100000, verbose=1)
```
