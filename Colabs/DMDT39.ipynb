```python
# Import necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Dataset path.
path = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/coastal_data_system_waves.csv'

# Load the dataset.
df = pd.read_csv(path)

# Date column to a DateTime object.
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

# Replace sensor -99.9 values by NaN.
df.replace(-99.9, float('nan'), inplace=True)

# Change missing values with the average of each column.
df = df.apply(lambda col: col.fillna(col.mean()), axis=0)

# Select the year 2018.
start_date = '2018-01-01 00:00:00'
end_date = '2018-12-31 11:59:59'
df_year = df[(df['Date/Time'] >= start_date) & (df['Date/Time'] <= end_date)]

# Extract the variables of interest.
df = df_year[['SST', 'Hs', 'Tz']].dropna()

# Create bubble chart.
plt.style.use('dark_background')
plt.figure(figsize=(10, 8))
plt.scatter(df['SST'], df['Hs'], s=df['Tz']*30, alpha=0.5, edgecolors='k')
plt.xlabel('Surface temperature (C°)')
plt.ylabel('Wave height (meters)')
plt.title('Relationship between TSS, Hs and Tz')
plt.grid(True)

# Add a legend for the third dimension (Tz).
plt.legend(labels=['Tz'], title='Tz (seconds)')

# Show plot.
plt.tight_layout()
plt.show()
```
