Make scatter plot of home score vs. away score of the last 10 years, with this dataset: fifa_results.csv

```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the CSV file into a Pandas DataFrame
data_file = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/fifa_results.csv'
df = pd.read_csv(data_file)
# parsing dates
df['date'] = pd.to_datetime(df['date'],errors='coerce',format='%Y-%m-%d')
df['Year'] = df['date'].dt.year

# Filter the DataFrame to include only the last 10 years of data
last_year = df['Year'].max()
df_last_10_years = df[df['Year'] >= last_year - 9]

# Plot the data as a scatter plot
fig, ax = plt.subplots()
ax.scatter(df_last_10_years['home_score'], df_last_10_years['away_score'])

# Set the title and axis labels
ax.set_title('Home score vs. away score (last 10 years)')
ax.set_xlabel('Home score')
ax.set_ylabel('Away score')

plt.show()
```
