```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Assuming your dataset is stored in a DataFrame called 'df'
# Convert DOB to datetime format

df = pd.read_csv("/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/hrdataset_v14.csv")
df['DOB'] = pd.to_datetime(df['DOB'])

# Calculate age for each employee
current_year = datetime.now().year
df['Age'] = current_year - df['DOB'].dt.year

# Create a histogram to visualize the age distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, color='blue', edgecolor='black')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.grid(True)
plt.show()

# Print summary statistics of age distribution
print("Summary Statistics of Age Distribution:")
print("Mean Age:", df['Age'].mean())
print("Median Age:", df['Age'].median())
print("Minimum Age:", df['Age'].min())
print("Maximum Age:", df['Age'].max())
```
