```python
# Primary data structure libraries
import pandas as pd
import numpy as np

# Primary plotting library
import matplotlib.pyplot as plt

# For ggplot-like style of plots
plt.style.use('ggplot')

# Read the csv file
df = pd.read_csv('/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/bank.csv', sep=';')

# Convert age to numeric
df["age"] = pd.to_numeric(df["age"])

# Create subplots using subplot2grid
fig = plt.figure(figsize=(14, 6))

ax1 = plt.subplot2grid((2,2),(0,0))
ax2 = plt.subplot2grid((2,2),(1,0))
ax3 = plt.subplot2grid((2,2),(0,1), rowspan=2)

# Histogram
ax1.hist(df["age"], bins=30, color='skyblue', edgecolor='black')
ax1.set_title('Histogram of Age')
ax1.set_xlabel('Age')
ax1.set_ylabel('Frequency')

# Box Plot
ax2.boxplot(df["age"], vert=False)
ax2.set_title('Box Plot of Age')
ax2.set_xlabel('Age')

# Density Plot
age_mean = df["age"].mean()
age_std = df["age"].std()
# calculate the density curve
age_range = np.arange(df["age"].min(), df["age"].max(), 0.1)
age_density = (1/ (np.sqrt(2 * np.pi * age_std**2))) * np.exp( - (age_range-age_mean)**2 / (2 * age_std**2))
# plot the curve
ax3.plot(age_range, age_density, color='skyblue')
# fill the area
ax3.fill_between(age_range, age_density, color='skyblue', alpha=0.5)
ax3.set_title('Density Plot of Age')
ax3.set_xlabel('Age')

# Improve the layout and add super title
plt.tight_layout()
fig.subplots_adjust(top=0.88)
plt.suptitle('Age Distribution', fontsize=16)

# Output the resultant plot
plt.show()
```
