```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_58mental_health.csv")


# Convert "Mental Health" to numerical
df['Mental Health'] = df['Mental Health'].apply(lambda x: 1 if x == 'Good' else 0)

# Select only ordinal features for box plot
ordinal_features = ['Data Availability', 'Affordable Housing', 'Educational Quality', 'Trust in Police', 'Roads and Sidewalks', 'Local Gatherings']

# Create subplots for box plots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
fig.suptitle("Box Plots of Ordinal Features vs. Mental Health")

for i, ax in enumerate(axes.flatten()):
    feature = ordinal_features[i]
    df[feature] = df[feature].astype('category')
    df.boxplot(column='Mental Health', by=feature, ax=ax)
    ax.set_title(f"Box Plot for {feature}")
    ax.set_xlabel(feature)
    ax.set_ylabel("Mental Health")

plt.tight_layout()
plt.show()
```
