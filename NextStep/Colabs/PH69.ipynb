```python
# importing libraries
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Get the angles from 0 to 2 pie (360 degree) in narray object
X = np.arange(0, np.pi*2, 0.05)

# Using built-in trigonometric function we can directly plot
# the given cosine wave for the given angles
Y1 = np.sin(X)
Y2 = np.cos(X)
Y3 = np.tan(X)


#Subplots
# Horizontal

# Initialise the subplot function using number of rows and columns
figure, (ax1, ax2, ax3) = plt.subplots(1, 3)

# For Sine Function
ax1.plot(X, Y1, 'tab:green')
ax1.set_title("Sine Function")

# For Cosine Function
ax2.plot(X, Y2, 'tab:orange')
ax2.set_title("Cosine Function")

# For Tangent Function
ax3.plot(X, Y3, 'tab:blue')
ax3.set_title("Tangent Function")

# set the spacing between subplots
plt.subplots_adjust(left=0, right=1.1, top=0.9, wspace=0.4, hspace=0.4)

# Combine all the operations and display
plt.show()

# Get the angles from 0 to 2 pie (360 degree) in narray object
X = np.arange(0, np.pi*4, 0.05)
Y1 = np.sin(X)
Y2 = np.cos(X)

# Plotting both the curves simultaneously
plt.plot(X, Y1,"--", color='g', label='sin') # "--" for dashed
plt.plot(X, Y2,":", color='orange', label='cos') # ":"  for dotted

# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Sine and Cosine functions")

# Adding legend, which helps us recognize the curve according to it's color
plt.legend()

# To load the display window
plt.show()

```
