```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_16commodity_prices.csv")

# Calculate price difference between North Sea and Gulf of Oman oil
df['price_difference'] = df['north_sea_oil_price'] - df['gulf_of_oman_oil_price']

# Create a normalized value for dates to use as colormap
norm_dates = (df.index - df.index.min()) / (df.index.max() - df.index.min())

# Create scatter plot with colormap based on normalized dates
plt.figure(figsize=(10, 6))
# Mulptiply 's' parameter by 5 to scale the points for visibility
plt.scatter(df['north_sea_oil_price'], df['gulf_of_oman_oil_price'], s=df['price_difference']*5,
            c=norm_dates, cmap='plasma')

plt.colorbar(label='Date Progression')
plt.xlabel('North Sea Oil Price')
plt.ylabel('Gulf of Oman Oil Price')
plt.title('Price Differences in Oil Markets')
plt.show()
```
