````python
#```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
df = pd.read_csv('data_94food_nutrition_data.csv')

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(heatmap)
plt.title('Correlation Heatmap')

# Add column names as ticks
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

# Show the plot
plt.tight_layout()
plt.show()
#```
````

````python
#```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data into a DataFrame
df = pd.read_csv('data_94food_nutrition_data.csv')

# Create a new column that shows whether a food has more than 10 sodium (mg).
df['high sodium'] = np.where(df['sodium (mg)'] > 10, 1, 0)

# Count high sodium food cases by food category
food_category_high_sodium_count = df.groupby(['food category', 'high sodium']).size().unstack()

# Create a stacked bar plot
plt.figure(figsize=(8, 6))
food_category_high_sodium_count.plot(kind='bar', stacked=True)
plt.title('High Sodium Cases by Food Category')
plt.ylabel('Count')
plt.xlabel('Food Category')
plt.legend(['Low sodium', 'High sodium']) # Add custom legends

# Show the plot
plt.tight_layout()
plt.show()
#```
````
