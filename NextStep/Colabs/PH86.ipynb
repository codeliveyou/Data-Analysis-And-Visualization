```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and axis
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Create data points for the first sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
u, v = np.meshgrid(u, v)
x1 = 3 * np.sin(v) * np.cos(u)
y1 = 3 * np.sin(v) * np.sin(u)
z1 = 3 * np.cos(v)

# Create data points for the second sphere
x2 = np.sin(v) * np.cos(u) + 5
y2 = np.sin(v) * np.sin(u) + 2
z2 = np.cos(v) + 1

# Plot the earth
ax.plot_surface(x1, y1, z1, color='b', alpha=0.5, cmap = plt.cm.viridis)

# Plot the moon
ax.plot_surface(x2, y2, z2, color='r', alpha=0.5, cmap = plt.cm.gist_yarg)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('The earth and the moon')

# Show the plot
plt.show()
```
