```python
# Importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Module used to create 3D plot

# Define the Lorenz system equations
def lorenz(x, y, z, s=10, r=28, b=2.667):
  # Calculate the derivatives using Lorenz equations
  dx = s * (y - x)
  dy = r * x - y - x * z
  dz = x * y - b * z

  return dx, dy, dz

# Set parameters for simulation
dt = 0.01 # Time step
num_steps = 10000 # Number of simulation steps

# Initial conditions
x, y, z = 0.0, 1.0, 1.05 # Initial values
x_vals, y_vals, z_vals = [x], [y], [z] # Lists to store trajectory data

# Perform simulation
for _ in range(num_steps):
  # Calculate derivatives using Lorenz equations
  dx, dy, dz = lorenz(x, y, z)

  # Update values using Euler's method
  x += dx * dt
  y += dy * dt
  z += dz * dt

  # Append updated values to trajectory lists
  x_vals.append(x)
  y_vals.append(y)
  z_vals.append(z)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8)) # Create a new figure
ax = fig.add_subplot(111, projection='3d') # Add a 3D subplot
ax.plot(x_vals, y_vals, z_vals, lw=0.5) # Plot the lorenz attractor trajectory
ax.set_title('Lorenz Attractor') # Set plot title
plt.show() # Display the plot
```
