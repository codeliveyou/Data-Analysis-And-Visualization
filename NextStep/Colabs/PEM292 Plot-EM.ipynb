```python
# Importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate theta values
theta_vals = np.linspace(0, 2 * np.pi, 400) # Generate an array of theta values ranging from 0 to 2*pi

# Calculate r values based on the function
r_vals = 2 + 3 * np.cos(4 * theta_vals) # Calculate r value using the equation r = 2 + 3 * cos(4 * theta)

# Create a polor plot
plt.figure(figsize=(8, 8)) # Create a new figure with specified size
plt.polar(theta_vals, r_vals) # Create a polar plot using theta and r values
plt.title('Polar Plot of r = 2 + 3 * cos(4 * theta)') # Set plot title
plt.show() # Display the plot
```
