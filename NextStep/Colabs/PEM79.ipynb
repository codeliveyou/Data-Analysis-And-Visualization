```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data_3car_data.csv")

# Calculate average fuel efficiency for each year
average_fuel_efficiency = df.groupby('year_made')['fuel_efficiency'].mean()

# Create a time series chart
plt.figure(figsize=(10, 6))
plt.plot(average_fuel_efficiency.index, average_fuel_efficiency.values, marker='o')
plt.title('Average Fuel Efficiency of Cars Over the Years')
plt.xlabel('Year Made')
plt.ylabel('Average Fuel Efficiency (mpg)')
plt.grid(True)
plt.show()
```
