```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/data_4vehicle_data.csv")

# Group data by country of origin and year, then calculate the average fuel economy for each group
average_fuel_economy_by_country = df.groupby(['country_of_origin', 'year_manufactured'])['fuel_economy'].mean().reset_index()

# Create a time series chart for each country of origin
plt.figure(figsize=(12, 8))
sns.set_palette('tab10')  # Set color palette
for country in average_fuel_economy_by_country['country_of_origin'].unique():
    country_data = average_fuel_economy_by_country[average_fuel_economy_by_country['country_of_origin'] == country]
    plt.plot(country_data['year_manufactured'], country_data['fuel_economy'], marker='o', label=country)

plt.title('Fuel Economy Trend by Country of Origin')
plt.xlabel('Year Manufactured')
plt.ylabel('Average Fuel Economy')
plt.legend(title='Country of Origin', loc='best')
plt.grid(True)
plt.show()
```
