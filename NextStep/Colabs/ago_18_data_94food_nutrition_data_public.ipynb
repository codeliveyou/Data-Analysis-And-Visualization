````python
#```python
#Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
#df = pd.read_csv("data_94food_nutrition_data.csv")

# Group by category and get averages
avg_values = df.groupby('food category').mean(numeric_only=True).reset_index()

X = avg_values.columns[1:]
Y = avg_values[avg_values['food category'] == 'Fruit']
Z = avg_values[avg_values['food category'] == 'Vegetable']

X_axis = np.arange(len(X))

# Extract average values for 'Fruit' and 'Vegetable' categories
Y = avg_values[avg_values['food category'] == 'Fruit'].iloc[0, 1:]
Z = avg_values[avg_values['food category'] == 'Vegetable'].iloc[0, 1:]

# Create the bar plot
plt.bar(X_axis - 0.2, Y, 0.4, label='Fruits')
plt.bar(X_axis + 0.2, Z, 0.4, label='Vegetables')
plt.xticks(X_axis, X, rotation=45, ha='right')
plt.xlabel("Variable")
plt.ylabel("Average amount")
plt.title("Average values per food category")
plt.legend()
plt.show()
#```
````
