```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_105country_internet_data.c")

# Create a dual-axis bar plot
fig, ax1 = plt.subplots(figsize=(12, 8))

ax1.bar(df['Country Name'], df['Mobile Subscribers'], color='blue', label='Mobile Subscribers')
ax1.set_xlabel('Country')
ax1.set_ylabel('Mobile Subscribers', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticklabels(df['Country Name'], rotation=45, ha='right')

ax2 = ax1.twinx()
ax2.bar(df['Country Name'], df['Total Internet Users'], color='green', label='Total Internet Users')
ax2.set_ylabel('Total Internet Users', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('Mobile Subscribers vs Total Internet Users by Country')

ax1.legend(loc='upper left', frameon=False)
ax2.legend(loc='upper right', frameon=False)

plt.show()
```
