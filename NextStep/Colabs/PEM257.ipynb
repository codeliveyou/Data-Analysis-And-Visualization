```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_7car_data.csv")

# Create histograms for each origin
plt.figure(figsize=(10, 6))
for origin in df['origin'].unique():
    subset = df[df['origin'] == origin]
    plt.hist(subset['engine_size'], alpha=0.5, label=origin)

# Add labels and title
plt.xlabel('Engine Size')
plt.ylabel('Frequency')
plt.title('Engine Size Distribution per Origin')
plt.legend(title='Origin')

# Show the plot
plt.show()
```
