```python
# Import libraries
import pandas as pd

# Read the data
df = pd.read_csv('/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_35business_data.csv')

# Get holidays
df_holidays = df[df.is_holiday]

# Among holidays, get weekdays
df_holiweekdays = df_holidays.loc[~df.day_of_week.isin(["Saturday", "Sunday"])]

# Calculate the mean of gas price
avg_gas_price = df_holiweekdays.gas_cost_per_gallon.mean()

# Display the result
print(f"The average gallon gas price is {avg_gas_price}")
```
