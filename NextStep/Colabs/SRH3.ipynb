```python
from google.colab import drive

drive.mount("/content/gdrive", force_remount=True)

# Import necessary library
import pandas as pd

# Load data from the "data_140ds_salaries_02.csv" file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_140ds_salaries_02.csv")

# Group the dataframe by the job title and calculate the minimum, maximum salary for each group
grouped_df = df.groupby('job_title')['salary_in_usd'].agg(['min', 'max'])

# Iterate over each row in the grouped dataframe
for index, row in grouped_df.iterrows():
  # Print the job title and salary range for each row
  print(f"Job Title: {index}")
  # Check if the minimum and maximum salary are the same
  if row['min'] == row['max']:
    # Print "Salary: value" if the minimum and maximumm salary are the same
    print(f"Salary:", row['min'])
  else:
    print(f"Salary Range: {row['min']} - {row['max']}")
  print()  # Add an empty line for formatting
```
