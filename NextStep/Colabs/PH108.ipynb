```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

comaps = ["Reds", "Oranges", "Wistia"]

# Parameters for the torus
for j, R in enumerate([13,8,3]):

    r = 1  # Minor radius
    theta = np.linspace(0, 2 * np.pi, 100)  # Angle along the major circle
    phi = np.linspace(0, 2 * np.pi, 100)    # Angle along the minor circle
    theta, phi = np.meshgrid(theta, phi)

    # Parametric equations for torus
    x = (R + r * np.cos(theta)) * np.cos(phi)
    y = (R + r * np.cos(theta)) * np.sin(phi)
    z = np.sin(theta)


    ax.plot_surface(x, y, z, cmap=comaps[j])


# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Torus')

ax.set_zlim(-10, 10)
ax.set_ylim(-15, 15)
ax.set_xlim(-15, 15)

# Show the plot
plt.show()
```
