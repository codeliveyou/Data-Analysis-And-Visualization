```python
import matplotlib.pyplot as plt
import numpy as np

colors = plt.cm.Blues(np.linspace(0.4, 0.99, 10))

sup_x = 100

fig, ax = plt.subplots(figsize=(16, 5))
plt.ylim(-2, 7)
plt.xlim(0, sup_x)
plt.title("The waves of the Ocean")

x = np.linspace(0, sup_x, 15*sup_x)

for i in range(1, 10):
    # Calculate the array
    array = np.sin(x/4 + i*0.5)*(7-i)

    # Replace values less than -1 with -1
    array[array < -1] = -1

    ax.plot(x, array, color=colors[-i])

# Set background gray
plt.gca().set_facecolor('gainsboro')
```
