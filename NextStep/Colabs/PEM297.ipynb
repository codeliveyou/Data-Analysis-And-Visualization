```python
import numpy as np
import matplotlib.pyplot as plt

#the theta values
theta = np.linspace(0,50*np.pi,1000)

#polar equation
polar = np.cos(5*theta) * np.sin(10*theta)

#polar form to cartesian conversion

x = polar * np.cos(theta)
y = polar * np.sin(theta)


fig,axis = plt.subplots(subplot_kw = {'projection':'polar'})


axis.plot(theta,polar,label="Flower pattern")

plt.legend()
plt.show()
```
