Could you provide a visual representation of the distribution of repetitions for each exercise in our gym's workout data? This would help us understand the repetition range for each exercise.


```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
url = '/content/gdrive/MyDrive/DataAnalysisTasks/Datasets/data_76workout_data.csv'
data = pd.read_csv(url)

# Grouping by 'Exercise_Name' and getting unique 'Repetitions'
repetitions_per_exercise = data.groupby('Exercise_Name')['Repetitions'].unique()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(repetitions_per_exercise.index, repetitions_per_exercise.apply(len))
plt.xlabel('Exercise Name')
plt.ylabel('Number of Unique Repetition Ranges')
plt.title('Distribution of Repetitions per Exercise')
plt.xticks(rotation=90)
plt.show()
```
