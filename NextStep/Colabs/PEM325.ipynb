```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_105country_internet_data.c")

# Create a bubble chart
plt.figure(figsize=(12, 8))
plt.scatter(df['Internet Penetration Rate'], df['Fixed Broadband Subscribers'],
            s=df['Mobile Subscribers']*0.1, c=df['Total Internet Users'], cmap='coolwarm', alpha=0.7)

plt.xlabel('Internet Penetration Rate')
plt.ylabel('Fixed Broadband Subscribers')
plt.title('Bubble Chart: Internet Penetration vs Fixed Broadband')
plt.colorbar(label='Total Internet Users')
offset = 0.8
for i, iso_code in enumerate(df['ISO Code']):
    plt.text(df['Internet Penetration Rate'][i] + offset, df['Fixed Broadband Subscribers'][i] + offset, iso_code, fontsize=8)

plt.show()
```
