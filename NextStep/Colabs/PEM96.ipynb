```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_58mental_health.csv")

data = []

# Create pie charts for each feature
features = ['Data Availability', 'Affordable Housing', 'Educational Quality', 'Trust in Police', 'Roads and Sidewalks', 'Local Gatherings']
plt.figure(figsize=(15,10))

# Creating pie charts
for i, feature in enumerate(features):
    # Using sub-plots for each pie chart
    plt.subplot(2, 3, i+1)
    feature_data = df.groupby(feature)['Mental Health'].value_counts().unstack().fillna(0)
    print(feature_data)
    plt.pie(feature_data['Good'], labels=feature_data.index, autopct='%1.1f%%')
    plt.title(f"Mental Health Distribution for {feature}")

plt.tight_layout()
plt.show()
```
