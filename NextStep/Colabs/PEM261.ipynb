```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_28store_weekly_sales.csv")

# Create a scatter plot
plt.figure(figsize=(12, 8))
offset = 0.4
plt.scatter(df['Temperature'], df['Weekly_Sales'], s=df['Unemployment']*10, c=df['Holiday_Flag'], cmap='coolwarm')
plt.colorbar(label='Holiday Flag')
plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.title('Sales vs. Temperature Scatter Plot')
for i, row in df.iterrows():
  plt.text(row['Temperature'] + offset, row['Weekly_Sales'] + offset, row['Store'])
plt.show()
```
