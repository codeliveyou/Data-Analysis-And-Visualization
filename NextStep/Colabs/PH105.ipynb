**Prompt** I would like kernel density estimation plots for cases per 100 population and total cases, along with their respective amounts in bar charts by country in this dataset data_42covid_data.csv. These pairs of graphics must be one on top of the other and in red and lime colors respectively with a black background. You only have to select the data from the year 2023. I need the code in python.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset as pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_42covid_data.csv")

# Filter by year 2023
filt=(df['year']==2023 )

# Create new Dataframe with filtered data
data_2023=df[filt]
data_2023=data_2023[['country_name','cases_per_100,000','total_cases' ]]
data_2023=data_2023.set_index('country_name')

# Set style seaborn
plt.style.use('dark_background')


# Define subplot layout and figure size
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 18))

# Add features plots to subplots
data_2023['cases_per_100,000' ].plot(kind='kde', ax=axes[0, 0], color='red' )
data_2023['cases_per_100,000' ].plot(kind='bar',ax=axes[0,1], color='red')
data_2023['total_cases' ].plot(kind='kde', ax=axes[1, 0], color='lime')
data_2023['total_cases' ].plot(kind='bar',ax=axes[1,1], color='lime')

# Add titles and axis labels by graph
axes[0,0].set_title('Cases per 100,000', fontsize = 12)
axes[0,1].set_xlabel('Country', fontsize = 12)
axes[1,0].set_title('Total cases', fontsize = 12)
axes[1,1].set_xlabel('Country', fontsize = 12)

# Shows the grid with the sub graphs
plt.show()
```
