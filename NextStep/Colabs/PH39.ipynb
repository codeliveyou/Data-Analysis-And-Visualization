###PROMPT
Based on the file data_69flower_data.csv, making the assumption that the area is calculated as the multiplication of the lenght and the width of the sepal or the petal, I want you to calculate the respective area for the sepal and petals and plot them in a 3D scatterplot including also the relation between the areas, using different colors and markers for each flower type.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the dataset from the csv file
df = pd.read_csv("/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_69flower_data.csv")

# Calculate the areas and their relations
df['sepal_area']=df['sepal_length_in_cm']*df['sepal_width_in_cm']
df['petal_area']=df['petal_length_in_cm']*df['petal_width_in_cm']
df['area_relations']=df['sepal_area']/df['petal_area']

# Group the data by the flower types
grouped_df=df.groupby(['flower_type'])

# Create colors and markers lists for using for each flower type
colors=['b','g','r']
markers=['o','v','P']

# Create the plot
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')

# Iterate over the flower types to generate each plot
for (key, group),index in zip(grouped_df,range(len(grouped_df))):
    ax.scatter(group['petal_area'],group['sepal_area'],group['area_relations'], color=colors[index], marker=markers[index], label=key)

# Set the plot title
plt.title("Iris Flowers Sepal and Petals area")

# Set the axis label
ax.set_xlabel('Petal Area')
ax.set_ylabel('Sepal Area', loc='top')
ax.set_zlabel('Area Relations')

# Set tight layout to make sure the axis labels are shown correctly
plt.tight_layout()

# Add the legends
plt.legend(title="Flower Types")

# Show the plot
plt.show()
```
