```python
from google.colab import drive

drive.mount("/content/gdrive", force_remount=True)

# Import necessary library
import pandas as pd

# Load data from the "data_101commodity_prices.csv" file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_101commodity_prices.csv")

# Create a new dataframe with only the Sugar Price columns
sugar_df = df[['European Sugar Price', 'US Sugar Price', 'World Sugar Price']].copy()

# Calculate the price difference between the US Sugar Price and the World Sugar Price
sugar_df.loc[:, 'US vs World Sugar Price'] = sugar_df['US Sugar Price'] - sugar_df['World Sugar Price']

# Calculate the price difference between the European Sugar Price and the World Sugar Price
sugar_df.loc[:, 'European vs World Sugar Price'] = sugar_df['European Sugar Price'] - sugar_df['World Sugar Price']

# Create a new dataframe with only the Sugar Price difference columns
sugar_diff_df = sugar_df[['US vs World Sugar Price', 'European vs World Sugar Price']].copy()

# Print the new dataframe
print(sugar_diff_df)
```
