````python
```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("data_132heart_01.csv")
# Calculate the correlation matrix for all numerical columns
correlation_matrix = df.corr()

# Draw the heatmap by setting the figure size, ticks and colorbar
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', aspect='auto')
plt.title('Correlation Heatmap')
plt.xticks(np.arange(correlation_matrix.shape[0]), correlation_matrix.columns, rotation=90)
plt.yticks(np.arange(correlation_matrix.shape[0]), correlation_matrix.columns)
plt.colorbar(label='Correlation Coefficient')
plt.tight_layout()
plt.show()
```python
````
