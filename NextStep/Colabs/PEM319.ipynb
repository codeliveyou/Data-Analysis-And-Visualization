```python
import matplotlib.pyplot as plt
import numpy as np

signal_duration = 0.5 # In seconds
rate = 500                # In Hz

#time between two samples (seconds)
dt = 1./rate

#frequencey between points (Hz)
df = 1 / signal_duration

time_vector = np.linspace(0, signal_duration, int(signal_duration * rate))
num_samples = len(time_vector)

#generate signal
signal = np.sin(2 * np.pi * 50 * time_vector) + np.sin(2 *np.pi * 120* time_vector + np.pi/4)

#fourier transform
ft = np.fft.fft(signal)

#meaningful frequencies in fourier transform
freq = df*np.arange(0,(num_samples-1)/2.,dtype ='d')
num_frequencies = len(freq)


fig,axes = plt.subplots(nrows = 2, ncols = 1)

#plot input data signal against time
plt.subplot(2,1,1)
plt.plot(time_vector,signal,label='input data', color='blue')
plt.xlabel('time [s]')
plt.ylabel('Amplitude')
axes[0].set_title('Input signal')
plt.grid(True)

#plot frequency spectrum
plt.subplot(2,1,2)
plt.plot(freq,abs(ft[0:num_frequencies]),label='abs(fourier transform)',color='orange')
plt.xlabel('frequency [Hz]')
plt.ylabel('abs(DFT(Amplitude))')
axes[1].set_title('Fourier spectrum')

plt.subplots_adjust(hspace=0.7)

plt.grid(True)
plt.show()
```
