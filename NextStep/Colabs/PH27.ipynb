```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_111football_match_data.csv")

# Filter the DataFrame to include only rows with both Offsides and Red Cards not equal to 0
df_offsides_redcards = df[(df["Offsides"] != 0) & (df["Red Cards"] != 0)]

# Calculate the ratio of Accurate Passes to Total Passes and add it as a new column
df["ratio"] = df["Passes Accurate"] / df["Total Passes"]

# Get the maximum of the whole 2 columns to plot
rango = df[["ratio", "Passes Accuracy"]].max().max()

# Create a scatter plot of "ratio" on the x-axis and "Passes Accuracy" on the y-axis
# Points above the 45-degree slope will be red, and points below will be green
plt.scatter(df.ratio, df["Passes Accuracy"], c=['red' if y > x else 'green' for x, y in zip(df.ratio, df["Passes Accuracy"])])

# Add a 45-degree slope line to the plot
plt.plot([0, rango], [0, rango], color='blue', linestyle='--')

# Set plot labels and title
plt.xlabel("Passes Accurate / Total Passes Ratio")
plt.ylabel("Passes Accuracy")
plt.title("Passes Accuracy vs. Passes Accurate / Total Passes Ratio")

# Show the plot
plt.show()
```
