```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/data_4vehicle_data.csv")

# Create a histogram for engine displacements for each country of origin
plt.figure(figsize=(12, 8))
sns.set_palette('tab10')  # Set color palette
for country in df['country_of_origin'].unique():
    country_data = df[df['country_of_origin'] == country]
    sns.histplot(data=country_data, x='engine_displacement', bins=15, kde=True, label=country)

plt.title('Distribution of Engine Displacements by Country of Origin')
plt.xlabel('Engine Displacement')
plt.ylabel('Frequency')
plt.legend(title='Country of Origin', loc='upper right')
plt.grid(True)
plt.show()
```
