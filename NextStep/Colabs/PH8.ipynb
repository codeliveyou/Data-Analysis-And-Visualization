I want you to plot a flower with 18 violet petals with center at (3,3), radius 9, violet petals with center at (3,3) radius 2 and 4 yellow petals with center at (3,3) radius 1, fill it with the given color and include a grid.

```python
# Imports
import matplotlib.pyplot as plt
import numpy as np

# Flower configuration parameters
center_x, center_y = 3, 3
petals_1, r_1 = 18, 3
petals_2, r_2 = 9, 2
petals_3, r_3 = 4, 1


def calculate_petals_coordinates(petals, r, center_x, center_y):

    # 1000 values equally spaced at 360 degrees
    t = np.linspace(0, 2 * np.pi, 1000)

    # Calculate r representing the distance from the center to the curve as a function of the angle t and the number of petals.
    half_petals = petals / 2
    r = np.sqrt(np.abs(np.cos(half_petals * t))) * r

    # The coordinates are calculated
    x = center_x + r * np.cos(t)
    y = center_y + r * np.sin(t)
    return x, y


# The plot is created
x, y = calculate_petals_coordinates(petals_1, r_1, center_x, center_y)
plt.fill(x, y, color="purple", alpha=0.5)

x, y = calculate_petals_coordinates(petals_2, r_2, center_x, center_y)
plt.fill(x, y, color="purple", alpha=0.5)

x, y = calculate_petals_coordinates(petals_3, r_3, center_x, center_y)
plt.fill(x, y, color="yellow", alpha=0.5)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(
    f"Flower centered at ({center_x},{center_y}) with {petals_1} petals, radius {r_1}, {petals_2} petals, radius {r_2}, {petals_3} petals, radius {r_3} "
)
plt.grid(True)
plt.axis("equal")
plt.show()
```
