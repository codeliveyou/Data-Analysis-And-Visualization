```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def draw_rectangles(rectangles):
    fig, ax = plt.subplots(figsize=(25,18))

    # coordinates of the bottom left vertix of the first square
    x, y = 0, 0

    colorin = 0
    colores = ["r", "brown", "b", "purple", "green", "yellow"]

    for i, width in enumerate(rectangles[:-4]):
        print(width)
        rect = Rectangle((x, y), width, width, fill=True, color=colores[colorin % 6])
        ax.add_patch(rect)

        # Calculate the center of the rectangle
        center_x = x + width / 2
        center_y = y + width / 2

        # Write the rectangle side value at the center
        ax.text(center_x, center_y, str(width), ha='center', va='center', color='white', fontsize=20)

        # in this `if` statement we handle the counterclockwise rotations
        if colorin % 4 == 0:
            x += width
        if colorin % 4 == 1:
            y += width
            x += width - rectangles[i + 1]
        if colorin % 4 == 2:
            y += rectangles[i + 2]
            x += -rectangles[i + 1]
        if colorin % 4 == 3:
            y -= rectangles[i + 1]

        colorin += 1

    plt.xlim(-1, 1.8* sum(rectangles[:1]) + 1)
    plt.ylim(-1, max(rectangles) + 1)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')
    plt.show()

# reversed Fibonacci numbers
rectangle_dimensions = [89, 55, 34, 21, 13, 8, 5, 3, 2, 1, 1, 0, 0]

# call the function
draw_rectangles(rectangle_dimensions)
```
