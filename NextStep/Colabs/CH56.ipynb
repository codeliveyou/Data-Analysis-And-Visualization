```python
# Import the libraries
import pandas as pd
import numpy as np

# Read the data
df = pd.read_csv('/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_115birth_statistics.csv')

# Get only the States of California, Florida & Texas
df_fl_ca_tx = df[df.state_code.isin(['CA', 'FL', 'TX'])].reset_index()

# Get females
df_fl_ca_tx_females = df_fl_ca_tx.loc[df.sex=='Female'].reset_index()

# Get the distribution of the Highest Education Level
response_dict = df_fl_ca_tx_females.highest_education_level.value_counts().to_dict()

all = sum(response_dict.values())

ratio = np.round(100*(all-response_dict['Bachelor'])/all, 2)

print(f"Among female births from California, Florida or Texas, {ratio}% that have an education level greater than Bachelor. The distribution is\n\n {response_dict}")
```
