```python
# Import the libraries
import pandas as pd
import numpy as np

# Read the data
df = pd.read_csv('/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_115birth_statistics.csv')

# Get females
df_females = df.loc[df.sex=='Female'].reset_index()

# Split the string column
df_city = df_females.birth_location.str.split(expand=True)

# Determine the maximum number of generated columns
max_columns = df_city.shape[1]

# Create new column names for the generated columns
new_column_names = [f'split_{i+1}' for i in range(max_columns)]

# Add the generated columns to the original DataFrame
df_females[new_column_names] = df_city

# Get rows where city name has only 1 word
df_females = df_females[df_females.split_2.isna()]

# print result
print(df_females.number_of_births.median())
```
