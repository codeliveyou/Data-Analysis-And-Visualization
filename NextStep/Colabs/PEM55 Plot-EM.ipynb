```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Import for 3D scatter plot

# load the dataset
data = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_0car_data.csv")

# Create a 3D scatter plot of weight, acceleration, and horsepower
fig = plt.figure(figsize=(10, 8)) # Create a new figure
ax = fig.add_subplot(111, projection= '3d') # Add a 3D subplot

# Scatter plot with weight, acceleration, and horsepower
ax.scatter(data['weight'], data['acceleration'], data['horsepower'], c='purple', marker= 'o', alpha=0.7) # Create a scatter plot
ax.set_xlabel('Weight') # Label for the x-axis
ax.set_ylabel('Acceleration') # Label for the y-axis
ax.set_zlabel('Horsepower') # Label for the z-axis
ax.set_title('3D Scatter Plot of Weight, Acceleration and Horsepower') # Title of the plot
plt.show() # Show the scatter plot
```
