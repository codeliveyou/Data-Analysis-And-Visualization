```python
from google.colab import drive

drive.mount("/content/gdrive", force_remount=True)

# Import necessary library
import pandas as pd

# Load data from the "world_cups_data.csv" file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_139exam_score_02.csv")

# Calculate the average scores of males and females in reading and writing
avg_scores = df.groupby("Gender")[["ReadingScore", "WritingScore"]].mean()

# Print the average scores
print("Average Scores:")
print(avg_scores)

# Create a new table for people whose difference between their average score and their own score is less than 10
new_table = df.loc[abs(df["ReadingScore"] - avg_scores.loc[df["Gender"], "ReadingScore"]) < 10]
new_table = new_table.loc[abs(new_table["WritingScore"] - avg_scores.loc[new_table["Gender"], "WritingScore"]) < 10]
new_table = new_table[["LunchType", "PracticeSport", "TransportMeans"]]

# Print the new table
print("\nNew Table:")
print(new_table)
```
