```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Convert the string data to a DataFrame
data_string = """car_model,fuel_economy_mpg,engine_displacement_liters,engine_displacement_cubic_inches,horsepower,weight_lbs,acceleration_0_to_60_seconds,manufacturing_year,country_of_origin
Tesla Model 3,39,0.28,140,286,3500,5.1,2017,USA
Toyota Camry,30,2.5,150,181,3300,7.9,2022,Japan
Honda Civic,31,1.5,99,158,2700,8.2,2023,Japan
Ford F-150,18,3.5,281,375,6000,6.2,2022,USA
Chevy Silverado,19,5.3,355,420,6200,6.0,2023,USA
Toyota Corolla,33,1.8,121,139,2500,9.0,2022,Japan
Honda Accord,32,2.0,159,192,3200,8.4,2023,Japan
Nissan Altima,34,2.5,182,188,3000,8.6,2022,Japan
Hyundai Sonata,35,2.4,178,178,3100,8.8,2023,South Korea
Kia Optima,36,2.4,178,178,3100,8.8,2022,South Korea"""

# Convert the string data to a DataFrame
from io import StringIO
data_io = StringIO(data_string)
df = pd.read_csv(data_io)

# 1. Plot a bar plot showing the average horsepower for cars from different countries of origin.
grouped = df.groupby('country_of_origin')['horsepower'].mean().sort_values()
plt.figure(figsize=(8, 6))
plt.bar(grouped.index, grouped.values, color=['blue', 'green', 'red'])
plt.xlabel('Country of Origin')
plt.ylabel('Average Horsepower')
plt.title('Average Horsepower by Country of Origin')
plt.xticks(rotation=45)
plt.show()

# 2. Plot a scatter plot to visualize the relationship between the weight of the cars and their acceleration from 0 to 60 seconds.
plt.figure(figsize=(8, 6))
plt.scatter(df['weight_lbs'], df['acceleration_0_to_60_seconds'], color='blue', alpha=0.6)
plt.xlabel('Weight (lbs)')
plt.ylabel('Acceleration 0 to 60 seconds')
plt.title('Weight vs. Acceleration from 0 to 60 seconds')
plt.show()
```
