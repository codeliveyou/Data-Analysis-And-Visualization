````python
#```python
# Import libraries
import pandas as pd
import numpy as np

# Load the data into a DataFrame
df = pd.read_csv('data_94food_nutrition_data.csv')

# Filter out rows where calories (kcal) is between 20 and 50 and food category is fruit
filtered_data = df[(df['calories (kcal)'] >= 20) & (df['calories (kcal)'] <= 50) & (df['food category'] == 'Fruit')]

# Create a column indicating whether a food is above 10 sodium (mg)
filtered_data['high sodium'] = np.where(filtered_data['sodium (mg)'] > 10, 1, 0)

# Create a pivot table to calculate the median carbohydrates (g) for high sodium by fruits
pivot_table = pd.pivot_table(filtered_data, values='carbohydrates (g)', index=['food category', 'high sodium'], aggfunc=np.median)

# Print the results
print("Average carbohydrate (g) for high sodium (over 10 mg) \n fruits with calories (kcal) between 20 and 50:\n",
      pivot_table)
#```
````
