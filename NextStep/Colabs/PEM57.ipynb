```python
import matplotlib.pyplot as plt

#data sets
subjects = ['Computer','English','Chemistry','Physics','Biology','Maths', 'History']

student_1_marks = [70,60,90,80,70,65,80]

student_2_marks = [72,80,80,70,70,85,90]

#plot the marks
plt.plot(subjects,student_1_marks,label='Student 1 marks', marker = 'o', markerfacecolor='purple')

plt.plot(subjects,student_2_marks,label='Student 2 marks',marker = 'o', markerfacecolor='green')

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.title('Marks comparision')

plt.legend()
plt.show()
```
