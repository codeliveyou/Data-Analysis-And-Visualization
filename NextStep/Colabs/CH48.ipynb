```python
# Import libraries
import pandas as pd
import numpy as np

# Read the data
df = pd.read_csv('/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_35business_data.csv')

# Get days with at least 500 total sales.
df_500 = df[df.total_sales >= 500]

# Group by holidays or not holidays, get mean of consumer_price_index
df_holi_vs_non = df_500.groupby("is_holiday")["consumer_price_index"].mean()

# Transform to dictionary
df_holi_vs_non_dict = df_holi_vs_non.to_dict()

# print result
if df_holi_vs_non_dict[True] >= df_holi_vs_non_dict[False]:
    ratio_diff = 100*(df_holi_vs_non_dict[True]/df_holi_vs_non_dict[False] - 1)

    print(f"Among days with at least 500 total sales, the average consumer_price_index during holidays is {np.round(ratio_diff, 2)}% higher.")

else:
    ratio_diff = 100*(1 - df_holi_vs_non_dict[True]/df_holi_vs_non_dict[False])

    print(f"Among days with at least 500 total sales, the average consumer_price_index  during holidays is {np.round(ratio_diff, 2)}% lower.")
```
