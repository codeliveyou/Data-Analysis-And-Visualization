```python
# Import Necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("data_76workout_data.csv")

# Initialise dictionary to count each target area
targetAreaDictionary={}
for index, row in df.iterrows():
  targetAreaDictionary[row['Target_Area']]=0

# Calculate total workout durations of each target area
for index, row in df.iterrows():
  parts=row['Workout_Duration'].split();
  targetAreaDictionary[row['Target_Area']]+=(int)(parts[0])

# Iterate dictionary to create lists
targetAreas=list(targetAreaDictionary.keys())
workoutDurations=list(targetAreaDictionary.values())

# Create graph
plt.bar(targetAreas, workoutDurations)
plt.xlabel('Target Area')
plt.ylabel('Workout Duration')
plt.title('Bar Plot of Category Values')

# Show graph
plt.show()
```
