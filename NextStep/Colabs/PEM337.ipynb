```python
# Necessary imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dataset as a string
data = """age,sex,chest pain type,resting blood pressure,serum cholestoral,fasting blood sugar,resting electrocardiographic results,max heart rate,exercise induced angina,heart disease
72,0,3,130,328,0,2,110,1,1
68,0,3,125,555,1,0,169,1,1
57,1,2,124,261,0,0,141,0,2
66,0,4,128,263,1,0,105,1,2
74,1,3,115,279,0,2,122,1,1
69,1,4,130,179,0,2,140,0,2
56,0,3,130,256,1,0,142,1,2
57,1,4,115,247,0,0,143,0,1
60,0,3,140,269,0,2,170,1,1
69,1,2,143,399,0,0,158,0,2
59,1,4,135,266,0,0,161,0,2"""

# Converting the string data to a pandas DataFrame
from io import StringIO
df = pd.read_csv(StringIO(data))

# Plotting
plt.figure(figsize=(10,6))

for gender, marker in zip([0, 1], ['o', '^']):
    subset = df[df['sex'] == gender]
    for disease, color in zip([1, 2], ['red', 'green']):
        plt.scatter(subset[subset['heart disease'] == disease]['age'],
                    subset[subset['heart disease'] == disease]['max heart rate'],
                    color=color, marker=marker, label=f"{'Male' if gender == 1 else 'Female'} - {'Disease' if disease == 1 else 'No Disease'}")

plt.title("Age vs. Max Heart Rate by Gender and Heart Disease Status")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
