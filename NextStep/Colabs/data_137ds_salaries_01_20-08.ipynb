````python
#```python
# Import libraries
import pandas as pd
import numpy as np

# Load the data into a DataFrame
df = pd.read_csv('data_137ds_salaries_01.csv')

# Calculate the threshold for low, medium, and high salaries
salary_threshold = df['salary_in_usd'].sum() / 3

# Define a function to categorize salaries
def categorize_salary(salary):
  if salary < salary_threshold:
    return 'Low'
  elif salary >= salary_threshold and salary < 2 * salary_threshold:
    return 'Medium'
  else:
    return 'High'

# Apply the categorize_salary function to create a new column called 'salary_category'
df['salary_category'] = df['salary_in_usd'].apply(categorize_salary)

# Create a pivot table to calculate the average salary in USD by job title
pivot_table = pd.pivot_table(df, values='salary_in_usd',
                             index=['job_title', 'salary_category'],
                             aggfunc=np.mean)

# Print the results
print("Average salary in USD by job title:\n",
      pivot_table)
#```
````
