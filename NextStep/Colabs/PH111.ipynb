```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw first circunference
radius1 = 4
theta = np.linspace(0, 2 * np.pi, 100)

# Parametric equations for the circumference
x_circle1 = radius1 * np.cos(theta)
y_circle1 = radius1 * np.sin(theta)
z_circle1 = np.zeros_like(theta)

ax.plot(x_circle1, y_circle1, z_circle1, color='red', label=f'Radius {radius1}')


# Draw the torus helix

# Minor radius
r = 1

# Angle along the major circle
theta = np.linspace(0, 2 * np.pi, 100)

# Angle along the minor circle
phi = np.linspace(0, 2 * np.pi, 100)

# Parametric equations for torus
minor_radius = 0.8
curls = 10

x = (radius1 + minor_radius*np.cos(curls*theta))*np.cos(phi)
y = (radius1 + minor_radius*np.cos(curls*theta))*np.sin(phi)
z =  minor_radius*np.sin(curls*theta)

# Plot the second curve
ax.plot(x, y, z, color='blue', label=f'Radius {minor_radius}')


# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Torus helix around circumference')

ax.set_zlim(-2, 2)
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)

# Show the plot
plt.legend()
plt.show()
```
