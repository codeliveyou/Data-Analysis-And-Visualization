```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def draw_rectangles(rectangles):
    """
    Here we draw a sequence of purple and yellow rectangles whose sides are powers of 2, and they reduce by half when we get closer to the origin (0,0).
    """

    fig, ax = plt.subplots(figsize=(25,18))

    for i, width in enumerate(rectangles):

        # we draw the purple and yellow rectangles
        if i%2==0:
          rect = Rectangle((width, width), width, width, fill=True, color="purple")
          ax.add_patch(rect)

          rect = Rectangle((0,0), width, width, fill=True, color="purple")
          ax.add_patch(rect)

          rect = Rectangle((0, width), width, width, fill=True, color="yellow")
          ax.add_patch(rect)

          rect = Rectangle((width,0), width, width, fill=True, color="yellow")
          ax.add_patch(rect)

        # for odd number of iterations, we invert the positions of purple and yellow colors
        else:
          rect = Rectangle((0, width), width, width, fill=True, color="purple")
          ax.add_patch(rect)

          rect = Rectangle((0,width), width, width, fill=True, color="purple")
          ax.add_patch(rect)

          rect = Rectangle((0, 0), width, width, fill=True, color="yellow")
          ax.add_patch(rect)

          rect = Rectangle((width, width), width, width, fill=True, color="yellow")
          ax.add_patch(rect)

        border = Rectangle((0, 0), 2 * width, 2 * width, fill=False, edgecolor="white")
        ax.add_patch(border)

    plt.xlim(-1, 2*rectangles[0] + 1)
    plt.ylim(-1, 2*rectangles[0] + 1)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')
    plt.tight_layout()
    plt.show()

rectangle_dimensions = [2**l for l in range(4,-4,-1)]

draw_rectangles(rectangle_dimensions)
```
