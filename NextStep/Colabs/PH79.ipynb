```python
import matplotlib.pyplot as plt
import numpy as np

# shades of red
colors = plt.cm.Reds(np.linspace(0.4, 0.99, 10))

def red_sinplot():
    """
    Here we we plot red waves that get darker from bottom to front.
    """

    # create fig, ax objects
    fig, ax = plt.subplots(figsize=(10,6))

    # x-axis vector
    x = np.linspace(0, 21, 10*21)

    # plot the waves
    for i in range(1, 10):
        ax.plot(x, np.sin(x + i*0.5)*(7-i), color=colors[-i])

    return ax

ax = red_sinplot()
```
