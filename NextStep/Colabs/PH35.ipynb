Based on the file data_126exam_score_03.csv, I want you to plot in a 3d scatter plot the exams scores for female students, plotting the data with different colors and markers according to the weekly hours of study. Setting the size of the plot as 12x12. Make sure that all the axis labels are shown correctly, adding an additional padding if necessary

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the dataset from the csv file
df = pd.read_csv("/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_126exam_score_03.csv")

# Group the data by the weekly study hours
grouped_df=df.groupby(['WklyStudyHours'])

# Create colors and markers lists for using for each value of the study hours
colors=['b','g','r']
markers=['x','v','P']

# Create the plot
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(projection='3d')
# Iterate over the study hours to generate each plot
for (key, group),index in zip(grouped_df,range(len(grouped_df))):
    ax.scatter(group['MathScore'],group['ReadingScore'],group['WritingScore'], color=colors[index], marker=markers[index], label=key)

# Set the plot title
plt.title("Female Students Scores Plot")

# Set the axis label
ax.set_xlabel('Math Score')
ax.set_ylabel('Reading Score', loc='top')
ax.set_zlabel('Writing Score')

# Set tight layout to make sure the axis labels are shown correctly
plt.tight_layout()

# Add the legends
plt.legend(title="Weekly Study Hours")

plt.show()

```
