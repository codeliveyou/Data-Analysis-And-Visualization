```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_132heart_01.csv")

# Separate data for heart disease status taking 1 as has heart disease and 2 as that has not.
no_heart_disease = df[df['heart disease'] == 1]
has_heart_disease = df[df['heart disease'] == 2]

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Create histogram for resting blood pressure
ax1.hist(no_heart_disease['resting blood pressure'], bins=20, color='green', alpha=0.7, label='No Heart Disease')
ax1.hist(has_heart_disease['resting blood pressure'], bins=20, color='red', alpha=0.7, label='Has Heart Disease')
ax1.set_title('Resting Blood Pressure Distribution')
ax1.set_xlabel('Resting Blood Pressure')
ax1.set_ylabel('Frequency')
ax1.legend()

# Create scatter plot for serum cholestoral vs. resting blood pressure
ax2.plot(df['resting blood pressure'], df['serum cholestoral'], 'o', label='All')
ax2.plot(no_heart_disease['resting blood pressure'], no_heart_disease['serum cholestoral'], 'go', label='No Heart Disease')
ax2.plot(has_heart_disease['resting blood pressure'], has_heart_disease['serum cholestoral'], 'ro', label='Has Heart Disease')
ax2.set_title('Serum Cholestoral vs. Resting Blood Pressure')
ax2.set_xlabel('Resting Blood Pressure')
ax2.set_ylabel('Serum Cholestoral')
ax2.legend()

plt.tight_layout()
plt.show()
```
