```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data_3car_data.csv")

# Calculate correlation between engine size and acceleration
correlation = df['engine_size'].corr(df['acceleration'])

# Plotting the scatter chart
plt.figure(figsize=(10, 6))
plt.scatter(df['engine_size'], df['acceleration'], color='green', alpha=0.7)
plt.xlabel('Engine Size')
plt.ylabel('Acceleration')
plt.title(f'Scatter Plot: Engine Size vs Acceleration\nCorrelation: {correlation:.2f}')
plt.grid(True)
plt.show()
```
