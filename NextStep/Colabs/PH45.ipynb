###PROMPT
Based on the file data_6car_data.csv, calculate the relation of the acceleration and the weight of the car and store it as the relative power. Then plot in a barplot the horsepower of each car using different colors for each country of origin, and over the bar plot I want a lineplot of the relative power that has its own y axis with a different scale

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the dataset from the csv file
df = pd.read_csv("/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_6car_data.csv")

# Calculate the relative power
df['relative_power']=df['horsepower']/df['weight']

# Group the data by the origin
grouped_df=df.groupby(['origin'])

# Create colors and markers lists for using for each origin
colors=['b','g','r','y']
markers=['o','v','P','x']

# Create the plot
fig, ax1 = plt.subplots(figsize=(10,10))
labels=[]

# Iterate over the car origin to generate each plot
for (key, group),index in zip(grouped_df,range(len(grouped_df))):
    ax1.bar(x=group['make_and_model'],height=group['horsepower'],color=colors[index], label=key)
    labels.append(key)

# Set the plot title
plt.title("Car Models Horsepower vs Relative Power")

# Set the axis label
ax1.set_xlabel('Make and Model')
ax1.set_ylabel('Horse Power')
ax1.tick_params(axis='x', rotation = 45)

# instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()

# Sort the df to match with the grouped_df indexes
sorted_df=df.sort_values('origin')

# Plot the relative power lines
color = 'black'
ax2.set_ylabel('Relative Power', color=color)
ax2.plot(sorted_df['make_and_model'], sorted_df['relative_power'], color='black')
ax2.tick_params(axis='y', labelcolor=color)


# Add the legends
legend_labels = [f"{label}" for label, color in zip(labels, colors)]
ax1.legend(labels=legend_labels, title="Car Origin", loc='upper left')

# Set tight layout to make sure the axis labels are shown correctly
plt.tight_layout()

plt.legend(labels=['relative power'])

# Show the plot
plt.show()
```
