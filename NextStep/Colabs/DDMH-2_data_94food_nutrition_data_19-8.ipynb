````python
#```python
# Import the necessary libraries
import pandas as pd

# Load the data into a DataFrame
df = pd.read_csv('data_94food_nutrition_data.csv')

# Filter out rows where calories (kcal) are above 30
filtered_data = df[(df['calories (kcal)'] >= 30)]

# Group by food name and get the maximum value of carbohydrate
max_values = filtered_data.groupby('food name')['carbohydrates (g)'].max().reset_index()

# Sort the results by maximum value of carbohydrate in descending order
sorted_max_values = max_values.sort_values(by='carbohydrates (g)', ascending=False)

# Show the results as a Data Frame table
pd.DataFrame(sorted_max_values)
#```
````
