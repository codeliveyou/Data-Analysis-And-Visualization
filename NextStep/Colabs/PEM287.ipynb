```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_28store_weekly_sales.csv")

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Create a line chart
plt.figure(figsize=(10, 6))
stores = df['Store'].unique()
for store in stores:
    data = df[df['Store'] == store]
    plt.scatter(data['Date'], data['Weekly_Sales'], label=store)
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.title('Sales Over Time')
plt.legend()
plt.show()
```
