```python

# Importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

    # Read the workout data from the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_127car_info_02.csv")



# Sort data by case number
sorted_data = df.sort_values(by='name')

# Plot a time series chart
plt.plot(sorted_data['displacement'], sorted_data['name'], marker='o')
plt.xlabel('Displacement')
plt.ylabel('Vehicle_name')
plt.title('Displacement Time Series by Vehicle Name')
plt.grid()
plt.show()
```
