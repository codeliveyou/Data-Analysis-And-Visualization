````python
#```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
df = pd.read_csv('data_132heart_01.csv')

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(heatmap)
plt.title('Correlation Heatmap')

# Add column names as ticks
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

# Show the plot
plt.tight_layout()
plt.show()
#```
````

# Task PH58 (Plot-H)

## Prompt

Using the data in the data_132heart_01.csv file, create a stacked bar chart that counts heart disease cases by sex with custom labels.

## data

[image of the query]

````python
#```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
df = pd.read_csv('data_132heart_01.csv')

# Count heart disease cases by sex
sex_heart_disease_count = df.groupby(['sex', 'heart disease']).size().unstack()

# Create a stacked bar plot
plt.figure(figsize=(8, 6))
sex_heart_disease_count.plot(kind='bar', stacked=True)
plt.title('Heart Disease Cases by Sex')
plt.ylabel('Count')
plt.xlabel('Sex')
plt.xticks([0, 1], ['Male', 'Female'], rotation=0)  # Add custom labels for sex
plt.legend(['No Disease', 'Disease'])

# Show the plot
plt.tight_layout()
plt.show()
#```
````
