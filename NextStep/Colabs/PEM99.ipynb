```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_58mental_health.csv")


# Create histograms
features = ['Data Availability', 'Affordable Housing', 'Educational Quality', 'Trust in Police', 'Roads and Sidewalks', 'Local Gatherings']
df['Mental Health'] = df['Mental Health'].apply(lambda x: 1 if x == 'Good' else 0)
plt.figure(figsize=(15,10))

for i, feature in enumerate(features):
    plt.subplot(2, 3, i+1)  # arranging plots in 2 rows and 3 columns
    plt.hist([df[df['Mental Health'] == 0][feature], df[df['Mental Health'] == 1][feature]], bins=3, label=['Bad', 'Good'], alpha=0.5)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(f"Histogram of Mental Health by {feature}")
    plt.tight_layout()
    plt.legend()

plt.tight_layout()
plt.show()
```
