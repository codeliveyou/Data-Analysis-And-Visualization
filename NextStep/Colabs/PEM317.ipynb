```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-2*np.pi,2*np.pi,1000)

# f(x) = tan x
f = np.tan(x)

# g(x) = cot x
g = 1/np.tan(x)

plt.ylim(-6,6)

#whenever the function tan x is rapidly decreasing and cot x is rapidly increasing, we make those points "not a number (nan)" to remove sharp spikes from the plot
f[:-1][np.diff(f)<0] = np.nan
g[:-1][np.diff(g)>0] = np.nan

plt.plot(x,f,label='tan x', color='purple')
plt.plot(x,g,label ='cot x', color='red')

plt.axvline(x=0,color='black',linewidth = 1)
plt.axhline(y=0,color='black',linewidth=1)

plt.grid(True)
plt.legend()
plt.show()
```
