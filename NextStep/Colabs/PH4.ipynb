```python
# import necesary libraries.
import numpy as np
import matplotlib.pyplot as plt


# parametrs
c = (0,0) # center.
w = 6 # width.
h = 3 # height.


# generate points
th = np.linspace(0,2*np.pi,50)
x = c[0] + w/2 * np.cos(th)
y = c[1] + h/2 * np.sin(th)

# Set background to black.
plt.style.use('dark_background')

# Render figure.
plt.figure

# plot elipse.
plt.plot(x,y)

# labels
plt.xlabel('x')
plt.ylabel('y')

# title
plt.title('Elipse')

# Show plot
plt.show()
```
