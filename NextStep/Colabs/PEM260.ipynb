```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_7car_data.csv")

# Group the data by year and origin and calculate the average fuel efficiency
grouped_data = df.groupby(['year', 'origin'])['fuel_efficiency'].mean().reset_index()

# Create a line chart for each origin
plt.figure(figsize=(10, 6))
origins = grouped_data['origin'].unique()
for origin in origins:
    data = grouped_data[grouped_data['origin'] == origin]
    plt.plot(data['year'], data['fuel_efficiency'], label=origin)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Average Fuel Efficiency')
plt.title('Yearly Progress of Fuel Efficiency')
plt.legend(title='Origin')

# Show the plot
plt.show()
```
