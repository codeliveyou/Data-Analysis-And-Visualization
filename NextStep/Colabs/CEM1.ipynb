```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -2pi to +2pi
x = np.linspace(-2*np.pi, 2*np.pi, 2000)

# Finding the sine and cosine values
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)',color='orange')
plt.plot(x, y_cos, label='cos(x)',color='purple')

# Adding grid and labels
plt.grid(True)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sine and cosine functions')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

#show legends
plt.legend()
plt.show()
```
