Hello. For the dataset "data_99commodity_prices.csv" I want to plot the variation of commodity prices over time on the same plot. Above are the prices of oil and coffee and below are the prices of tea and sugar. I want the background of the plot to be slategray, and the general title to be white and size 18. I want the title of each line chart to be gold. Include the code in the response.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the workout data from the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_99commodity_prices.csv")

# Set Date column as index
df.set_index('Date', inplace=True)

# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot Oil prices lines
ax = df[['Brent Crude Oil', 'Dubai Crude Oil']].plot(ax=axs[0, 0])
ax.set_title('Oil Prices', color='gold') # set subplot title
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # rotate x-labels

# Plot Coffee prices lines
ax = df[['Arabica Coffee', 'Robusta Coffee']].plot(ax=axs[0, 1])
ax.set_title('Coffee Prices',color='gold')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Plot Tea prices lines
ax = df[['Colombo Tea', 'Kolkata Tea']].plot(ax=axs[1, 0])
ax.set_title('Tea Prices', color='gold')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Plot Sugar prices lines
ax = df[['European Sugar', 'US Sugar', 'World Sugar']].plot(ax=axs[1, 1])
ax.set_title('Sugar Prices', color='gold')
plt.xticks(rotation=45)

# Adjust layout spacing
plt.tight_layout()

# Set figure title
fig.suptitle('Commodities Prices', fontsize=18, y=1.02, color='white')

# Set figure background
fig.patch.set_facecolor('slategray')

# Show the plot
plt.show()
```

Sure! Here is the code that plots the variation of commodity prices over time in the same plot. Above are the prices of oil and coffee and below are the prices of tea and sugar. The background of the plot is slategray, it has a general title in white and size 18 and the title of each line chart is gold.
