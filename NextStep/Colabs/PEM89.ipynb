```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_58mental_health.csv")


# Count "Good" and "Bad" mental health for each feature
mental_health_counts = df.groupby(['Data Availability', 'Affordable Housing', 'Educational Quality', 'Trust in Police', 'Roads and Sidewalks', 'Local Gatherings'])['Mental Health'].value_counts().unstack().fillna(0)

# Create a bar plot
mental_health_counts.plot(kind='bar', figsize = (10,5), stacked=True)
plt.title("Mental Health Counts by Different Features")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Mental Health")
plt.show()
```
