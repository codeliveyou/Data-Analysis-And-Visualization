```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data_3car_data.csv")

# Calculate average weight of cars from each country of origin
avg_weight_by_country = df.groupby("country_origin")["weight"].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
avg_weight_by_country.plot(kind="bar")
plt.title("Average Weight of Cars by Country of Origin")
plt.xlabel("Country of Origin")
plt.ylabel("Average Weight")
plt.xticks(rotation=45)
plt.tight_layout()

# Display the bar chart
plt.show()
```
