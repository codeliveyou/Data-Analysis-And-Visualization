Hello, I need to graph a yellow circle of radius 1 and center at the point (0,0). Then, within this circumference, I would like to graph 30 equilateral triangles, the first with the following coordinates (the points in red and the letters in black must be in the graph):

P = (0.0)

Q = (cos(120), sin(120))

R = (cos(240), sin(240))

The next 29 triangles must be the first triangle rotated counterclockwise with a rotation angle of 12.

In addition, the triangles must change color, in this order:
red, blue, black, purple. Repeating this order.

Finally I need to plot a circle with center (0,0) and radius 0.5 over all of the above. This last circle must be painted all red inside.

Thank you!

```python
import numpy as np
import matplotlib.pyplot as plt


#colors

colors = ['red', 'blue', 'black', 'purple']

#circle - outside
theta = np.linspace(0, 2*np.pi, 100)
x1 = np.cos(theta)
y1 = np.sin(theta)
plt.plot(x1, y1, color = 'yellow')


#circle - inside
circle = plt.Circle((0,0),0.5, fc='red',ec="red")
plt.gca().add_patch(circle)

#rotation alpha
alpha = 12

#Triangles
for n in range(30):
  x = np.array([0 + n*alpha,120 + n*alpha,240 + n*alpha,0 + n*alpha])
  y = np.array([0 + n*alpha,120 + n*alpha,240 + n*alpha,0 + n*alpha])
  x = np.cos( x * np.pi / 180 )
  y = np.sin( y * np.pi / 180 )
  plt.plot( x, y, color=colors[n%4] )

plt.text(np.cos(0), np.sin(0), 'P', fontsize = 15)
plt.plot(np.cos(0), np.sin(0), marker="o", markersize=5, markeredgecolor="red", markerfacecolor="red")
plt.text(np.cos(120* np.pi / 180), np.sin(120* np.pi / 180), 'Q', fontsize = 15)
plt.plot(np.cos(120* np.pi / 180), np.sin(120* np.pi / 180), marker="o", markersize=5, markeredgecolor="red", markerfacecolor="red")
plt.text(np.cos(240* np.pi / 180), np.sin(240* np.pi / 180), 'R', fontsize = 15)
plt.plot(np.cos(240* np.pi / 180), np.sin(240* np.pi / 180), marker="o", markersize=5, markeredgecolor="red", markerfacecolor="red")
plt.show()



```
