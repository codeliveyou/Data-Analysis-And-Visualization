```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_58mental_health.csv")

# Convert ordinal data to numerical
ordinal_mapping = {'Low': 0, 'Medium': 1, 'High': 2}
df['Educational Quality'] = df['Educational Quality'].map(ordinal_mapping)
df['Trust in Police'] = df['Trust in Police'].map(ordinal_mapping)

# Create a scatter plot
# Add offset to display the overlapping point
offset = 0.015
plt.figure(figsize=(5, 6))
plt.subplots_adjust(right=2)
plt.scatter(df['Educational Quality'], df['Mental Health'], label="Educational Quality")
plt.scatter(df['Trust in Police'] + offset, df['Mental Health'], label="Trust in Police")
plt.xlabel("Ordinal Features")
plt.ylabel("Mental Health")
plt.title("Scatter Plot of Ordinal Features vs. Mental Health")
plt.legend()
plt.show()
```
