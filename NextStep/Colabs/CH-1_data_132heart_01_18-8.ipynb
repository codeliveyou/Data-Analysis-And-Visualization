````python
#```python
# Import libraries
import pandas as pd
import numpy as np

# Load the data into a DataFrame
df = pd.read_csv('data_132heart_01.csv')

# Filter out rows where age is between 57 and 73 and heart disease is present
filtered_data = df[(df['age'] >= 57) & (df['age'] <= 73) & (df['heart disease'] == 1)]

# Create a column indicating whether an individual has tachycardia (above 100)
filtered_data['tachycardia'] = np.where(filtered_data['max heart rate'] > 100, 1, 0)

# Create a pivot table to calculate the median resting blood pressure by sex and tachycardia
pivot_table = pd.pivot_table(filtered_data, values='resting blood pressure', index=['sex', 'tachycardia'], aggfunc=np.median)

# Print the results
print("Median resting blood pressure for tachycardia (above 100) in individuals \n between the ages 57 and 73 and heart disease is present by sex:\n",
      pivot_table)
#```
````

````python
#```python
# Import libraries
import pandas as pd

# Load the data into a DataFrame
df = pd.read_csv('data_132heart_01.csv')

# Define mapping dictionaries for categorical columns
chest_pain_type_map = {0: 'asymptomatic', 1: 'atypical angina', 2: 'non-anginal pain', 3: 'typical angina'}
rest_ecg_type_map = {0: 'left ventricular hypertrophy', 1: 'normal', 2: 'ST-T wave abnormality'}

# Apply mapping to the columns
df['chest pain type'] = df['chest pain type'].map(chest_pain_type_map)
df['resting electrocardiographic results'] = df['resting electrocardiographic results'].map(rest_ecg_type_map)

# Show the modified DataFrame
df.head(3)
#```
````
