```python
import pandas as pd
import matplotlib.pyplot as plt
from adjustText import adjust_text

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_7car_data.csv")

# Code for scatter plot based on matplotlib scatter function
plt.figure(figsize=(14, 8))
scatter = plt.scatter(df['horsepower'], df['fuel_efficiency'], s=df['engine_size']*10, c=df['year'], cmap='viridis')

# Add labels and title
plt.xlabel('Horsepower')
plt.ylabel('Fuel Efficiency')
plt.title('Fuel Efficiency vs. Horsepower')
plt.colorbar(scatter, label='Year')

# Add labels to data points
for i, row in df.iterrows():
  texts = [plt.text(row['horsepower'], row['fuel_efficiency'], f"{row['manufacturer']} {row['model']}", fontsize=8) for i, row in df.iterrows()]
  adjust_text(texts)
# Show the plot
plt.show()
```
