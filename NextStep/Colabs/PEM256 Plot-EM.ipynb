```python
# importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the vector field equations
def vector_field(x, y):
    u = -y # Calculate the x-component of the vector field
    v = x  # Calculate the y-component of the vector field
    return u, v

# Generate x and y values for the grid
x_vals = np.linspace(-3, 3, 20) # Generate x values
y_vals = np.linspace(-3, 3, 20) # Generate y values
X, Y = np.meshgrid(x_vals, y_vals) # Create a grid of (x, y) values

# Calculate vector field values
U, V = vector_field(X, Y) # Calculate x-component (U) and y-component (V) of vector field

# Create a streamplot
plt.figure(figsize=(10, 8)) # Create a new figure
plt.streamplot(X, Y, U, V, color='blue', density=1.5) # Create streamplot using U and V component
plt.title('Streamplot of Vector Field') # Set plot title
plt.xlabel('x') # Label for x-axis
plt.ylabel('y') # Label for y-axis
plt.show() # Display the plot
```
