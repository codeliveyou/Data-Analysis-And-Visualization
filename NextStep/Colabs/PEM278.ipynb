```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_28store_weekly_sales.csv")

# Group data by Store and Holiday_Flag and calculate total Weekly_Sales
grouped_data = df.groupby(['Store', 'Holiday_Flag'])['Weekly_Sales'].sum().unstack()

# Create a bar plot
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('Store')
plt.ylabel('Weekly Sales')
plt.title('Holiday Sales Bar Plot')
plt.legend(title='Holiday_Flag', labels=['Non-Holiday', 'Holiday'])
plt.show()
```
