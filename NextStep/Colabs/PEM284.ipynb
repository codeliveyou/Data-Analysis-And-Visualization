```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_28store_weekly_sales.csv")

# Create a histogram with density plot
plt.figure(figsize=(10, 6))
plt.hist(df[df['Holiday_Flag'] == 0]['Unemployment'], label='Non-Holiday', alpha=0.5, density=True)
plt.hist(df[df['Holiday_Flag'] == 1]['Unemployment'], label='Holiday', alpha=0.5, density=True)
plt.xlabel('Unemployment Rate')
plt.ylabel('Density')
plt.title('Unemployment Rate Distribution')
plt.legend(title='Holiday_Flag', labels=['Non-Holiday', 'Holiday'])
plt.show()
```
