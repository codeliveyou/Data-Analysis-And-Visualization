```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_121iris.csv')

# List of columns to analyze
columns_to_analyze = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']

# Group data by 'Species'
grouped = df.groupby('Species')

# Create a figure and axes
fig, axes = plt.subplots(nrows=len(columns_to_analyze), ncols=1, figsize=(8, 12))

# Iterate through the columns and create bar plots
for idx, column in enumerate(columns_to_analyze):
    ax = axes[idx]

    # Calculate mean and standard deviation for each species
    mean_values = grouped[column].mean()
    std_values = grouped[column].std()

    # Create bar plots
    mean_values.plot(kind='bar', yerr=std_values, ax=ax, capsize=4)

    # Set labels and title
    ax.set_ylabel(column)
    ax.set_title(f'Mean and Standard Deviation of {column} by Species')
    ax.set_xticklabels(mean_values.index, rotation=45)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
