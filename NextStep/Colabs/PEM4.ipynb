```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_132heart_01.csv")

# Filter the given data for males and females
male_data = df[df['sex'] == 0]['resting blood pressure']
female_data = df[df['sex'] == 1]['resting blood pressure']

# Lets now draw a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([male_data, female_data], labels=['Male', 'Female'])
plt.xlabel('Sex')
plt.ylabel('Resting Blood Pressure')
plt.title('Dispersion of Resting Blood Pressure Among Males and Females')
plt.show()
```
