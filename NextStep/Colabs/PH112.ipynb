```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate coordinates for the triangle
triangle_radius = 1.0
triangle_center = (0, 0)
triangle_vertices = [
    (
        triangle_center[0] + triangle_radius * np.cos(2 * np.pi * i / 3),
        triangle_center[1] + triangle_radius * np.sin(2 * np.pi * i / 3)
    )
    for i in range(3)
]


# Calculate coordinates for the pentagon
pentagon_radius = triangle_radius * np.sqrt(2 + np.sqrt(5)) / 2
pentagon_center = (0, 0)
pentagon_vertices = [
    (
        pentagon_center[0] + pentagon_radius * np.cos(2 * np.pi * i / 5),
        pentagon_center[1] + pentagon_radius * np.sin(2 * np.pi * i / 5)
    )
    for i in range(5)
]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the shapes
triangle = plt.Polygon(triangle_vertices, edgecolor='blue', fill=True)
pentagon = plt.Polygon(pentagon_vertices, edgecolor='yellow', fill=True, color="cyan")

# Calculate circumcircle
circumcenter = (0, 0)
circumcircle_radius = pentagon_radius
circumcircle = plt.Circle(circumcenter, circumcircle_radius, color='red', fill=True, linestyle='dashed')

# Add shapes and circumcircle to the plot
ax.add_patch(circumcircle)
ax.add_patch(pentagon)
ax.add_patch(triangle)

# Set axis limits and aspect ratio
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_aspect('equal')

# Display the plot
plt.title('Shapes with a Common Circumcircle')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
