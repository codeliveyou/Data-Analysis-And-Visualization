```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_28store_weekly_sales.csv")

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Fuel_Price'], df['CPI'], s=df['Weekly_Sales']*0.01, c=df['Temperature'], cmap='coolwarm')
plt.colorbar(label='Temperature')
plt.xlabel('Fuel Price')
plt.ylabel('Consumer Price Index')
plt.title('Fuel Price vs. Consumer Price Index (CPI) Scatter Plot')
for i, row in df.iterrows():
    plt.text(row['Fuel_Price'], row['CPI'], row['Store'])
plt.show()
```
