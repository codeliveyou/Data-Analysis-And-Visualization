```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data_4vehicle_data.csv")

# Extract relevant columns
horsepower = df['horsepower']
acceleration = df['acceleration_in_seconds']

# Create a scatter chart
plt.figure(figsize=(10, 6))
plt.scatter(horsepower, acceleration, color='b', alpha=0.7)
plt.title('Horsepower vs Acceleration')
plt.xlabel('Horsepower')
plt.ylabel('Acceleration (in seconds)')
plt.grid(True)
plt.show()
```
