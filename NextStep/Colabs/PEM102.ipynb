```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data_4vehicle_data.csv")

# Extract relevant columns
engine_displacement = df['engine_displacement']
weight_in_pounds = df['weight_in_pounds']

# Create a scatter chart
plt.figure(figsize=(10, 6))
plt.scatter(engine_displacement, weight_in_pounds, color='b', alpha=0.7)
plt.title('Engine Displacement vs Vehicle Weight')
plt.xlabel('Engine Displacement')
plt.ylabel('Weight (in pounds)')
plt.grid(True)
plt.show()
```
