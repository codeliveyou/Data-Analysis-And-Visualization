Given the following file that contains different exercises with their duration for each part of the body, I need you to make me a bar graph where I can see for each area of the body ('Targeted Area') how many exercises of each type of duration there are (10, 20 or 30 minutes). I also want the bar corresponding to the shortest exercise to be blue, the bar corresponding to the 20-minute exercise to be orange, and the bar corresponding to the longest exercise to be red.

```python
# Import libraries
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


# Create axis
axes = [6, 6, 6]

# Create Data
data = np.ones(axes, dtype=bool)

# Control Transparency
alpha = 1

# Control colour
colors = np.empty(axes + [4], dtype=np.float32)

colors[0] = [0, 1, 1, alpha]  # sky blue
colors[1] = [0, 1, 1, alpha]  # sky blue
colors[2] = [1, 1, 1, alpha]  # white
colors[3] = [1, 1, 1, alpha]  # white
colors[4] = [0, 1, 1, alpha]  # sky blue
colors[5] = [0, 1, 1, alpha]  # sky blue



# Plot figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Voxels is used to customizations of
# the sizes, positions and colors.
ax.voxels(data, facecolors=colors, edgecolors='grey')





```
