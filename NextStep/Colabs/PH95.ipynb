```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/gdrive/MyDrive/ICE-V2-Dataset/Datasets/data_121iris.csv')

# Color dictionary
color_dict = {'Iris-setosa': "mediumslateblue", 'Iris-versicolor': "purple", 'Iris-virginica': "red"}

# Marker dictionary
marker_dict = {'Iris-setosa': 'o', 'Iris-versicolor': 'o', 'Iris-virginica': 's'}

# Create subplots for sepals and petals
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Scatterplot for sepals
ax1 = axes[0]
for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    ax1.scatter(subset['SepalWidth'], subset['SepalLength'], c=color_dict[species], label=species,
                marker=marker_dict[species], alpha=0.8)

ax1.set_xlabel('Sepal Width')
ax1.set_ylabel('Sepal Length')
ax1.set_title('Sepal Scatterplot')
ax1.legend()

# Scatterplot for petals
ax2 = axes[1]
for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    ax2.scatter(subset['PetalWidth'], subset['PetalLength'], c=color_dict[species], label=species,
                marker=marker_dict[species], alpha=0.8)

ax2.set_xlabel('Petal Width')
ax2.set_ylabel('Petal Length')
ax2.set_title('Petal Scatterplot')
ax2.legend()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
