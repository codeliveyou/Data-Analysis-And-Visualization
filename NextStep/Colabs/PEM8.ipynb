* Prompt: Graph the function sin(6t) in polar coordinates.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate t values
t = np.linspace(0, 2 * np.pi, 1000)

# Calculate r values using the function sin(6t)
r = np.sin(6 * t)

# Create a polar plot
plt.figure(figsize=(10, 10))
plt.polar(t, r, label=r"$r = \sin(6t)$")
plt.legend()
plt.title("Polar Plot of $r = \sin(6t)$")
plt.grid(True)

# Show the plot
plt.show()
```
