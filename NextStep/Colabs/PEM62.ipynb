```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy

def f(x):
  return x**2;


def g(x):
  return x**(1/3)


x = np.linspace(-1,3,1000)

plt.plot(x,f(x))
plt.plot(x,g(x))

plt.fill_between(x,f(x),g(x),where = [(x>=0) and (x<=1) for x in x], color='blue')

x = sy.Symbol('x')

# area between the curves
area_between_curves = sy.integrate(g(x)-f(x),(x,0,1))
print("Area between the curves = ",area_between_curves)
```
