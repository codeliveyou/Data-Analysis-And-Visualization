```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data_3car_data.csv")

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.bar(df['car_name'], df['engine_size'], color='blue', alpha=0.7)
plt.xlabel('Car Names')
plt.ylabel('Engine Size')
plt.title('Distribution of Engine Sizes Among Cars')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```
