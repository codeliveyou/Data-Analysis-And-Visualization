**Prompt** Hello, I would like you to graph a vector function of the helix of radius 3, in an interval of the z axis that goes from -10 to 10. The curve should be in red and the names of the axes in green. The graph must have a legend for the curve.

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Create figure size and axis
fig = plt.figure(figsize =(18, 9))
axs = plt.axes(projection ='3d')

# Create variables and their working interval
t = np.linspace(-10, 10, 400)
x = 3 * np.sin(t)
y = 3 * np.cos(t)
z = t

# Plot vectorial helix
axs.plot(x, y, z, label='Curve=(3*sin(t) , 3*cos(t) , t)', color='red')

# Set labels and legend values
label_color='green'
axs.set_xlabel('$X$', fontsize=10, color=label_color )
axs.set_ylabel('$Y$', fontsize=10, color=label_color)
axs.set_zlabel('$Z$', fontsize=10, color=label_color)
axs.legend()

# Show plot
plt.show()
```
