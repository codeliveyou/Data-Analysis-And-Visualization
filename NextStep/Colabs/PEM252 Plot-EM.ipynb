```python
# Importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f(x, y)
def f(x, y):
  # Calculate the value of the function f(x, y) for given x and y
  return np.sin(np.sqrt(x**2 + y**2))

# Generate x and y values for the grid
x_vals = np.linspace(-5, 5, 400) # Generate x values
y_vals = np.linspace(-5, 5, 400) # Generate y values
X, Y = np.meshgrid(x_vals, y_vals) # Create a grid of (x, y) values
Z = f(X, Y) # Calculate Z values using the defined function

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8)) # Create a new figure
ax = fig.add_subplot(111, projection='3d') # Add a 3D subplot
ax.plot_surface(X, Y, Z, cmap='viridis') # Plot the 3D surface
ax.set_title('3D Surface Plot of f(x, y)') # Set the plot title
ax.set_xlabel('x') # Label for x-axis
ax.set_ylabel('y') # Label for y-axis
ax.set_zlabel('f(x, y)') # Label for z-axis
plt.show() # Display the plot
```
