```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from io import StringIO

# Dataset as string
data = """sex,shell_length_mm,shell_width_mm,shell_height_mm,body_weight_g,soft_tissue_weight_g,visceral_weight_g,shell_weight_g,age_years
male,10,12,3,100,50,20,30,1
female,11,13,4,110,60,25,35,2
male,12,14,5,120,70,30,40,3
female,13,15,6,130,80,35,45,4
male,14,16,7,140,90,40,50,5
female,15,17,8,150,100,45,55,6
male,16,18,9,160,110,50,60,7
female,17,19,10,170,120,55,65,8
male,18,20,11,180,130,60,70,9
female,19,21,12,190,140,65,75,10
male,20,22,13,200,150,70,80,11
female,21,23,14,210,160,75,85,12
male,22,24,15,220,170,80,90,13
female,23,25,16,230,180,85,95,14
male,24,26,17,240,190,90,100,15
female,25,27,18,250,200,95,105,16
male,26,28,19,260,210,100,110,17
female,27,29,20,270,220,105,115,18
male,28,30,21,280,230,110,120,19
female,29,31,22,290,240,115,125,20"""

# Convert the string data to DataFrame
df = pd.read_csv(StringIO(data))

# Calculate the average body weight and soft tissue weight for male specimens
male_avg_body_weight = df[df['sex'] == 'male']['body_weight_g'].mean()
male_avg_soft_tissue_weight = df[df['sex'] == 'male']['soft_tissue_weight_g'].mean()

# Calculate the average body weight and soft tissue weight for female specimens
female_avg_body_weight = df[df['sex'] == 'female']['body_weight_g'].mean()
female_avg_soft_tissue_weight = df[df['sex'] == 'female']['soft_tissue_weight_g'].mean()

# Display the results
print(f"Average body weight for male specimens: {male_avg_body_weight:.2f}g")
print(f"Average soft tissue weight for male specimens: {male_avg_soft_tissue_weight:.2f}g")
print(f"Average body weight for female specimens: {female_avg_body_weight:.2f}g")
print(f"Average soft tissue weight for female specimens: {female_avg_soft_tissue_weight:.2f}g")

# Comparison
if male_avg_body_weight > female_avg_body_weight:
    print("Males have a higher average body weight than females.")
elif male_avg_body_weight < female_avg_body_weight:
    print("Females have a higher average body weight than males.")
else:
    print("Males and females have the same average body weight.")

if male_avg_soft_tissue_weight > female_avg_soft_tissue_weight:
    print("Males have a higher average soft tissue weight than females.")
elif male_avg_soft_tissue_weight < female_avg_soft_tissue_weight:
    print("Females have a higher average soft tissue weight than males.")
else:
    print("Males and females have the same average soft tissue weight.")
```
