I want you to draw me a flower with 18 petals with center at (1,1)

```python
# Imports
import matplotlib.pyplot as plt
import numpy as np

# Flower configuration parameters
center_x, center_y = 1, 1
petals = 18

# 1000 values equally spaced at 360 degrees
t = np.linspace(0, 2 * np.pi, 1000)

# Calculate r representing the distance from the center to the curve as a function of the angle t and the number of petals.
half_petals = petals / 2
r = np.sqrt(np.abs(np.cos(half_petals * t)))

# The coordinates are calculated
x = center_x + r * np.cos(t)
y = center_y + r * np.sin(t)

# The plot is created
plt.plot(x, y, marker="")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(f"Flower of {petals} and center ({center_x},{center_y})")
plt.grid(True)
plt.axis("equal")
plt.show()
```
