```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data_3car_data.csv")

# Filter cars with fuel efficiency greater than 30 mpg
high_efficiency_cars = df[df['fuel_efficiency'] > 30]

# Count the number of cars with fuel efficiency greater than 30 mpg
num_high_efficiency_cars = high_efficiency_cars.shape[0]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(high_efficiency_cars['car_name'], high_efficiency_cars['fuel_efficiency'], color='blue')
plt.xlabel('Car Names')
plt.ylabel('Fuel Efficiency (mpg)')
plt.title('Number of Cars with Fuel Efficiency > 30 mpg')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the bar chart
plt.show()

# Print the count of cars with fuel efficiency greater than 30 mpg
print(f'Number of cars with fuel efficiency greater than 30 mpg: {num_high_efficiency_cars}')
```
