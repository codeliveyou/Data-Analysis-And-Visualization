```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_34unemployment_rate.csv")

# Select a weekday
selected_weekday = "Monday"
selected_weekday_data = df[df['weekday'] == selected_weekday]

# Calculate the total revenue for the selected weekday
total_revenue = selected_weekday_data['revenue'].sum()

# Set up the pie chart
plt.figure(figsize=(8, 8))
plt.title(f"Revenue Distribution on {selected_weekday}")

# Create the pie chart
wedges, labels, autopcts = plt.pie(selected_weekday_data['revenue'], labels=selected_weekday_data['business_name'], autopct='%1.1f%%', startangle=90)

# Add data labels
for i, p in enumerate(wedges):
    business_name = selected_weekday_data.iloc[i]['business_name']
    revenue_percentage = (selected_weekday_data.iloc[i]['revenue'] / total_revenue) * 100
    unemployment_rate = selected_weekday_data.iloc[i]['unemployment_rate']

    data_label = f"{business_name}\n{revenue_percentage:.1f}%\nUnemployment Rate: {unemployment_rate}%"
    p.set_label(data_label)

plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.axis('equal')
plt.show()
```
