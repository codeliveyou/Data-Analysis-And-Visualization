Hello. Given the dataset "data_80shell_measurements.csv" I want to plot a 3d scatter plot on length, width and height of the shells. Let the male gender points be blueviolet colored and circular in shape. I want the female gender points to be goldenrod color and diamond shape. That the plot title be size 19, in bold brown color and italic serif style. The background of the figure should be wheat, the background of the plot should be beige and the names of the axes should be brown. Include the code in the response.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

# Read the workout data from the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_80shell_measurements.csv")

# Create figure
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111, projection = '3d')

# Filter data by gender
male = df[df['Gender'] == 'Male']
female = df[df['Gender'] == 'Female']

# Set male data
x1 = male['Shell Length']
y1 = male['Shell Width']
z1 = male['Shell Height']

# Set female data
x2 = female['Shell Length']
y2 = female['Shell Width']
z2 = female['Shell Height']

# Points with form and style requirements
ax.scatter(x1, y1, z1, c='blueviolet',
           marker='o', label='male', alpha = 0.9)
ax.scatter(x2, y2, z2, c='goldenrod',
           marker='D', label='female', alpha = 0.9)

# Set labels
ax.set_xlabel('Length', color = 'brown')
ax.set_ylabel('Width', color='brown')
ax.set_zlabel('Height', color='brown')

# Show legends
plt.legend()

# Set title
plt.title("3D Scatter Plot relating Lenght, Width and Height of Shells",
          fontsize=19, fontweight='bold', fontdict={'family': 'serif', 'style': 'italic'},
          color='brown')

# Set the plot background beige
plt.gca().set_facecolor('beige')

# Set the figure background color
fig.patch.set_facecolor('wheat')

# Adjust spacing
plt.tight_layout()

# Show the plot
plt.show()
```

Sure! Here is the code that plots a 3d scatter plot on length, width and height of the shells that makes the male gender points blueviolet colored and circular shaped and the female gender points goldenrod colored and diamond shaped. The plot title is size 19, in bold brown color and italic serif style. The background of the figure is wheat, the background of the plot is beige and the names of the axes are brown.
