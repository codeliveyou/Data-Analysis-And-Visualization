# Task CH (Calculation-H)

## Prompt

Using the data in `data_104unemployment_rate.csv`,
tell me, in general, which days have higher total sales, weekends or weekdays?

## data

data in file

## dataset

data_104unemployment_rate.csv

## Model response

Sure! In order to know which days have higher sales in general, we can compute the average for weekdays and for weekends, and compare them. The average of total sales for weekdays is 2000.00, while the average of total sales for weekends is 2333.33. This means that, in general, sales are higher during the weekends.

I hope this is helpful!

````python
#```python
#Import necessary libraries
import pandas as pd

# Load the data into a DataFrame
#df = pd.read_csv("data_104unemployment_rate.csv")

# Create a new column to separate weekdays from weekends
df['is_weekend'] = (
    (df['day_of_week'] == 'Saturday') | (df['day_of_week'] == 'Sunday')
)

# Calculate the average for each group of days
grouped_data = df.groupby('is_weekend')['total_sales'].mean().round(2).reset_index()

#print(f'The fruit with the best protein to carb ratio is {fruit_name} with a', \
#      f'ratio of {ratio:.2f}.')

weekdays_avg = grouped_data[~grouped_data['is_weekend']]
weekends_avg = grouped_data[grouped_data['is_weekend']]

weekdays_avg_sales = weekdays_avg['total_sales'].iloc[0]
weekends_avg_sales = weekends_avg['total_sales'].iloc[0]

# Print the results
print("The average of total sales for weekdays is {:.2f}".format(weekdays_avg_sales))
print("The average of total sales for weekends is {:.2f}".format(weekends_avg_sales))

if weekdays_avg_sales > weekends_avg_sales:
    print("The highest sales are in weekdays.")
else:
    print("The highest sales are in weekends.")
#```
````

# Task PH73 (Plot-H)

## Prompt

The data is in `data_104unemployment_rate.csv`. I want you to make a bar plot of price of gas and unemployment rate, but plot a separate bar for weekends and weekdays.

## data

data in file

## dataset

data_104unemployment_rate.csv

## Model response

Sure! Here is a bar plot of price of gas and unemployment rate with a different bar for weekends and weekdays:

[Image of the query]

I hope this is helpful!

````python
#```python
#Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
#df = pd.read_csv("data_104unemployment_rate.csv")

# Group by category and get averages of price of gas and unemployment rate
avg_values = df.groupby('business_day')[
                        ['price_of_gas', 'unemployment_rate']
                        ].mean(numeric_only=True).reset_index()

X = avg_values.columns[1:]
Y = avg_values[avg_values['business_day']]
Z = avg_values[~avg_values['business_day']]

X_axis = np.arange(len(X))

# Extract average values for weekdays and weekends categories
Y = avg_values[avg_values['business_day']].iloc[0, 1:]
Z = avg_values[~avg_values['business_day']].iloc[0, 1:]

# Create the bar plot
plt.bar(X_axis - 0.2, Y, 0.4, label='weekday')
plt.bar(X_axis + 0.2, Z, 0.4, label='weekend')

plt.xticks(X_axis, X)
plt.xlabel("Variable")
plt.ylabel("Average amount")
plt.title("Average values per type of business day")
plt.legend()
plt.show()
#```
````
