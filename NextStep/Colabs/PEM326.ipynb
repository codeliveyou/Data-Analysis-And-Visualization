```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the workout data from the CSV file
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_96birth_data.csv")

#filter the average_birth_weight
birth_weight = df[df['average_birth_weight'] >= 7.5]

# Configure and display the pie chart
plt.figure(figsize = (6, 6))
plt.pie(birth_weight['birth_count'], labels = birth_weight['state'], autopct = '%1.1f%%', startangle = 140 )
plt.title('Pie chart representation of birth counts in various states')
plt.axis('equal')
plt.show()


```
