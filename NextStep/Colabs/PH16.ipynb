**Prompt** I need you to draw a heliptic paraboloid of unit radius oriented in the direction of the positive x axis. The graphing interval of y and z must be between -8 and 8. I would like red labels on the axes and a title that describes the function please.

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Create vectors data source, using puter product in x and y axis, formula in z axis
scale_grid=64
y = np.outer(np.linspace(-8, 8, scale_grid), np.ones(scale_grid))
z = y.copy().T
x = y **2 + z**2

# Create color map
color_map = plt.get_cmap('plasma')

# Create figure size and axis
fig = plt.figure(figsize =(18, 9))
axs = plt.axes(projection ='3d')

# Create surface plot
surf = axs.plot_surface(x, y, z, cmap = color_map)

# Set titles and labels
color='red'
axs.set_title('Function x = y²+ z²')
axs.set_xlabel('$X$', fontsize=12, color=color )
axs.set_ylabel('$Y$', fontsize=12, color=color)
axs.set_zlabel('$Z$', fontsize=12, color=color)

# show plot
plt.show()
```
