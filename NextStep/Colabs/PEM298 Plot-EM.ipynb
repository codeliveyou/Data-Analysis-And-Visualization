```python
# Importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the function z = sin(sqrt(x^2 + y^2)) + cos(sqrt(x^2 + y^2))
def f(x, y):
  return np.sin(np.sqrt(x**2 + y**2)) + np.cos(np.sqrt(x**2 + y**2))

# Generate x and y values for the grid
x_vals = np.linspace(-5, 5, 400)  # Generate 400 evenly spaced x values from -5 to 5
y_vals = np.linspace(-5, 5, 400)  # Generate 400 evenly spaced y values from -5 to 5
X, Y = np.meshgrid(x_vals, y_vals)  # Create a grid of x and y values using meshgrid
Z = f(X, Y)  # Calculate the corresponding z values using the defined function

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))  # Create a new figure with a specified size
ax = fig.add_subplot(111, projection='3d')  # Add a 3D subplot to the figure
ax.plot_surface(X, Y, Z, cmap='viridis')  # Create a surface plot with the calculated x, y and z values
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2)) + cos(sqrt(x^2 + y^2))')  # Set the plot title
ax.set_xlabel('x')  # Label for the x-axis
ax.set_ylabel('y')  # Label for the y-axis
ax.set_zlabel('z')  # Label for the z-axis
plt.show()  # Display the plot
```
