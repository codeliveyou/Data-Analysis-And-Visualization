```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_16commodity_prices.csv")
df['date'] = pd.to_datetime(df['date'])

locations = ['colombo_tea_price', 'kolkata_tea_price', 'mombasa_tea_price']
colors = ['blue', 'green', 'orange']

# Create multi-bar plot
plt.figure(figsize=(12, 6))
for i, loc in enumerate(locations):
    plt.bar(df['date'], df[loc], width=20, color=colors[i], label=loc)

plt.xlabel('Date')
plt.ylabel('Tea Price')
plt.title('Tea Price Distribution across Regions')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
