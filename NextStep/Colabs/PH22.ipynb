**Prompt** Hello, I need you to graph a spiral in three dimensions with polar coordinates in the color lime in such a way that the theta angle runs from 0 to 10 pi and that the rho value runs from 0 to 20 units in the direction of the z axis. The graph must have a title and the axes their names.

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Create figure size and axis
fig = plt.figure(figsize =(18, 9))
axs = plt.axes(projection ='3d')

# Create variables and their working interval
theta = np.linspace(0, 10 * np.pi, 400)
z = np.linspace(0, 20, 400)
rho = z
x = rho * np.sin(theta)
y = rho * np.cos(theta)

# Plot vectorial spiral
axs.plot(x, y, z, color='lime')

# Set labels and legend values
axs.set_title('Spiral')
axs.set_xlabel('$X$', fontsize=12)
axs.set_ylabel('$Y$', fontsize=12)
axs.set_zlabel('$Z$', fontsize=12)

# Show plot
plt.show()
```
