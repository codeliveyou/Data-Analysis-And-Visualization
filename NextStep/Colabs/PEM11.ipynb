```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_34unemployment_rate.csv")

# Set up the scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot
scatter = plt.scatter(df['avg_temp'], df['gas_price'], s=df['revenue'], c=df['unemployment_rate'], cmap='coolwarm', alpha=0.9)

# Add colorbar
colorbar = plt.colorbar(scatter)
colorbar.set_label('Unemployment Rate')

# Set labels and title
plt.xlabel('Average Temperature')
plt.ylabel('Gas Price')
plt.title('Scatter Plot of Business')

# Add labels for each point
for i, row in df.iterrows():
    plt.text(row['avg_temp'], row['gas_price'], row['business_name'], fontsize=8, ha='center', va='center')

plt.grid(True)
plt.show()
```
