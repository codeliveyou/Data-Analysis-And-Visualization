Make me the necessary bar charts for the data set "data_100commodity_prices.csv" showing the average of each commodity for each day of the week.

```python
# Imports
import matplotlib.pyplot as plt

# Load dataset
try:
    data = pd.read_csv("data_100commodity_prices.csv")
except Exception as e:
    print(e)
df = data

# Cast days Date to Datetime
df["Date"] = pd.to_datetime(df["Date"])

# Configure subplots
fig, axes = plt.subplots(
    nrows=len(df.columns[1:-3]), ncols=1, figsize=(7, 2.5 * len(df.columns[1:-3]))
)

for idx, column in enumerate(df.columns[1:-3]):
    # The mean per day is calculated
    df_ = df.groupby("Date")[column].mean().reset_index(name="mean")

    # Day field creation
    df_["Day"] = df_["Date"].dt.strftime("%A")

    # Day_number field creation
    df_["Day_number"] = df_["Date"].dt.dayofweek

    # Sort by Day_number
    df_ = df_.sort_values(by="Day_number", ascending=True)

    # Add subplot
    ax = axes[idx]
    ax.bar(df_["Day"], df_["mean"], color="blue")
    ax.set_title(column)
    ax.set_xlabel("Day")
    ax.set_ylabel("Price")
    ax.grid(True)

# Plot
plt.tight_layout()
plt.show()
```
