```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data_4vehicle_data.csv")

# Calculate the average fuel economy for each vehicle
average_fuel_economy_per_vehicle = df.groupby('vehicle_name')['fuel_economy'].mean()

# Create a bar chart to visualize the average fuel economy for each vehicle
plt.figure(figsize=(12, 8))
average_fuel_economy_per_vehicle.plot(kind='bar', color='blue')
plt.ylabel('Average Fuel Economy')
plt.xlabel('Vehicle Name')
plt.title('Average Fuel Economy for Each Vehicle')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
