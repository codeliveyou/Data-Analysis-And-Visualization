```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def draw_rectangles(rectangles):
    """
    Draw a pattern of rectangles with given dimensions and display it using Matplotlib.

    Parameters:
        rectangles (list): List of rectangle side lengths.

    Returns:
        None
    """
    # Create a new figure and axis with a specified figsize
    fig, ax = plt.subplots(figsize=(25, 18))

    # Iterate through the list of rectangle side lengths
    for i, width in enumerate(rectangles):
        # Create a border rectangle around the center rectangle
        border = Rectangle((0, 0), 2 * width, 2 * width, fill=False, edgecolor="black")

        # Create the center filled rectangle
        rect = Rectangle((width, width), width, width, fill=True, color="black")

        # Add the border and center rectangles to the plot
        ax.add_patch(border)
        ax.add_patch(rect)

    # Create a center rectangle (for the smallest rectangle)
    rect = Rectangle((0, 0), width, width, fill=True, color="black")
    ax.add_patch(rect)

    # Set the x and y limits for the plot
    plt.xlim(-1, 2 * rectangles[0] + 1)
    plt.ylim(-1, 2 * rectangles[0] + 1)

    # Set the aspect ratio and turn off axis labels
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.show()

# Generate a list of rectangle dimensions (powers of 2 from 2^4 to 2^-4)
rectangle_dimensions = [2 ** l for l in range(4, -4, -1)]

# Call the function to draw the rectangles
draw_rectangles(rectangle_dimensions)
```
