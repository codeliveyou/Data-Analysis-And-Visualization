```python
from google.colab import drive

drive.mount("/content/gdrive", force_remount=True)

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset file into a pandas dataframe
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_132heart_01.csv")

# Analyze the distribution of serum cholesterol by age
plt.figure(figsize=(10, 6))
plt.hist(df["serum cholestoral"], bins=20)
plt.xlabel("Serum Cholesterol")
plt.ylabel("Frequency")
plt.title("Distribution of Serum Cholesterol by Age")
plt.show()

# Group the data by age with a range of 10
df["age group"] = pd.cut(df["age"], range(20, 100, 10), right=False)
grouped = df.groupby("age group")["serum cholestoral"].mean()
print(grouped)
```
