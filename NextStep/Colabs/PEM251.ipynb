```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/gdrive/MyDrive/Datasets/data_7car_data.csv")

# Create a boxplot
plt.figure(figsize=(10, 6))
ax = df.boxplot(column='acceleration', by='origin', patch_artist=True)

# Add labels and title
plt.xlabel('Origin')
plt.ylabel('Acceleration')
plt.title('Origin-based Boxplot of Acceleration')

# Highlight outliers
for box in ax.artists:
    box.set_facecolor('lightblue')

# Show the plot
plt.show()
```
