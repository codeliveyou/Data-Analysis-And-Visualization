```python
# import necesary libraries.
import pandas as pd
import matplotlib.pyplot as plt

# path to the file.
path = '/content/data_133exam_score_01.csv'


# Load Dataset in to pandas df.
df = pd.read_csv(path)

# Select data.
m_scre = df[df['Gender'] == 'male']['MathScore']
f_scre = df[df['Gender'] == 'female']['MathScore']

# Set plot.
plt.figure(figsize=(8,4))

# Render plot for density.
m_scre.plot(kind='density',label='male')
f_scre.plot(kind='density',label='female')

# Add title.
plt.title('Mathematics score density')

# Add labels.
plt.xlabel('Mathematics scores')
plt.ylabel('Density')
plt.legend()

# Show plot.
plt.grid() # Add grid
plt.tight_layout() # Scale plot
plt.show() # Render plot
```
