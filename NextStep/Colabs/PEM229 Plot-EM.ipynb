```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the Mandelbrot function
def mandelbrot(c, max_iter):
  # Compute the Mandelbrot value for a complex number c
  z = c
  for n in range(max_iter):
    if abs(z) > 2:
      return n
    z = z * z + c
  return max_iter

# Set parameters for Madelbort Set visualization
width, height = 800, 800 # Dimensions of the image
x_min, x_max = -2.5, 1.5 # Range of the real values
y_min, y_max = -2, 2 # Range of imaginary values

# Generate arrays of x and y values for the grid
x_vals = np.linspace(x_min, x_max, width)
y_vals = np.linspace(y_min, y_max, height)
pixels = np.zeros((width, height))  # Create an array to store Mandelbrot values

# Calculate Mandelbrot values for each pixel
for x_idx, x in enumerate(x_vals):
  for y_idx, y in enumerate(y_vals):
    c = x + y * 1j # Form a complex number using x and y
    pixels[x_idx, y_idx] = mandelbrot(c, 100) # Calculate Mandelbrot value

# Create the Mandelbrot Set plot
plt.figure(figsize=(10, 8)) # Create a new figure

# Diaplay the Mandelbrot values using logarithm for better visualization
plt.imshow(np.log(pixels.T + 1), extent=(x_min, x_max, y_min, y_max), cmap='inferno', aspect='auto')
plt.colorbar(label='Log Iterations') # Add a colorbar with label
plt.xlabel('Real') # Label for x_axis
plt.ylabel('Imaginary') # Label for y_axis
plt.title('Mandelbrot Set') # Set plot title
plt.show() # Display the plot
```
